### Your question: thin vs. full SVD — when does the third column matter?

#### The two forms

For an $m \times n$ matrix $A$ with rank $r \leq \min(m, n)$:

| Form | $U$ | $\Sigma$ | $V^T$ | Also called |
|---|---|---|---|---|
| **Thin (economy) SVD** | $m \times r$ | $r \times r$ | $r \times n$ | compact, reduced |
| **Full SVD** | $m \times m$ | $m \times n$ | $n \times n$ | — |

For your matrix: $m = 3$, $n = 2$, $r = 2$.

- **Thin SVD:** $U$ is $3 \times 2$, $\Sigma$ is $2 \times 2$, $V^T$ is $2 \times 2$. This is what you computed.
- **Full SVD:** $U$ is $3 \times 3$, $\Sigma$ is $3 \times 2$ (extra row of zeros), $V^T$ still $2 \times 2$.

Both give the same product $A$. The extra column of $U_{\text{full}}$ is multiplied by the zero row of $\Sigma_{\text{full}}$, contributing nothing to the reconstruction.

---

#### When you need the full $U$

The extra column(s) of $U$ span the **left null space** of $A$ (i.e. $\text{null}(A^T)$). You need them when:

1. **Analysing the four fundamental subspaces** — the left null space comes from the extra columns of $U$ (covered in Q2).
2. **Solving least-squares problems geometrically** — the residual $\mathbf{b} - A\hat{\mathbf{x}}$ lies in the left null space.
3. **Defining an orthonormal basis for all of $\mathbb{R}^m$** — the columns of $U_{\text{thin}}$ only span the column space of $A$.

For the purpose of this question (compute and verify $A = U\Sigma V^T$), the thin SVD is sufficient.

---

#### Finding the third column: cross product and the general method

Your instinct — take the cross product — works perfectly in $\mathbb{R}^3$:

$$\mathbf{u}_3 = \mathbf{u}_1 \times \mathbf{u}_2 = \begin{pmatrix}2/3 \\ 2/3 \\ 1/3\end{pmatrix} \times \begin{pmatrix}-1/\sqrt{2} \\ 1/\sqrt{2} \\ 0\end{pmatrix} = \frac{1}{3\sqrt{2}}\begin{pmatrix}-1 \\ -1 \\ 4\end{pmatrix}$$

Unit norm check: $\frac{1}{18}(1 + 1 + 16) = 1$ ✓

**The general method** (any dimension): **Gram-Schmidt**. Take any vector $\mathbf{w}$ not in the span of $\{\mathbf{u}_1, \mathbf{u}_2\}$, project out existing components, normalise:

$$\tilde{\mathbf{u}}_3 = \mathbf{e}_3 - (\mathbf{e}_3 \cdot \mathbf{u}_1)\mathbf{u}_1 - (\mathbf{e}_3 \cdot \mathbf{u}_2)\mathbf{u}_2, \qquad \mathbf{u}_3 = \frac{\tilde{\mathbf{u}}_3}{\|\tilde{\mathbf{u}}_3\|}$$

**Another route:** the left null space satisfies $A^T \mathbf{u}_3 = \mathbf{0}$. Solve this directly to get the third column without Gram-Schmidt.

**NumPy note:** `numpy.linalg.svd(A)` returns the full SVD by default; pass `full_matrices=False` for thin.

**Key takeaway:** The thin SVD is the right default. Extend to full when you need a basis for all of $\mathbb{R}^m$ — most commonly for the four fundamental subspaces or least-squares geometry. Cross product works in $\mathbb{R}^3$; Gram-Schmidt works generally.

```python
import numpy as np

A = np.array([[3, 1],
              [1, 3],
              [1, 1]], dtype=float)

# Verify A^T A
AtA = A.T @ A
print("A^T A:")
print(AtA)

# Eigendecomposition of A^T A (eigh guarantees real eigenvalues for symmetric matrices)
eigenvalues, V = np.linalg.eigh(AtA)
eigenvalues = eigenvalues[::-1]   # descending order
V = V[:, ::-1]
print(f"\nEigenvalues of A^T A: {eigenvalues}  (expected [18, 4])")
print(f"Singular values: {np.sqrt(eigenvalues)}  (expected [3*sqrt(2)={3*np.sqrt(2):.6f}, 2.0])")
print(f"\nV:\n{np.round(V, 6)}")

# Compute U via u_i = (1/sigma_i) A v_i
sigmas = np.sqrt(eigenvalues)
U_thin = np.column_stack([A @ V[:, i] / sigmas[i] for i in range(len(sigmas))])
print(f"\nU (thin):\n{np.round(U_thin, 6)}")
print(f"\nExpected u1 = [2/3, 2/3, 1/3]:   {[round(2/3,6), round(2/3,6), round(1/3,6)]}")
print(f"Expected u2 = [-1/sqrt2, 1/sqrt2, 0]: {[round(-1/np.sqrt(2),6), round(1/np.sqrt(2),6), 0]}")

# Verify reconstruction
A_rec = U_thin @ np.diag(sigmas) @ V.T
print(f"\nMax reconstruction error: {np.max(np.abs(A - A_rec)):.2e}")

# Full SVD via numpy
U_full, s_full, Vt_full = np.linalg.svd(A, full_matrices=True)
print(f"\n--- Full SVD ---")
print(f"U_full shape: {U_full.shape}  (vs U_thin: {U_thin.shape})")
print(f"Third column of U (left null space):\n  {np.round(U_full[:, 2], 6)}")
print(f"A^T u3 = {np.round(A.T @ U_full[:, 2], 10)}  (should be [0, 0])")

# Verify u3 via cross product
u3_cross = np.cross(U_thin[:, 0], U_thin[:, 1])
print(f"\nu3 via cross product: {np.round(u3_cross, 6)}")
print(f"  norm = {np.linalg.norm(u3_cross):.6f}  (should be 1.0)")
print(f"  u3.u1 = {u3_cross @ U_thin[:,0]:.2e},  u3.u2 = {u3_cross @ U_thin[:,1]:.2e}  (both 0)")
```
