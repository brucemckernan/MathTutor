### Your question: thin vs. full SVD — when does the third column matter?

#### The two forms

For an $m \times n$ matrix $A$ with rank $r \leq \min(m, n)$:

| Form | $U$ | $\Sigma$ | $V^T$ | Also called |
|---|---|---|---|---|
| **Thin (economy) SVD** | $m \times r$ | $r \times r$ | $r \times n$ | compact, reduced |
| **Full SVD** | $m \times m$ | $m \times n$ | $n \times n$ | — |

For your matrix: $m = 3$, $n = 2$, $r = 2$.

- **Thin SVD:** $U$ is $3 \times 2$, $\Sigma$ is $2 \times 2$, $V^T$ is $2 \times 2$. This is what you computed.
- **Full SVD:** $U$ is $3 \times 3$, $\Sigma$ is $3 \times 2$ (extra row of zeros), $V^T$ still $2 \times 2$.

Both give the same product $A$. The extra column of $U_{\text{full}}$ is multiplied by the zero row of $\Sigma_{\text{full}}$, contributing nothing to the reconstruction.

---

#### When you need the full $U$

The extra column(s) of $U$ span the **left null space** of $A$ (i.e. $\text{null}(A^T)$). You need them when:

1. **Analysing the four fundamental subspaces** — the left null space comes from the extra columns of $U$ (covered in Q2).
2. **Solving least-squares problems geometrically** — the residual $\mathbf{b} - A\hat{\mathbf{x}}$ lies in the left null space.
3. **Defining an orthonormal basis for all of $\mathbb{R}^m$** — the columns of $U_{\text{thin}}$ only span the column space of $A$.

For the purpose of this question (compute and verify $A = U\Sigma V^T$), the thin SVD is sufficient.

---

#### Finding the third column: cross product and the general method

Your instinct — take the cross product — works perfectly in $\mathbb{R}^3$:

$$\mathbf{u}_3 = \mathbf{u}_1 \times \mathbf{u}_2 = \begin{pmatrix}2/3 \\ 2/3 \\ 1/3\end{pmatrix} \times \begin{pmatrix}-1/\sqrt{2} \\ 1/\sqrt{2} \\ 0\end{pmatrix} = \frac{1}{3\sqrt{2}}\begin{pmatrix}-1 \\ -1 \\ 4\end{pmatrix}$$

Unit norm check: $\frac{1}{18}(1 + 1 + 16) = 1$ ✓

**The general method** (any dimension): **Gram-Schmidt**. Take any vector $\mathbf{w}$ not in the span of $\{\mathbf{u}_1, \mathbf{u}_2\}$, project out existing components, normalise:

$$\tilde{\mathbf{u}}_3 = \mathbf{e}_3 - (\mathbf{e}_3 \cdot \mathbf{u}_1)\mathbf{u}_1 - (\mathbf{e}_3 \cdot \mathbf{u}_2)\mathbf{u}_2, \qquad \mathbf{u}_3 = \frac{\tilde{\mathbf{u}}_3}{\|\tilde{\mathbf{u}}_3\|}$$

**Another route:** the left null space satisfies $A^T \mathbf{u}_3 = \mathbf{0}$. Solve this directly to get the third column without Gram-Schmidt.

**NumPy note:** `numpy.linalg.svd(A)` returns the full SVD by default; pass `full_matrices=False` for thin.

**Key takeaway:** The thin SVD is the right default. Extend to full when you need a basis for all of $\mathbb{R}^m$ — most commonly for the four fundamental subspaces or least-squares geometry. Cross product works in $\mathbb{R}^3$; Gram-Schmidt works generally.

````python
import numpy as np

A = np.array([[3, 1],
              [1, 3],
              [1, 1]], dtype=float)

# Verify A^T A
AtA = A.T @ A
print("A^T A:")
print(AtA)

# Eigendecomposition of A^T A (eigh guarantees real eigenvalues for symmetric matrices)
eigenvalues, V = np.linalg.eigh(AtA)
eigenvalues = eigenvalues[::-1]   # descending order
V = V[:, ::-1]
print(f"\nEigenvalues of A^T A: {eigenvalues}  (expected [18, 4])")
print(f"Singular values: {np.sqrt(eigenvalues)}  (expected [3*sqrt(2)={3*np.sqrt(2):.6f}, 2.0])")
print(f"\nV:\n{np.round(V, 6)}")

# Compute U via u_i = (1/sigma_i) A v_i
sigmas = np.sqrt(eigenvalues)
U_thin = np.column_stack([A @ V[:, i] / sigmas[i] for i in range(len(sigmas))])
print(f"\nU (thin):\n{np.round(U_thin, 6)}")
print(f"\nExpected u1 = [2/3, 2/3, 1/3]:   {[round(2/3,6), round(2/3,6), round(1/3,6)]}")
print(f"Expected u2 = [-1/sqrt2, 1/sqrt2, 0]: {[round(-1/np.sqrt(2),6), round(1/np.sqrt(2),6), 0]}")

# Verify reconstruction
A_rec = U_thin @ np.diag(sigmas) @ V.T
print(f"\nMax reconstruction error: {np.max(np.abs(A - A_rec)):.2e}")

# Full SVD via numpy
U_full, s_full, Vt_full = np.linalg.svd(A, full_matrices=True)
print(f"\n--- Full SVD ---")
print(f"U_full shape: {U_full.shape}  (vs U_thin: {U_thin.shape})")
print(f"Third column of U (left null space):\n  {np.round(U_full[:, 2], 6)}")
print(f"A^T u3 = {np.round(A.T @ U_full[:, 2], 10)}  (should be [0, 0])")

# Verify u3 via cross product
u3_cross = np.cross(U_thin[:, 0], U_thin[:, 1])
print(f"\nu3 via cross product: {np.round(u3_cross, 6)}")
print(f"  norm = {np.linalg.norm(u3_cross):.6f}  (should be 1.0)")
print(f"  u3.u1 = {u3_cross @ U_thin[:,0]:.2e},  u3.u2 = {u3_cross @ U_thin[:,1]:.2e}  (both 0)")
````

## Q2 — SVD and the four fundamental subspaces

For a matrix $A$ with SVD $A = U \Sigma V^T$, express each of the four fundamental subspaces in terms of the columns of $U$ and $V$. How does the rank of $A$ appear in the SVD?

### Feedback

**Overall: Partially correct — good instincts on rank and the column space, but two significant errors on the left null space and the claim about rank.**

---

#### Rank — ✓ Mostly correct, but one wrong claim

You correctly stated that rank equals the number of nonzero singular values, and correctly concluded this matrix has rank 2. Both right.

However, you then said "rank is the minimum of $(m, n)$." This is **not** generally true. Rank can be anything from 0 to $\min(m,n)$. When rank $= \min(m,n)$ we say the matrix has *full rank*; this is just a special case. The fact that this particular $3 \times 2$ matrix has rank 2 means it happens to be full rank — it is not a consequence of rank being defined as $\min(m,n)$.

---

#### Column space — ✓ Correct

The first $r$ columns of $U$ span the **column space** (range) of $A$. You stated this correctly.

---

#### Left null space — ✗ Incorrect

You wrote "there are no vectors in the left null space of this matrix." This is wrong.

The left null space of $A$ is $\text{null}(A^T)$, and its dimension is $m - r = 3 - 2 = 1$. There is exactly **one** left null space vector — it is the **third column of $U$** (the one you computed at the end of Q1: $\frac{1}{3\sqrt{2}}(-1, -1, 4)^T$).

The general rule is:
- First $r$ columns of $U$ → **column space** of $A$
- Last $m - r$ columns of $U$ → **left null space** of $A$ (i.e. $\text{null}(A^T)$)

You correctly noted in Q1 that the third column could be found and even computed it — so you had the right piece; you just misidentified it as a vector belonging to an empty subspace.

You also wrote the left null space vectors would be in "columns $r \leq n$ of $U$." The indices are wrong. The left null space columns are the *last* $m - r$ columns (indices $r+1$ through $m$, not the first $r$).

---

#### Row space — ✓ Correct (with a note)

The first $r$ columns of $V$ span the **row space** of $A$. You said "the rows of $V$ and therefore columns of $V^T$" — slightly roundabout. It is cleaner to say: the first $r$ **columns of $V$** span the row space (equivalently, the first $r$ rows of $V^T$).

---

#### Null space — ✗ Missing

You did not mention the **null space** of $A$ (also called the right null space). The last $n - r$ columns of $V$ span $\text{null}(A)$.

For this matrix: $n = 2$, $r = 2$, so $n - r = 0$ — the null space is trivial (only the zero vector). So nothing is missing numerically, but the conceptual description should include it.

---

#### The complete picture

Let $A = U \Sigma V^T$ with rank $r$:

| Subspace | Basis | Dimension |
|---|---|---|
| Column space $= \text{col}(A)$ | First $r$ columns of $U$ | $r$ |
| Left null space $= \text{null}(A^T)$ | Last $m - r$ columns of $U$ | $m - r$ |
| Row space $= \text{col}(A^T)$ | First $r$ columns of $V$ | $r$ |
| Null space $= \text{null}(A)$ | Last $n - r$ columns of $V$ | $n - r$ |

These four subspaces are orthogonal in pairs: column space $\perp$ left null space (both in $\mathbb{R}^m$), and row space $\perp$ null space (both in $\mathbb{R}^n$). The orthogonality comes directly from the columns of $U$ and $V$ being orthonormal — the fundamental theorem of linear algebra falls out of the SVD for free.

The rank $r$ appears as the number of nonzero diagonal entries in $\Sigma$. Each nonzero $\sigma_i$ contributes one dimension to both the column space and the row space.

#### Convention note: does B left-multiply or right-multiply?

The standard mathematical convention is that a matrix acts on **column vectors from the left**:

$$\mathbf{y} = B\mathbf{x}, \qquad \mathbf{x} \in \mathbb{R}^{2 \times 1}$$

This is the default assumption when $B$ is given without qualification.

**Why the code uses `@ B.T`.** NumPy stores collections of points as row vectors (shape `(N, 2)`). To apply $B$ to all $N$ points simultaneously, transpose the column-vector equation:

$$(B\mathbf{x})^T = \mathbf{x}^T B^T$$

So `circle @ B.T` is *not* a different convention — it is the **same** left-multiplication written out for row-vector storage. You are not right-multiplying by $B$; you right-multiply by $B^T$ precisely *because* $B$ is a left-multiplier of columns.

**Does this mean we apply the rows of $B$ to the row vectors?** Not quite. Each row of `circle @ B.T` is $\mathbf{x}^T B^T$, which is the transpose of $B\mathbf{x}$. The columns of $B^T$ (= rows of $B$) are dotted with each row-point, but only as a bookkeeping consequence of the storage layout — the underlying operation is still $B$ acting on column vectors from the left.

**Rule of thumb:** whenever NumPy code writes `X @ M.T`, read it in standard math notation as "$M$ acts on each point from the left."
