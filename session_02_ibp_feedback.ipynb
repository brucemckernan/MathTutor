## Q2 â€” Iterated IBP

**Question:** Evaluate $\int x^2 \cos x \, dx$.

### Result: Correct âś“

**First IBP â€” LIATE assignment:**

$$u = x^2 \implies du = 2x \, dx$$
$$dv = \cos x \, dx \implies v = \sin x$$

$$\int x^2 \cos x \, dx = x^2 \sin x - \int 2x \sin x \, dx$$

Correct. The residual integral $\int 2x \sin x \, dx$ is still a product of Algebraic Ă— Trig, so a second round of IBP is needed â€” you correctly recognised this.

**Second IBP â€” LIATE assignment:**

$$u = 2x \implies du = 2 \, dx$$
$$dv = \sin x \, dx \implies v = -\cos x$$

$$\int 2x \sin x \, dx = 2x(-\cos x) - \int (-\cos x)(2) \, dx = -2x \cos x + 2\sin x$$

Substituting back:

$$\int x^2 \cos x \, dx = x^2 \sin x - (-2x \cos x + 2\sin x) = x^2 \sin x + 2x \cos x - 2\sin x + C$$

Your answer matches this exactly.

### One small thing to note

You dropped the $+ C$ in your final line. It won't cost you marks in this context, but make a habit of including it â€” especially once we reach definite integrals and improper integrals where boundary terms matter.

### Why iterated IBP terminates here

Each round of IBP reduces the power of $x$ by 1: $x^2 \to 2x \to 2$ (a constant). Once the algebraic factor is a constant, the remaining integral ($\int 2\cos x \, dx = 2\sin x$) is immediate. A polynomial of degree $n$ paired with $\sin x$ or $\cos x$ will always require exactly $n$ rounds. This is a reliable pattern to remember.

```python
from sympy import *
x = symbols('x')
result = integrate(x**2 * cos(x), x)
print(result)  # should be x**2*sin(x) + 2*x*cos(x) - 2*sin(x)
```

## Q3 â€” Cyclic IBP

**Question:** Evaluate $\int e^x \sin x \, dx$.

### Result: Almost â€” one sign error, correct technique âś“

Your overall approach is exactly right: apply IBP twice, notice the original integral reappears, then solve algebraically. That is the cyclic IBP pattern. The error is a single wrong derivative at the very first step, and it propagates through to give the wrong sign in the final answer.

---

### Where the error is

You wrote $u = \sin x$ and then $du = -\cos x \, dx$.

The derivative of $\sin x$ is $+\cos x$, not $-\cos x$. ($-\cos x$ is the derivative of $\cos x$.)

$$u = \sin x \implies du = \cos x \, dx \quad \checkmark$$

---

### Correct working

**First IBP** â€” LIATE: Trig ahead of Exponential, so $u = \sin x$:

$$u = \sin x, \quad du = \cos x \, dx$$
$$dv = e^x \, dx, \quad v = e^x$$

$$\int e^x \sin x \, dx = e^x \sin x - \int e^x \cos x \, dx$$

The residual integral is still a product of two functions of $x$, so apply IBP again. Assign $u = \cos x$ to keep Trig ahead of Exponential:

**Second IBP:**

$$u = \cos x, \quad du = -\sin x \, dx$$
$$dv = e^x \, dx, \quad v = e^x$$

$$\int e^x \cos x \, dx = e^x \cos x - \int e^x (-\sin x) \, dx = e^x \cos x + \int e^x \sin x \, dx$$

**Substitute back:**

$$\int e^x \sin x \, dx = e^x \sin x - \left(e^x \cos x + \int e^x \sin x \, dx\right)$$

$$\int e^x \sin x \, dx = e^x \sin x - e^x \cos x - \int e^x \sin x \, dx$$

**Recognise the cycle and solve algebraically** â€” add $\int e^x \sin x \, dx$ to both sides:

$$2\int e^x \sin x \, dx = e^x(\sin x - \cos x)$$

$$\boxed{\int e^x \sin x \, dx = \frac{e^x(\sin x - \cos x)}{2} + C}$$

---

### What your answer produced

With $du = -\cos x \, dx$, the first IBP gave $e^x \sin x + \int e^x \cos x \, dx$ (a $+$ where there should be a $-$). The subsequent algebra was correct given that line, but the sign propagated through to give $\frac{e^x(\sin x + \cos x)}{2}$ instead of $\frac{e^x(\sin x - \cos x)}{2}$.

---

### Why either LIATE assignment works here (but must stay consistent)

In the cyclic case, choosing $u = e^x$ (Exponential) instead of $u = \sin x$ also works â€” you get a different intermediate expression, but the same final answer. The critical rule is that **both** applications of IBP must assign $u$ to the same "type" of factor. If you start with $u = \sin x$, the second round must use $u = \cos x$ (Trig again). If you flip mid-way and choose $u = e^x$ in the second round, you undo the first step and get $0 = 0$ â€” no information.

---

### Summary

| $du$ for $u = \sin x$ | $-\cos x \, dx$ âś— | $\cos x \, dx$ âś“ |
| Cyclic recognition | âś“ | âś“ |

| Final answer | $\frac{e^x(\sin x + \cos x)}{2}$ âś— | $\frac{e^x(\sin x - \cos x)}{2}$ âś“ |

```python
from sympy import *
x = symbols('x')
result = integrate(exp(x) * sin(x), x)
print(result)  # should be exp(x)*(sin(x) - cos(x))/2
```

## Q4 â€” The invisible $dv$

**Question:** Evaluate $\int \ln x \, dx$.

### Result: Correct âś“

Perfect execution of the invisible-$dv$ trick. You identified that there is no direct antiderivative for $\ln x$, spotted the multiplication-by-1 device, and carried the IBP through cleanly.

**LIATE assignment:**

$$u = \ln x \implies du = \frac{1}{x} \, dx$$
$$dv = 1 \, dx \implies v = x$$

**IBP formula:**

$$\int \ln x \, dx = x \ln x - \int x \cdot \frac{1}{x} \, dx = x \ln x - \int 1 \, dx = x \ln x - x + C$$

$$\boxed{\int \ln x \, dx = x(\ln x - 1) + C}$$

Your factored form $x(\ln x - 1) + C$ is correct and the conventional way to write it.

---

### Why this works

$\ln x$ sits at the top of LIATE (Logarithmic â€” highest priority for $u$). The problem is that there appears to be nothing left for $dv$. The trick is to write the integrand as $\ln x \cdot 1$ and assign $dv = 1 \, dx$, so $v = x$. This is always valid: every integrand has an implicit factor of 1.

The reason this assignment is productive: differentiating $\ln x$ gives $\frac{1}{x}$, a simpler function. Multiplying $v = x$ by $\frac{1}{x}$ in the residual integral $\int v \, du$ cancels the $x$, leaving $\int 1 \, dx$ â€” trivially integrable. The same device works for $\int \arctan x \, dx$, $\int \arcsin x \, dx$, and similar lone-function integrals.

```python
from sympy import *
x = symbols('x')
result = integrate(ln(x), x)
print(result)  # should be x*log(x) - x
```

## Q5 â€” Improper integral with IBP

**Question:** Evaluate $\displaystyle\int_0^\infty x e^{-x} \, dx$, or explain why it diverges.

### Result: Correct answer, but two sign errors that cancel âš ď¸Ź

Your final answer of **1 is correct**, and the high-level method is exactly right: introduce a limit to handle the infinite upper bound, apply IBP with LIATE, then evaluate each piece using L'HĂ´pital. However, the antiderivative of $e^{-x}$ is wrong at two points in your working, and the errors happen to cancel. That kind of lucky cancellation is worth catching.

---

### The sign error

You wrote: *"$dv = e^{-x}\,dx$ and $v = e^{-x}$"*

The antiderivative of $e^{-x}$ is $-e^{-x}$, not $e^{-x}$. (The chain rule: $\frac{d}{dx}(-e^{-x}) = -e^{-x} \cdot (-1) = e^{-x}$.)

$$dv = e^{-x}\,dx \implies v = -e^{-x}$$

The same error reappears when you evaluated $\int e^{-x}\,dx = e^{-x}$; it should be $-e^{-x}$.

---

### Correct working

**IBP** with $u = x$, $du = dx$, $dv = e^{-x}\,dx$, $v = -e^{-x}$:

$$\int x e^{-x}\,dx = x(-e^{-x}) - \int (-e^{-x})\,dx = -xe^{-x} + \int e^{-x}\,dx = -xe^{-x} - e^{-x} + C = -(x+1)e^{-x} + C$$

**Improper integral:**

$$\lim_{b\to\infty}\Bigl[-(x+1)e^{-x}\Bigr]_0^b = \lim_{b\to\infty}\Bigl[-(b+1)e^{-b}\Bigr] - \bigl[-(0+1)e^0\bigr] = \lim_{b\to\infty}\bigl[-(b+1)e^{-b}\bigr] + 1$$

As $b\to\infty$, $(b+1)e^{-b} \to 0$ by L'HĂ´pital (identical reasoning to yours):

$$\frac{b+1}{e^b} \xrightarrow{\text{L'H}} \frac{1}{e^b} \to 0$$

So the integral $= 0 + 1 = \boxed{1}$.

---

### Why your errors cancelled

Your intermediate expression was $\bigl[xe^{-x} - e^{-x}\bigr]_0^b$, giving $(be^{-b} - e^{-b}) - (0 - 1)$. The correct expression is $\bigl[-(x+1)e^{-x}\bigr]_0^b$, giving $-(b+1)e^{-b} - (-1)$. Both expressions have $xe^{-x}$ terms that vanish as $b\to\infty$, so both give 1 despite the sign difference. The final answer survives because L'HĂ´pital kills the problematic terms regardless of their sign.
