## Q1 — Changing order of integration

Evaluate $\displaystyle\int_0^1 \int_y^1 e^{x^2} \, dx \, dy$ by switching the order of integration.

### Feedback

**Correct throughout. Clean geometric reasoning and the u-substitution is executed well.**

---

### Region identification — ✓

You correctly identified the region. The original limits say:
- outer integral: $0 \leq y \leq 1$
- inner integral: $y \leq x \leq 1$

This is the triangle with vertices $(0,0)$, $(1,0)$, $(1,1)$, bounded by three edges: the diagonal $y = x$, the vertical line $x = 1$, and the horizontal line $y = 0$. You identified all three boundaries correctly (via the sketch). ✓

---

### Switching the order — ✓

Describing the same triangle with $x$ as the outer variable:
- $x$ runs from $0$ to $1$ (same constants)
- for each fixed $x$, $y$ runs from $0$ to $x$ (below the diagonal, above the x-axis)

$$\int_0^1 \int_0^x e^{x^2} \, dy \, dx \qquad \checkmark$$

This is the key step. Once the $x$ is on the outside, the integrand $e^{x^2}$ is constant with respect to $y$, so the inner integral is just $e^{x^2}$ times the length of the $y$-interval.

---

### Evaluating the inner integral — ✓

$$\int_0^x e^{x^2} \, dy = e^{x^2} \big[y\big]_0^x = x e^{x^2} \qquad \checkmark$$

The factor of $x$ emerges naturally here from the upper limit $y = x$. It is not introduced artificially — it is the width of the $y$-interval at a given $x$, and it is precisely what makes the outer integral tractable by substitution.

---

### u-substitution — ✓

$$\int_0^1 x e^{x^2} \, dx$$

Setting $u = x^2$, $du = 2x \, dx$, so $x \, dx = \frac{du}{2}$. Bounds: $x=0 \Rightarrow u=0$; $x=1 \Rightarrow u=1$. ✓

$$\frac{1}{2}\int_0^1 e^u \, du = \frac{1}{2}\Big[e^u\Big]_0^1 = \frac{e - 1}{2} \approx 0.8591 \qquad \checkmark$$

---

### Two small presentation points

**1. Stray $dy\,dx$ after the inner integral is evaluated.**

In the line
$$\int_{x=0}^{x=1} \left[e^{x^2} y\right]_{y=0}^{y=x} \, dy\,dx$$
the $dy$ should not appear — you have already integrated with respect to $y$ and are about to evaluate the bracket. The line should read:
$$\int_0^1 \left[e^{x^2} y\right]_0^x dx = \int_0^1 x e^{x^2} \, dx$$

**2. The $x$'s in the u-sub line.**

You wrote $\int e^u \cdot x \cdot \frac{du}{2x}$, which is fine — the $x$'s do cancel. It is slightly cleaner to cancel them in the same line before writing the integral:
$$x \, dx = \frac{du}{2} \implies \int_0^1 x e^{x^2} \, dx = \frac{1}{2}\int_0^1 e^u \, du$$

Neither of these affects the logic or the answer.

---

### Why switching the order works here

It is worth pausing on why $e^{x^2}$ has no elementary antiderivative in $x$ but becomes easy once we have $x e^{x^2}$. The function $e^{x^2}$ does not pair with a $dx$ alone to give a recognisable derivative. But $x e^{x^2}\,dx = \frac{1}{2}d(e^{x^2})$ — it is the exact derivative of $\frac{1}{2}e^{x^2}$. Switching the order of integration manufactured the extra $x$ factor that made the substitution work. Recognising this pattern — "I need an $x$ in front of $e^{x^2}$, and reversing the order gives me one" — is the diagnostic that tells you to switch order rather than attempt the original integral.

---

### Summary

| Step | Mark | Note |
|---|---|---|
| Region correctly identified as triangle | ✓ | |
| Three bounding edges stated | ✓ | $y=x$, $x=1$, $y=0$ |
| Switched limits: $x$ from 0 to 1, $y$ from 0 to $x$ | ✓ | |
| Inner integral evaluated to $x e^{x^2}$ | ✓ | |
| u-sub $u = x^2$ set up | ✓ | |
| $x\,dx = du/2$, bounds unchanged $[0,1]$ | ✓ | |
| Final answer $\frac{e-1}{2}$ | ✓ | |
| Stray $dy\,dx$ after bracket evaluation | — | Minor notation only |

```python
import numpy as np
from scipy import integrate

# Verify by numerical integration in the original order
inner = lambda x, y: np.exp(x**2)
result_original, _ = integrate.dblquad(inner, 0, 1, lambda y: y, lambda y: 1)

# And in the switched order
inner_switched = lambda y, x: np.exp(x**2)
result_switched, _ = integrate.dblquad(inner_switched, 0, 1, 0, lambda x: x)

exact = (np.e - 1) / 2

print(f"Original order (numerical):  {result_original:.10f}")
print(f"Switched order (numerical):  {result_switched:.10f}")
print(f"Exact value (e-1)/2:         {exact:.10f}")
print(f"e = {np.e:.6f},  (e-1)/2 = {exact:.6f}")
```
