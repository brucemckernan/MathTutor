## Q1 — Changing order of integration

Evaluate $\displaystyle\int_0^1 \int_y^1 e^{x^2} \, dx \, dy$ by switching the order of integration.

### Feedback

**Correct throughout. Clean geometric reasoning and the u-substitution is executed well.**

---

### Region identification — ✓

You correctly identified the region. The original limits say:
- outer integral: $0 \leq y \leq 1$
- inner integral: $y \leq x \leq 1$

This is the triangle with vertices $(0,0)$, $(1,0)$, $(1,1)$, bounded by three edges: the diagonal $y = x$, the vertical line $x = 1$, and the horizontal line $y = 0$. You identified all three boundaries correctly (via the sketch). ✓

---

### Switching the order — ✓

Describing the same triangle with $x$ as the outer variable:
- $x$ runs from $0$ to $1$ (same constants)
- for each fixed $x$, $y$ runs from $0$ to $x$ (below the diagonal, above the x-axis)

$$\int_0^1 \int_0^x e^{x^2} \, dy \, dx \qquad \checkmark$$

This is the key step. Once the $x$ is on the outside, the integrand $e^{x^2}$ is constant with respect to $y$, so the inner integral is just $e^{x^2}$ times the length of the $y$-interval.

---

### Evaluating the inner integral — ✓

$$\int_0^x e^{x^2} \, dy = e^{x^2} \big[y\big]_0^x = x e^{x^2} \qquad \checkmark$$

The factor of $x$ emerges naturally here from the upper limit $y = x$. It is not introduced artificially — it is the width of the $y$-interval at a given $x$, and it is precisely what makes the outer integral tractable by substitution.

---

### u-substitution — ✓

$$\int_0^1 x e^{x^2} \, dx$$

Setting $u = x^2$, $du = 2x \, dx$, so $x \, dx = \frac{du}{2}$. Bounds: $x=0 \Rightarrow u=0$; $x=1 \Rightarrow u=1$. ✓

$$\frac{1}{2}\int_0^1 e^u \, du = \frac{1}{2}\Big[e^u\Big]_0^1 = \frac{e - 1}{2} \approx 0.8591 \qquad \checkmark$$

---

### Two small presentation points

**1. Stray $dy\,dx$ after the inner integral is evaluated.**

In the line
$$\int_{x=0}^{x=1} \left[e^{x^2} y\right]_{y=0}^{y=x} \, dy\,dx$$
the $dy$ should not appear — you have already integrated with respect to $y$ and are about to evaluate the bracket. The line should read:
$$\int_0^1 \left[e^{x^2} y\right]_0^x dx = \int_0^1 x e^{x^2} \, dx$$

**2. The $x$'s in the u-sub line.**

You wrote $\int e^u \cdot x \cdot \frac{du}{2x}$, which is fine — the $x$'s do cancel. It is slightly cleaner to cancel them in the same line before writing the integral:
$$x \, dx = \frac{du}{2} \implies \int_0^1 x e^{x^2} \, dx = \frac{1}{2}\int_0^1 e^u \, du$$

Neither of these affects the logic or the answer.

---

### Why switching the order works here

It is worth pausing on why $e^{x^2}$ has no elementary antiderivative in $x$ but becomes easy once we have $x e^{x^2}$. The function $e^{x^2}$ does not pair with a $dx$ alone to give a recognisable derivative. But $x e^{x^2}\,dx = \frac{1}{2}d(e^{x^2})$ — it is the exact derivative of $\frac{1}{2}e^{x^2}$. Switching the order of integration manufactured the extra $x$ factor that made the substitution work. Recognising this pattern — "I need an $x$ in front of $e^{x^2}$, and reversing the order gives me one" — is the diagnostic that tells you to switch order rather than attempt the original integral.

---

### Summary

| Step | Mark | Note |
|---|---|---|
| Region correctly identified as triangle | ✓ | |
| Three bounding edges stated | ✓ | $y=x$, $x=1$, $y=0$ |
| Switched limits: $x$ from 0 to 1, $y$ from 0 to $x$ | ✓ | |
| Inner integral evaluated to $x e^{x^2}$ | ✓ | |
| u-sub $u = x^2$ set up | ✓ | |
| $x\,dx = du/2$, bounds unchanged $[0,1]$ | ✓ | |
| Final answer $\frac{e-1}{2}$ | ✓ | |
| Stray $dy\,dx$ after bracket evaluation | — | Minor notation only |

```python
import numpy as np
from scipy import integrate

# Verify by numerical integration in the original order
inner = lambda x, y: np.exp(x**2)
result_original, _ = integrate.dblquad(inner, 0, 1, lambda y: y, lambda y: 1)

# And in the switched order
inner_switched = lambda y, x: np.exp(x**2)
result_switched, _ = integrate.dblquad(inner_switched, 0, 1, 0, lambda x: x)

exact = (np.e - 1) / 2

print(f"Original order (numerical):  {result_original:.10f}")
print(f"Switched order (numerical):  {result_switched:.10f}")
print(f"Exact value (e-1)/2:         {exact:.10f}")
print(f"e = {np.e:.6f},  (e-1)/2 = {exact:.6f}")
```

## Q2 — Double integral over a general region

$\displaystyle\iint_R xy \, dA$ where $R$ is the region bounded by $y = x$ and $y = x^2$.

### Feedback

**Correct throughout. Clean setup, correct inner evaluation, correct outer integral.**

---

### Intersection points — ✓

Setting $y = x$ equal to $y = x^2$:

$$x = x^2 \implies x^2 - x = 0 \implies x(x-1) = 0 \implies x = 0 \text{ or } x = 1$$

Intersection points $(0,0)$ and $(1,1)$. ✓

On $(0,1)$, $x > x^2$ (e.g. at $x = \frac{1}{2}$: $\frac{1}{2} > \frac{1}{4}$), so $y = x$ lies above $y = x^2$ throughout the region. ✓

#### One word to fix: "bounding rectangle"

You described the region as "creating a bounding rectangle". The region is not a rectangle — it is the curvilinear area between the two curves. The $[0,1] \times [0,1]$ square is a bounding *box* that contains the region, but the region itself is a curved sliver. The correct phrasing is simply: the region bounded above by $y = x$ and below by $y = x^2$, for $0 \leq x \leq 1$.

---

### Limits and order of integration — ✓

Integrating $y$ first (inner), $x$ second (outer):

$$\int_0^1 \int_{x^2}^{x} xy \, dy \, dx$$

- Outer: $x$ from $0$ to $1$ (the $x$-extent of the region) ✓
- Inner: $y$ from $x^2$ (lower curve) to $x$ (upper curve) ✓

Your reasoning for this order — that $y = x^2$ as a $y$-limit is cleaner than $x = \sqrt{y}$ — is correct and worth noting as a general principle: when a boundary is a function of the outer variable, the algebra stays polynomial.

---

### Inner integral — ✓

$$\int_{x^2}^{x} xy \, dy = x \left[\frac{y^2}{2}\right]_{x^2}^{x} = \frac{x}{2}\left(x^2 - x^4\right) = \frac{1}{2}\left(x^3 - x^5\right) \qquad \checkmark$$

---

### Outer integral — ✓

$$\frac{1}{2}\int_0^1 \left(x^3 - x^5\right) dx = \frac{1}{2}\left[\frac{x^4}{4} - \frac{x^6}{6}\right]_0^1 = \frac{1}{2}\left(\frac{1}{4} - \frac{1}{6}\right) = \frac{1}{2} \cdot \frac{2}{24} = \frac{1}{24} \qquad \checkmark$$

---

### Minor LaTeX note

In your notebook the closing bracket is written `\right]_{x=0}^{x=1$$` — a missing `}` before the final `$$`. No mathematical issue, just a typo to fix.

---

### Summary

| Step | Mark | Note |
|---|---|---|
| Intersection points $(0,0)$ and $(1,1)$ | ✓ | |
| $y = x$ above $y = x^2$ on $(0,1)$ confirmed | ✓ | |
| Limits set up correctly | ✓ | |
| Inner integral: $\frac{1}{2}(x^3 - x^5)$ | ✓ | |
| Outer integral: $\frac{1}{2}[\frac{x^4}{4} - \frac{x^6}{6}]_0^1$ | ✓ | |
| Final answer $\frac{1}{24}$ | ✓ | |
| "Bounding rectangle" phrasing | — | Region is a curved area, not a rectangle |

```python
import numpy as np
from scipy import integrate

# Verify numerically
f = lambda y, x: x * y
result, _ = integrate.dblquad(f, 0, 1, lambda x: x**2, lambda x: x)

print(f"Numerical result:  {result:.10f}")
print(f"Exact value 1/24:  {1/24:.10f}")
print(f"Match: {np.isclose(result, 1/24)}")
```

## Q3 — Polar coordinates

$\displaystyle\iint_R (x^2 + y^2) \, dA$ where $R$ is the disk $x^2 + y^2 \leq 4$.

### Feedback

**Correct throughout. The Jacobian is handled properly — the most common point of failure on this type of question.**

---

### Polar substitution — ✓

Using $r^2 = x^2 + y^2$, the integrand $x^2 + y^2$ becomes simply $r^2$. ✓

---

### Limits — ✓

The disk $x^2 + y^2 \leq 4$ has $r^2 \leq 4$, so $0 \leq r \leq 2$. ✓

Full disk centred at the origin: $0 \leq \theta \leq 2\pi$. ✓

---

### Jacobian — ✓

You correctly wrote $dA = r \, dr \, d\theta$, giving the integrand $r^2 \cdot r = r^3$. This is the single most important step in any polar coordinate calculation. Many errors come from writing $dA = dr \, d\theta$ and losing the factor of $r$. You got it right. ✓

---

### Evaluation — ✓

$$\int_0^{2\pi}\int_0^2 r^3 \, dr \, d\theta = \int_0^{2\pi} \left[\frac{r^4}{4}\right]_0^2 d\theta = \int_0^{2\pi} \frac{16}{4} \, d\theta = \int_0^{2\pi} 4 \, d\theta = \Big[4\theta\Big]_0^{2\pi} = 8\pi \qquad \checkmark$$

---

### One point to sharpen: where does the $r$ come from?

You described the $r$ in $r \, dr \, d\theta$ as "the inner integral of a polar coordinate integral which is always $r$". This phrasing slightly misrepresents what the $r$ is — it could suggest $r$ is a limit of integration rather than part of the area element.

The correct explanation: when you change variables from $(x, y)$ to $(r, \theta)$, the area element $dA$ transforms as

$$dA = dx \, dy = \left|\frac{\partial(x,y)}{\partial(r,\theta)}\right| dr \, d\theta = r \, dr \, d\theta$$

The factor $r$ is the **Jacobian determinant** of the polar coordinate transformation $x = r\cos\theta$, $y = r\sin\theta$. It accounts for the fact that a thin ring at radius $r$ has circumference $2\pi r$ — rings further from the origin are larger, so they carry more area per unit $dr$. The $r$ is not part of the integrand or the limits; it is part of the area measure.

This matters because in more general coordinate changes (cylindrical, spherical) you need to compute the Jacobian fresh each time rather than relying on a rule of thumb. Knowing *why* $r$ appears makes those transitions straightforward.

| Coordinates | Area/Volume element |
|---|---|
| Cartesian $(x,y)$ | $dx \, dy$ |
| Polar $(r,\theta)$ | $r \, dr \, d\theta$ |
| Cylindrical $(r,\theta,z)$ | $r \, dr \, d\theta \, dz$ |
| Spherical $(\rho,\phi,\theta)$ | $\rho^2 \sin\phi \, d\rho \, d\phi \, d\theta$ |

---

### Geometric interpretation — ✓

Your description — "the volume under the parabolic surface above the $xy$-plane bounded by the circular region" — is geometrically correct. The integral $\iint_R (x^2+y^2)\,dA$ is indeed the volume under the surface $z = x^2 + y^2$ (a **paraboloid of revolution**) over the disk $r \leq 2$. The more precise term is *paraboloid* rather than *parabolic surface*, since "parabolic" more commonly refers to a 2D curve.

---

### Summary

| Step | Mark | Note |
|---|---|---|
| $x^2 + y^2 \to r^2$ | ✓ | |
| $r$ bounds: 0 to 2 | ✓ | |
| $\theta$ bounds: 0 to $2\pi$ | ✓ | |
| Jacobian $r$ included: $dA = r\,dr\,d\theta$ | ✓ | Most common error point — handled correctly |
| Inner integral: $[r^4/4]_0^2 = 4$ | ✓ | |
| Outer integral: $4 \cdot 2\pi = 8\pi$ | ✓ | |
| Description of $r$ as Jacobian vs. limit | — | Worth knowing the reason, not just the rule |

```python
import numpy as np
from scipy import integrate

# Verify numerically in Cartesian
f = lambda y, x: x**2 + y**2
result, _ = integrate.dblquad(f, -2, 2,
                               lambda x: -np.sqrt(4 - x**2),
                               lambda x:  np.sqrt(4 - x**2))

print(f"Numerical (Cartesian):  {result:.8f}")
print(f"Exact value 8π:         {8 * np.pi:.8f}")
print(f"Match: {np.isclose(result, 8 * np.pi)}")

# Also show the Jacobian explicitly
print(f"\nJacobian of polar transform:")
print(f"  x = r cos θ,  y = r sin θ")
print(f"  ∂(x,y)/∂(r,θ) = |cos θ  -r sin θ| = r cos²θ + r sin²θ = r")
print(f"                   |sin θ   r cos θ|")
```
