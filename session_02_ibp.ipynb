## Q3 — Cyclic IBP

Evaluate $\int e^x \sin x \, dx$.

*Hint: Apply IBP twice and think about what you notice.*

$\int\ e^x\ sin\ x\ dx$ is product of two functions of x therfore use IBP

Prioritise Trig over Exponent in the choice of u:

$u\ =\ sin\ x$ and therefore $du\ =\ -\ cos\ x\ dx$

$dv\ = e^x$ and therefore $v\ =\ e^x$

Use the IBP formula gives:

$\int\ e^x\ sin\ x\ dx\ =\ sin\ x\ e^x\ -\ \int\ e^x(-\ cos\ x) dx$

We still have an integral on the right hand side that is the product of functions of x therefore use IBP again:


$u\ =\ -\ cos\ x$ and therefore $du\ =\ sin\ x\ dx$

$dv\ = e^x$ and therefore $v\ =\ e^x$

$\int\ e^x\ sin\ x\ dx\ =\ sin\ x\ e^x\ -\ (-\ cos\ x)e^x\ -\ \int\ e^x\ sin\ x\ dx$

We now see $\int\ e^x\ sin\ x\ dx$ on both sides. If we add this expression to both sides giving us $2\ \int\ e^x\ sin\ x\ dx$ on th rights hand side and then divide both sides by 2 we can arrive at:

$\int\ e^x\ sin\ x\ dx\ =\ \frac{sin\ x\ e^x\ +\ cos\ x\ e^x}{2}$
