### B3 — Python: Power Iteration

Implement **power iteration** from scratch (do not use `numpy.linalg.eig` for the iteration itself) to find the dominant eigenvalue and eigenvector of:

$$B = \begin{pmatrix} 4 & 1 \\ 2 & 3 \end{pmatrix}$$

**Requirements:**
- Start from $v_0 = (1, 1)^T$
- Normalise $v$ at each step (use the Euclidean norm)
- Run for 20 iterations
- Print the eigenvalue estimate and eigenvector at each iteration
- After the loop, verify your result using `numpy.linalg.eig`

*The eigenvalue estimate at step $k$ is $\lambda_k = v_k^T B v_k$ (the Rayleigh quotient).*

```python
import numpy as np
import math

B = np.array([[4, 1],
              [2, 3]], dtype=float)

vec1 = np.array([3, 1])

def EucNorm(v1):
    return math.sqrt(np.dot(v1, v1.T))

def Rayleigh(v1):
    return v1.dot(B @ v1.T) / EucNorm(B @ v1.T)

for i in range(20):
    vec1 = B @ vec1.T / EucNorm(vec1)
    print(f"eigenvalue estimate for step {i} is {Rayleigh(vec1)} from vector: {vec1}")

print(f"Final eigenvalue estimate is {Rayleigh(vec1)}")

print(f"eigenvalue from numpy.linalg.eig: {np.linalg.eig(B)}")
```
