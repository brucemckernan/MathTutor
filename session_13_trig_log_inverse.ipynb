## Section A — Trigonometric Substitution

## Q1

Evaluate $\displaystyle\int \frac{dx}{\sqrt{x^2 - 9}}$.

This is the **third trig-substitution pattern** — the one not covered in Session 8. The three patterns and their substitutions are:

| Form | Substitution | Identity |
|---|---|---|
| $\sqrt{a^2 - x^2}$ | $x = a\sin\theta$ | $1 - \sin^2\theta = \cos^2\theta$ |
| $\sqrt{a^2 + x^2}$ | $x = a\tan\theta$ | $1 + \tan^2\theta = \sec^2\theta$ |
| $\sqrt{x^2 - a^2}$ | $x = a\sec\theta$ | $\sec^2\theta - 1 = \tan^2\theta$ |
