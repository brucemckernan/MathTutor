**Question:**
$$\lim_{x \to 0} \frac{\sin(3x) - 3x}{x^3}$$

### Verdict

The limit **does exist** and equals $-\dfrac{9}{2}$. The key issue in your answer is that identifying a $\frac{0}{0}$ indeterminate form is correct, but the conclusions drawn from the sign analysis are not valid here — more on that below.

### What you got right

- Correctly recognised that simple factoring doesn't immediately simplify things.

### The sign analysis issue

You evaluated the function at $x = -1$ and $x = 1$ and observed opposite signs, leading you to conclude there is a vertical asymptote. There are two problems with this:

1. **The test points are too far from 0.** Sign analysis at $x = \pm 1$ tells you the sign of the function at those specific points — it says nothing about what happens as $x \to 0^\pm$. The function could do anything in between.

2. **Opposite-sided signs do not imply a vertical asymptote.** A function can pass through zero, or approach a finite limit, and still have opposite signs on either side. The classic example is $f(x) = x$: negative for $x < 0$, positive for $x > 0$, but $\lim_{x \to 0} x = 0$, not $\pm\infty$.

Opposite-sided signs approaching $\pm \infty$ require that the function actually *blows up*, which needs to be established separately.

### Approach 1 — L'Hôpital's Rule (applied three times)

L'Hôpital's rule applies whenever the limit is in $\frac{0}{0}$ or $\frac{\pm\infty}{\pm\infty}$ form. Here we apply it repeatedly until the indeterminate form is resolved.

**First application** — still $\frac{0}{0}$:
$$\lim_{x \to 0} \frac{\sin(3x) - 3x}{x^3} = \lim_{x \to 0} \frac{3\cos(3x) - 3}{3x^2}$$

**Second application** — still $\frac{0}{0}$:
$$= \lim_{x \to 0} \frac{-9\sin(3x)}{6x}$$

**Third application** — still $\frac{0}{0}$:
$$= \lim_{x \to 0} \frac{-27\cos(3x)}{6}$$

Now we can substitute $x = 0$ directly:
$$= \frac{-27\cos(0)}{6} = \frac{-27}{6} = \boxed{-\frac{9}{2}}$$

**Key takeaway:** When L'Hôpital's first application still gives $\frac{0}{0}$, you apply it again. It can be used as many times as needed, as long as the indeterminate form persists.

### Approach 2 — Taylor Series (often faster for $x \to 0$ limits)

Recall the Taylor series for $\sin(u)$ around $u = 0$:
$$\sin(u) = u - \frac{u^3}{6} + \frac{u^5}{120} - \cdots$$

Substituting $u = 3x$:
$$\sin(3x) = 3x - \frac{(3x)^3}{6} + \frac{(3x)^5}{120} - \cdots = 3x - \frac{27x^3}{6} + O(x^5)$$

Now subtract $3x$ and divide by $x^3$:
$$\frac{\sin(3x) - 3x}{x^3} = \frac{-\dfrac{27x^3}{6} + O(x^5)}{x^3} = -\frac{27}{6} + O(x^2)$$

Taking $x \to 0$:
$$\lim_{x \to 0} \frac{\sin(3x) - 3x}{x^3} = -\frac{27}{6} = \boxed{-\frac{9}{2}}$$

**Why this approach is worth knowing:** For limits of the form $\dfrac{f(x)}{g(x)}$ as $x \to 0$ where both sides vanish, expanding to the first non-cancelling term is often more direct than repeated L'Hôpital applications.

**Both substitution strategies are valid — but nonlinear systems require a verification step.**

You have two options when recovering $y$ after solving for $x$:

**Why does this happen with nonlinear systems?** In a linear system, back-substitution into any equation always produces a unique, consistent result — no spurious solutions can appear. In a nonlinear system, a single equation can have multiple roots (here $y^2 = 1$ has two), and not all of them need satisfy the full system. This is why verification is essential whenever you back-substitute into a single equation of a nonlinear system.

$$f_{xx}(1, 1) = 6(1) = 6 > 0$$

The second derivative test rule is:
- $D > 0$ and $f_{xx} > 0$ → **local minimum** (bowl opening upward)
- $D > 0$ and $f_{xx} < 0$ → **local maximum** (bowl opening downward)
- $D < 0$ → **saddle point** (regardless of $f_{xx}$)

So (1, 1) is indeed a local minimum, but because $f_{xx} = +6 > 0$, not because it's negative. The intuition about concavity is also inverted: $f_{xx} > 0$ means the function curves *upward* in the $x$-direction (concave up), which is consistent with a minimum.

| Point | $D$ | $f_{xx}$ | Classification |
|-------|-----|----------|----------------|
| (0, 0) | $36(0)(0) - 9 = -9 < 0$ | — | Saddle point |
| (1, 1) | $36(1)(1) - 9 = 27 > 0$ | $6(1) = 6 > 0$ | Local minimum |

The function $f(x,y) = x^3 + y^3 - 3xy$ is sometimes called the "monkey saddle" family — it has a well-known saddle at the origin and a single local minimum at (1, 1).

### Numerical verification

````python
import numpy as np
from scipy.optimize import fsolve

def f(x, y):
    return x**3 + y**3 - 3*x*y

def grad(xy):
    x, y = xy
    return [3*x**2 - 3*y, 3*y**2 - 3*x]

# Find critical points numerically from multiple starting guesses
starts = [(-1, -1), (0.1, 0.1), (1, 1), (2, 0), (0, 2), (1, -1)]
found = set()
for s in starts:
    sol = fsolve(grad, s, full_output=True)
    if sol[2] == 1:  # converged
        pt = tuple(np.round(sol[0], 6))
        found.add(pt)

print("Critical points found:")
for pt in sorted(found):
    x, y = pt
    fxx, fyy, fxy = 6*x, 6*y, -3
    D = fxx*fyy - fxy**2
    kind = "saddle" if D < 0 else ("local min" if fxx > 0 else "local max")
    print(f"  ({x:.4f}, {y:.4f})  D={D:.4f}  fxx={fxx:.4f}  → {kind}")

# Verify (1,-1) is NOT a critical point
print("\nVerification that (1,-1) is not a critical point:")
print(f"  grad f at (1,-1) = {grad([1,-1])}")
````

### The IBP assignment error

You wrote $u = x$ (with $v = -1/x$). Notice what that produces for the $uv$ term:
$$u \cdot v = x \cdot \left(-\frac{1}{x}\right) = -1 \quad \text{(a constant, not a function of } x\text{)}$$

Whenever the $uv$ term collapses to a constant, the entire term vanishes when evaluated at the bounds — that is a strong signal the assignment is wrong. The whole point of the $uv$ term is to contribute a boundary evaluation; if it's constant, it carries no information.

**The LIATE rule** (Logarithm, Inverse trig, Algebraic, Trig, Exponential) tells you which factor to assign to $u$: choose the one appearing **earlier** in the list. Here, $\ln x$ is a logarithm and $x^{-2}$ is algebraic:

$$u = \ln x, \quad du = \frac{1}{x}\,dx \qquad dv = x^{-2}\,dx, \quad v = -\frac{1}{x}$$

Now $uv = -\dfrac{\ln x}{x}$, which is a genuine function of $x$ and will produce a meaningful boundary term.

**Why does LIATE work here?** The goal of IBP is to replace the original integral with a simpler one. Choosing $u = \ln x$ works because *differentiating* it ($du = \frac{1}{x}\,dx$) reduces complexity in the remaining integral. Choosing $u = x$ would force you to *integrate* $\ln x$, making the remaining integral harder.

### Was your arithmetic correct given $u = x$?

Yes — once you wrote down the (incorrect) IBP expression, every subsequent step was mechanically correct. The error was entirely in the setup, not the execution.

To see why the setup is inconsistent, recall that IBP requires $u \cdot dv$ to equal the original integrand:

| Assignment | $u \cdot dv$ | Equals integrand? |
|---|---|---|
| $u = \ln x,\ dv = x^{-2}\,dx$ | $\ln x \cdot x^{-2}\,dx = \dfrac{\ln x}{x^2}\,dx$ | ✓ |
| $u = x,\ dv = x^{-2}\,dx$ | $x \cdot x^{-2}\,dx = \dfrac{1}{x}\,dx$ | ✗ |

With $u = x$, the factor $\ln x$ disappears from the integrand entirely. You were no longer working on $\int \frac{\ln x}{x^2}\,dx$ — you had inadvertently set up IBP for $\int \frac{1}{x}\,dx$.

From the expression $\left[-1 - \int -\frac{1}{x}\,dx\right]_1^b$ onward, your steps were all correct:
- $-\int -\frac{1}{x}\,dx = \ln|x|$ ✓  
- $[\ln|x| - 1]_1^b$ ✓  
- $(\ln b - 1) - (0 - 1) = \ln b$ ✓  
- $\lim_{b \to \infty} \ln b = \infty$ ✓  

You correctly evaluated $\int_1^b \frac{1}{x}\,dx$ and correctly found it diverges. The algebra was sound — it just operated on the wrong integral.

### Correct solution

With the correct assignment:
$$u = \ln x, \quad du = \frac{1}{x}\,dx \qquad dv = x^{-2}\,dx, \quad v = -\frac{1}{x}$$

Applying IBP ($\int u\,dv = uv - \int v\,du$):
$$\int_1^b \frac{\ln x}{x^2}\,dx = \left[-\frac{\ln x}{x}\right]_1^b - \int_1^b \left(-\frac{1}{x}\right)\frac{1}{x}\,dx = \left[-\frac{\ln x}{x}\right]_1^b + \int_1^b x^{-2}\,dx$$

$$= \left[-\frac{\ln x}{x}\right]_1^b + \left[-\frac{1}{x}\right]_1^b = \left[-\frac{\ln x}{x} - \frac{1}{x}\right]_1^b$$

Evaluating at the bounds:
$$= \left(-\frac{\ln b}{b} - \frac{1}{b}\right) - \left(-\frac{\ln 1}{1} - \frac{1}{1}\right) = -\frac{\ln b}{b} - \frac{1}{b} + 0 + 1$$

Taking the limit as $b \to \infty$:
- $\dfrac{\ln b}{b} \to 0$ (L'Hôpital: $\dfrac{(\ln b)'}{(b)'} = \dfrac{1/b}{1} \to 0$)
- $\dfrac{1}{b} \to 0$

$$\lim_{b \to \infty} \left(-\frac{\ln b}{b} - \frac{1}{b} + 1\right) = 0 - 0 + 1 = \boxed{1}$$

The integral **converges to 1**.

**Intuition for convergence:** $x^2$ in the denominator dominates the slow logarithmic growth of $\ln x$. Recall $\int_1^\infty x^{-p}\,dx$ converges for $p > 1$; multiplying by $\ln x$ (which grows slower than any power of $x$) doesn't change that.

**Eigenvector scaling:** Eigenvectors are only defined up to a non-zero scalar multiple.

*Why algebraically:* If $v$ is an eigenvector ($Av = \lambda v$), multiply both sides by any non-zero scalar $c$:
$$A(cv) = c(Av) = c(\lambda v) = \lambda(cv)$$
So $cv$ satisfies the same equation with the same $\lambda$ — it is also an eigenvector. This holds for any $c \neq 0$: negatives, fractions, irrational numbers.

*Why geometrically:* The eigenvector equation says the matrix acts by pure scaling along a certain **line through the origin**. The line is the eigenspace — the invariant direction. Any non-zero vector pointing along that line is a valid representative. Writing $(-1, 1)^T$ or $(1, -1)^T$ or $(-3, 3)^T$ are all ways of pointing along the same line $y = -x$.

*The one restriction:* $c = 0$ is excluded because the zero vector satisfies $A \cdot 0 = \lambda \cdot 0$ for *every* $\lambda$ simultaneously — it carries no information about the matrix. By definition, eigenvectors must be non-zero.

*Why this matters in practice:* When you compute eigenvectors by hand and check against NumPy, they will often look different:
- Your answer: $(-1, 1)^T$
- NumPy's output: $(0.707, -0.707)^T$ (unit-normalised, opposite sign)

These represent the **same eigenvector** — one is $-\frac{1}{\sqrt{2}}$ times the other. Without knowing about scaling, you might think your answer is wrong when it is fully correct. The check is: are they scalar multiples of each other? If yes, they are the same eigenvector. Different sources simply make different conventional choices: NumPy normalises to unit length; by hand we usually pick the simplest integers with no common factor.

**Why this matrix is well-behaved:** $A$ is real and symmetric. By the Spectral Theorem, symmetric matrices always have real eigenvalues and orthogonal eigenvectors. Notice that $(-1, 1)^T \cdot (1, 1)^T = -1 + 1 = 0$ — the two eigenvectors you found are orthogonal. This is not a coincidence. For symmetric matrices, $P$ can always be chosen to be an orthogonal matrix (columns are orthonormal), giving the decomposition $A = Q\Lambda Q^T$ where $Q^{-1} = Q^T$. That is the symmetric eigendecomposition, and it underpins PCA and SVD.

## B2 — SVD Conceptual: Feedback

**Question:** What do the singular values of a matrix represent geometrically? Think about what any matrix does to the unit sphere/unit circle.

### Verdict

Good answer — the structural picture ($U$, $\Sigma$, $V^T$; rotation/reflection + stretch) is correct, and the analogy to eigenvalues is reasonable. The one gap is the geometric punchline the question was specifically probing: **what shape does the unit circle become, and what do the singular values measure about that shape?** That answer is not stated explicitly.

### What you got right

- **Structure of SVD:** $A = U\Sigma V^T$ with two unitary matrices and a diagonal matrix. ✓
- **Role of $U$ and $V$:** They rotate and/or reflect; they are isometries (preserve norms and angles). ✓
- **Role of $\Sigma$:** The diagonal entries (singular values) perform the stretching. ✓
- **Analogy to eigenvalues:** Reasonable — singular values measure how much the matrix stretches along certain directions. ✓

### The missing geometric picture: unit circle → ellipse

The question asked specifically what a matrix does to the unit circle. Here is the complete answer:

**Any matrix $A$ maps the unit circle (or sphere) to an ellipse (or ellipsoid).** The singular values $\sigma_1 \geq \sigma_2 \geq \cdots \geq 0$ are exactly the **lengths of the semi-axes** of that ellipse. The corresponding columns of $U$ (left singular vectors) give the **directions** of those axes.

The three-factor decomposition makes this action explicit. Applying $A = U\Sigma V^T$ to a vector $x$ on the unit circle proceeds in three steps:

1. **$V^T$** — rotate/reflect the input. Since $V$ is orthogonal, $V^T x$ is still on the unit circle.
2. **$\Sigma$** — stretch along each coordinate axis by $\sigma_1, \sigma_2, \ldots$. This turns the circle into an axis-aligned ellipse with semi-axes $\sigma_1, \sigma_2$.
3. **$U$** — rotate/reflect again. This tilts the ellipse into its final orientation in the output space.

So $\sigma_1$ is the maximum distance any unit input can be mapped to (the largest semi-axis), and $\sigma_n$ is the minimum (the smallest semi-axis). If any $\sigma_i = 0$, the matrix collapses that dimension — the image lives in a lower-dimensional subspace.

$$\text{unit circle} \xrightarrow{V^T} \text{rotated circle} \xrightarrow{\Sigma} \text{axis-aligned ellipse} \xrightarrow{U} \text{tilted ellipse}$$

### Three clarifications on your answer

**1. "Singular value vectors" — left vs right singular vectors**

SVD gives two sets of singular vectors: the columns of $V$ (right singular vectors, living in the input space) and the columns of $U$ (left singular vectors, living in the output space). The stretching by $\sigma_i$ maps the $i$-th right singular vector $v_i$ to $\sigma_i$ times the $i$-th left singular vector $u_i$:
$$A v_i = \sigma_i u_i$$
This is the singular value equation — analogous to $Ax = \lambda x$ for eigenvectors, but with a different vector on each side. The "direction of stretch" in the output is $u_i$; the "input direction being stretched" is $v_i$.

**2. Singular values are always non-negative**

You said singular values are "analogous to eigenvalues" — that's true, but one key difference: singular values $\sigma_i \geq 0$ always. Eigenvalues can be negative or complex. The largest singular value $\sigma_1$ equals the matrix's 2-norm (operator norm): $\|A\|_2 = \sigma_1$.

**3. SVD works for any matrix; eigendecomposition does not**

Eigendecomposition $A = PDP^{-1}$ requires $A$ to be square (and further requires $n$ linearly independent eigenvectors). SVD $A = U\Sigma V^T$ works for any $m \times n$ matrix. This is what makes SVD the more general and widely applicable tool — it can decompose a $1000 \times 50$ data matrix, for instance, where eigendecomposition isn't even defined.

### Numerical verification

````python
import numpy as np
import matplotlib.pyplot as plt

# Use the matrix from B1 (symmetric — singular values = eigenvalues here)
A = np.array([[3, 1], [1, 3]], dtype=float)

U, s, Vt = np.linalg.svd(A)
print("Singular values:", s)
print("U =\n", U)
print("V^T =\n", Vt)
print("\nReconstruct A = U Σ V^T:")
print(U @ np.diag(s) @ Vt)

# Visualise: apply A to the unit circle
theta = np.linspace(0, 2 * np.pi, 300)
circle = np.array([np.cos(theta), np.sin(theta)])  # (2, 300)
ellipse = A @ circle                                # (2, 300)

fig, axes = plt.subplots(1, 2, figsize=(10, 4))
axes[0].plot(*circle, 'b')
axes[0].set_aspect('equal'); axes[0].set_title('Unit circle (input)')
axes[0].axhline(0, color='k', lw=0.5); axes[0].axvline(0, color='k', lw=0.5)

axes[1].plot(*ellipse, 'r')
# Draw semi-axes: columns of U scaled by singular values
for i in range(2):
    v = s[i] * U[:, i]
    axes[1].annotate('', xy=v, xytext=(0,0),
                     arrowprops=dict(arrowstyle='->', color='navy', lw=2))
    axes[1].text(v[0]*1.05, v[1]*1.05, f'σ={s[i]:.1f}', fontsize=10)
axes[1].set_aspect('equal'); axes[1].set_title('Image of unit circle under A (ellipse)')
axes[1].axhline(0, color='k', lw=0.5); axes[1].axvline(0, color='k', lw=0.5)
plt.tight_layout()
plt.show()
print(f"\nSemi-axis lengths = singular values: {s[0]:.4f}, {s[1]:.4f}")
````

**Question:** Implement power iteration to find the dominant eigenvalue and eigenvector of $B = \begin{pmatrix} 4 & 1 \\ 2 & 3 \end{pmatrix}$.

### Verdict

The code converges to the correct dominant eigenvalue (5) and eigenvector direction, but contains three errors — two of which accidentally cancel at convergence. The final answer is right; the method is not. This is worth understanding carefully, because bugs that only surface in the limit are the hardest kind to catch.

### What you got right

- Correct overall structure: multiply by $B$, normalise, repeat for 20 iterations. ✓
- Writing `EucNorm` explicitly rather than calling `np.linalg.norm` is fine — it shows you understand what the norm is. ✓
- Verifying the result against `numpy.linalg.eig` at the end — exactly the right sanity check. ✓
- The code produces the correct final eigenvalue estimate of 5. ✓

### Error 1 — Wrong starting vector

The question specified $v_0 = (1, 1)^T$, but you initialised with `np.array([3, 1])`.

This is the least consequential error — any non-zero starting vector that is not exactly orthogonal to the dominant eigenvector will converge to the same result. But it does not match the spec, and the convergence rate depends on the initial vector.

### Error 2 — Normalising by the wrong vector's norm

Your loop update:
```python
vec1 = B @ vec1.T / EucNorm(vec1)
```

Python evaluates the entire right-hand side before assigning, so `vec1` on the RHS is still the **old** vector. You are dividing the new vector $Bv_k$ by the norm of the **old** $v_k$, not the new one.

The correct power iteration update is:
$$v_{k+1} = \frac{B v_k}{\|B v_k\|}$$

In code:
```python
w = B @ vec1
vec1 = w / EucNorm(w)
```

With the correct update, `vec1` has Euclidean norm 1 after every step. With your update, the norm is **not** preserved — it drifts and eventually stabilises at $\|\lambda_1 u_1\| = \lambda_1 = 5$.

**Why it still converges in direction:** Even with the wrong normalisation, the iterate aligns with the dominant eigenvector. Once $v_k \approx \alpha u_1$ (where $u_1$ is the unit dominant eigenvector):
$$v_{k+1} = \frac{B(\alpha u_1)}{\|\alpha u_1\|} = \frac{\alpha \lambda_1 u_1}{\alpha} = \lambda_1 u_1$$
The iterate stabilises at $\lambda_1 u_1$ — the correct direction, but magnitude $\lambda_1$ rather than 1.

### Error 3 — Incorrect Rayleigh quotient formula

The question told you explicitly: the eigenvalue estimate is $\lambda_k = v_k^T B v_k$ (for unit-normalised $v_k$).

You wrote:
```python
def Rayleigh(v1):
    return v1.dot(B @ v1.T) / EucNorm(B @ v1.T)
```

This computes $\dfrac{v^T(Bv)}{\|Bv\|}$, which is neither the Rayleigh quotient nor any other standard estimator. It is an accidental hybrid of two formulas that are each valid on their own:

**Formula 1 — Rayleigh quotient** (what the question asked for):
$$R(v) = \frac{v^T B v}{v^T v}$$
For a unit-normalised $v$ this simplifies to $v^T B v$. Gives a signed eigenvalue estimate; converges to the eigenvalue itself, not just its magnitude. Works particularly well for symmetric matrices where it provably bounds the eigenvalues.

**Formula 2 — Norm ratio** (the other common estimator):
$$\frac{\|Bv\|}{\|v\|}$$
For a unit $v$ this is just $\|Bv\|$. Gives $|\lambda|$ (magnitude only). Simpler to compute and a natural consequence of convergence: once $v \approx u_1$, we have $Bv \approx \lambda_1 v$, so $\|Bv\| / \|v\| \approx |\lambda_1|$.

Your formula takes the numerator from Formula 1 and the denominator from Formula 2, giving $v^T Bv / \|Bv\|$. This is not standard. If your course presented the norm ratio as the eigenvalue estimator, that is a legitimate alternative — the two are equivalent at convergence for a positive dominant eigenvalue, but the Rayleigh quotient gives better intermediate estimates and handles negative eigenvalues correctly.

**Correct implementation (Rayleigh quotient):**
```python
def Rayleigh(v1):
    return v1 @ B @ v1          # v^T B v, valid when v1 is unit-normalised
```

**Correct implementation (norm ratio):**
```python
def norm_ratio(v1):
    return np.linalg.norm(B @ v1) / np.linalg.norm(v1)   # ||Bv|| / ||v||
```

### Why errors 2 and 3 cancel at convergence

This is the most instructive part. Both bugs are real errors, yet the final printed eigenvalue is correct. Here is why.

With Bug 2, after convergence `vec1 → λ₁ u₁` (the dominant eigenvector scaled to magnitude $\lambda_1 = 5$, not 1).

Now evaluate your Rayleigh formula on $v = \lambda_1 u_1$:

$$\frac{v^T(Bv)}{\|Bv\|} = \frac{(\lambda_1 u_1)^T (\lambda_1^2 u_1)}{\|\lambda_1^2 u_1\|} = \frac{\lambda_1^3 \|u_1\|^2}{\lambda_1^2 \|u_1\|} = \frac{\lambda_1^3}{\lambda_1^2} = \lambda_1 \checkmark$$

So your Rayleigh formula happens to return $\lambda_1$ when applied to a vector of magnitude $\lambda_1$ — not because it is the Rayleigh quotient, but because the magnitudes conspire to cancel correctly.

**During intermediate iterations**, the two errors do *not* cancel: the iterate hasn't yet converged to the eigenvector direction, and the printed eigenvalue estimates are unreliable. A correct implementation would give a good eigenvalue estimate from the very first step.

### Minor notes

**`vec1.T` on a 1D array is a no-op.** NumPy 1D arrays have no row/column distinction — `.T` returns the same array unchanged. It reads as though you are transposing to a column vector, but nothing happens. Use `.T` only on 2D arrays.

**`np.linalg.eig` returns a tuple.** Printing `np.linalg.eig(B)` shows the raw `(eigenvalues, eigenvectors)` tuple. Unpacking is cleaner:
```python
eigenvalues, eigenvectors = np.linalg.eig(B)
print("Eigenvalues:", eigenvalues)
```

import numpy as np
