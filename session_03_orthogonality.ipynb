## Q1 — Projection onto a vector

Let $\mathbf{u} = (2, 1, -1)^T$ and $\mathbf{v} = (1, 3, 1)^T$.

(a) Find the scalar projection of $\mathbf{v}$ onto $\mathbf{u}$.

(b) Find the vector projection of $\mathbf{v}$ onto $\mathbf{u}$.

(c) Find the component of $\mathbf{v}$ orthogonal to $\mathbf{u}$. Verify it is orthogonal.

**Your answer:**

Scalar projection uses the formula $\frac{v\ .\ u}{||u||}$ to project v onto u. This gives:

Working out the magnitude of u first:

$$||u||\ =\ \sqrt{2^2\ +\ 1^2\ +\ (-1)^2} = \sqrt{6}$$

Then the dot product of v and u:

$$\left(2, 1, -1 \right).\left(1,3,1\right)\ = 4$$

Applying the formula therefore gives:

$$\frac{4}{\sqrt{6}}\ which\ simplifies\ to\ \frac{2\sqrt{6}}{3}$$

Vector projectio uses the formula $\frac{v\ .\ u}{||u||^2}\ .\ u$

plugging in values we have already worked out (including that magnitude u squared gets rid of the square root meaning $||u||^2\ = \sqrt{6}^2 = \ 6$, gives:

$$\frac{2}{3}\left(2,1,-1\right)\ =\ \left(\frac{4}{3}, \frac{2}{3}, \frac{-1}{3}\right)$$

We can now find the component of v that is orthogonal to u by simply subtracting from v the pieces of v that are in the direction of u that we just worked out with vector projection:

$$\left(1,3,1\right)\ -\ \left(\frac{4}{3}, \frac{2}{3}, \frac{-1}{3}\right)\ =\ \left(\frac{-1}{3}, \frac{7}{3}, \frac{5}{3}\right)

and we can show it is orthogonal by simply taking the dot product of the resulting vector with u, if it is orthogonal the result will equal 0:

$$\left(\frac{-1}{3}, \frac{7}{3}, \frac{5}{3}\right)\ .\ \left(2,1,-1\right)\ =\ \frac{-2}{3}\ +\ \frac{7}{3}\ -\ \frac{5}{3}\ =\ 0$$

As the dot product equals zero this shows it is orthogonal to u



## Q2 — Gram-Schmidt

Apply Gram-Schmidt to the following vectors to produce an orthonormal basis for their span:

$$\mathbf{a}_1 = \begin{pmatrix}1\\1\\0\end{pmatrix}, \quad \mathbf{a}_2 = \begin{pmatrix}1\\0\\1\end{pmatrix}, \quad \mathbf{a}_3 = \begin{pmatrix}0\\1\\1\end{pmatrix}$$

Show each step: the unnormalised orthogonal vectors $\mathbf{q}_1', \mathbf{q}_2', \mathbf{q}_3'$ and the final unit vectors $\mathbf{q}_1, \mathbf{q}_2, \mathbf{q}_3$.

**Your answer:**

I will choose $q^{'}_1$ to be $a_1$ so $q^{'}_1 = \left(1,1,0 \right)$. Therefore to create $q_1$ I need to normalise by dividing by its magnitude $||q^{'}_1|| = \frac{1}{\sqrt{2}}$ giving $q_1=\left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right)$

To create $q_2$ I need to subtract from $a_2$ all its components that are in the direction of $q_1$ and then normalise it by dividing through by its magnitude:

Create $q^{'}_2$, project $a_2$ onto $q_1$ and subtract from $a_2$ (note in the projection I dont need to divide by the magnitude squared as I have normalised  $q^{'}_1$ to create $q_1$)

$$\left(1,0,1\right)\ . \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right) = \frac{1}{\sqrt{2}}$$
$$\frac{1}{\sqrt{2}} \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right) = \left(\frac{1}{2}, \frac{1}{2}, 0 \right)$$

therefore 

$$q^{'}_2 = \left(1,0,1\right) - \left(\frac{1}{2}, \frac{1}{2}, 0 \right) = \left(\frac{1}{2}, \frac{-1}{2}, 1 \right)$$

and normalising to get $q_2$

$$||q^{'}_2|| = \sqrt{\frac{1}{4} + \frac{1}{4} + 1} = \sqrt{\frac{3}{2}}$$

$$q_2 = \frac{1}{\sqrt{\frac{3}{2}}}\left(\frac{1}{2}, \frac{-1}{2}, 1\right)=\left(\frac{\sqrt{\frac{3}{2}}}{3}, \frac{- \sqrt{\frac{3}{2}}}{3}, \frac{2\sqrt{\frac{3}{2}}}{3}\right)$$

To create $q_3$ I need to subtract from $a_3$ all its components that are in the direction of $q_1$ and $q_2$ and then normalise it by dividing through by its magnitude:

Project $a_3$ onto $q_1$:

$$\left(0,1,1\right) . \left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right) = \frac{1}{\sqrt{2}}$$
$$\frac{1}{\sqrt{2}}\left(\frac{1}{\sqrt{2}}, \frac{1}{\sqrt{2}}, 0 \right) = \left(\frac{1}{2}, \frac{1}{2}, 0 \right)$$

Project $a_3$ onto $q_2$:

$$\left(0,1,1\right) . \left(\frac{\sqrt{\frac{3}{2}}}{3}, \frac{- \sqrt{\frac{3}{2}}}{3}, \frac{2\sqrt{\frac{3}{2}}}{3}\right) = \frac{\sqrt{\frac{3}{2}}}{3}$$
$$\frac{\sqrt{\frac{3}{2}}}{3}\left(\frac{\sqrt{\frac{3}{2}}}{3}, \frac{- \sqrt{\frac{3}{2}}}{3}, \frac{2\sqrt{\frac{3}{2}}}{3}\right)=\left(\frac{1}{6}, \frac{-1}{6}, 1\right)$$

subract projections from $a_3$ to get $q^{'}_3$:

$$q^{'}_3=\left(0,1,1\right) - \left(\frac{1}{2}, \frac{1}{2}, 0 \right) - \left(\frac{1}{6}, \frac{-1}{6}, 1\right) = \left(\frac{-2}{3}, \frac{2}{3}, 0\right)$$

Find magnitude of $q^{'}_3$:

$$||q^{'}_3|| = \sqrt{\frac{4}{9} + \frac{4}{9} + 0} = \sqrt{\frac{8}{9}}$$

Divide $q^{'}_3$ by its magnitude to find $q_3$:

$$q_3 = \frac{1}{\sqrt{\frac{8}{9}}} \left(\frac{-2}{3}, \frac{2}{3}, 0\right) = \left(\frac{-3\sqrt{\frac{8}{9}}}{4}, \frac{3\sqrt{\frac{8}{9}}}{4}, 0\right)$$


## Q3 — Projection matrix

Let $A = \begin{pmatrix}1 & 0\\1 & 1\\0 & 1\end{pmatrix}$.

(a) Write down the formula for the projection matrix $P$ onto the column space of $A$.

(b) Compute $P$.

(c) Project $\mathbf{b} = (1, 2, 3)^T$ onto $\text{col}(A)$.

(d) Verify that $\mathbf{b} - P\mathbf{b}$ is orthogonal to both columns of $A$.

**Your answer:**

The formula fto create the projection matrix is : $P_{proj} = A(A^TA)^{-1}A^T$ Assuming A's columns are linearly independent, which in the case they are given the off diagonal zeros.

I compute P as follows:

$$P = \begin{bmatrix} 1 & 0 \\ 1 & 1 \\ 0 & 1 \end{bmatrix} \left( \begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}\begin{bmatrix} 1 & 0 \\ 1 & 1 \\ 0 & 1 \end{bmatrix}\right)^{-1} \begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}$$
$$P = \begin{bmatrix} 1 & 0 \\ 1 & 1 \\ 0 & 1 \end{bmatrix} \begin{bmatrix} 2 & 1 \\ 1 & 2\end{bmatrix}^{-1} \begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}$$
$$P = \begin{bmatrix} 1 & 0 \\ 1 & 1 \\ 0 & 1 \end{bmatrix} \begin{bmatrix} \frac{2}{3} & \frac{-1}{3} \\ \frac{-1}{3} & \frac{2}{3}\end{bmatrix}\begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}$$
$$P = \begin{bmatrix} \frac{2}{3} & \frac{-1}{3} \\ \frac{1}{3} & \frac{1}{3} \\ \frac{-1}{3} & \frac{2}{3} \end{bmatrix}\begin{bmatrix} 1 & 1 & 0 \\ 0 & 1 & 1 \end{bmatrix}$$
$$P = \begin{bmatrix} \frac{2}{3} & \frac{1}{3} & \frac{-1}{3}\\ \frac{1}{3} & \frac{2}{3} & \frac{1}{3} \\ \frac{-1}{3} & \frac{1}{3} & \frac{2}{3} \end{bmatrix}$$

I can now use P to project $b = (1,2,3)^T$ onto the column space of A:

$$Pb = \begin{bmatrix} \frac{2}{3} & \frac{1}{3} & \frac{-1}{3}\\ \frac{1}{3} & \frac{2}{3} & \frac{1}{3} \\ \frac{-1}{3} & \frac{1}{3} & \frac{2}{3} \end{bmatrix}\begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} = \begin{bmatrix} \frac{1}{3} \\ \frac{8}{3} \\ \frac{7}{3}\end{bmatrix}$$

I can now compute $b - Pb$ as follows:

$$b - Pb = \begin{bmatrix} 1 \\ 2 \\ 3 \end{bmatrix} - \begin{bmatrix} \frac{1}{3} & \frac{8}{3} & \frac{7}{3} \end{bmatrix} = \begin{bmatrix} \frac{2}{3} & \frac{-2}{3} & \frac{2}{3} \end{bmatrix}$$

I can proove these are orthoganal to the columns of A by taking the dot product of this vector with each column vector of A and it should equal zero in both cases:

$$\begin{bmatrix} 1 & 1 & 0 \end{bmatrix} . \begin{bmatrix} \frac{2}{3} & \frac{-2}{3} & \frac{2}{3} \end{bmatrix} = \frac{2}{3} + \frac{-2}{3} + 0 = 0$$

$$\begin{bmatrix} 0 & 1 & 1 \end{bmatrix} . \begin{bmatrix} \frac{2}{3} & \frac{-2}{3} & \frac{2}{3} \end{bmatrix} = 0 + \frac{-2}{3} + \frac{2}{3} = 0$$

As the dot product of $ b - Pb $ is zero with both columns of A it is orthogonal to both columns.




## Q4 — Least squares

Three data points: $(0, 1)$, $(1, 2)$, $(2, 2.5)$. Find the best-fit line $y = c_0 + c_1 x$ in the least-squares sense by setting up and solving the normal equations $A^T A \hat{x} = A^T \mathbf{b}$.

**Your answer:**

I create a Design matrix from the x values and by initialising the intercept to 1 (by convention) this gives me:

$$X = \begin{bmatrix} 1 & 0 \\ 1 & 1 \\ 1 & 2\end{bmatrix}$$

The coefficients I am solving for are noted in a vector $\beta$:

$$\beta = \begin{bmatrix}C_0 \\ C_1\end{bmatrix}$$

The observed outputs y are used to compse a vector y:

$$y = \begin{bmatrix}1 \\ 2 \\ 2.5\end{bmatrix}$$

I can solve for $\beta$ by left multiplying both sides of the equation $X\beta = y$ by the left inverse of the matrix X (we use the left inverse $(X^TX)^{-1}$ as the X matrix is a tall matrix) $\beta = (X^TX)^{-1}X^Ty$:

First I find $X^TX$

$$\begin{bmatrix}1 & 1 & 1 \\ 0 & 1 & 2\end{bmatrix} \begin{bmatrix} 1 & 0 \\ 1 & 1 \\ 1 & 2\end{bmatrix} = \begin{bmatrix}3 & 3 \\ 3 & 5\end{bmatrix}$$

I then find the inverse $(X^TX)^{-1}$:

$$det((X^TX)) = \frac{1}{3\ .\ 5 - 3\ .\ 3} = \frac{1}{6}$$
$$cofactor\ matrix = \begin{bmatrix}5 & -3 \\ -3 & 3\end{bmatrix}$$

therefore 

$$(X^TX)^{-1} = \frac{1}{6}\begin{bmatrix}5 & -3 \\ -3 & 3\end{bmatrix}$$

I can now use this as the left inverse in $\beta = (X^TX)^{-1}y$:

$$\frac{1}{6}\begin{bmatrix}5 & -3 \\ -3 & 3\end{bmatrix}\begin{bmatrix}1 & 1 & 1 \\ 0 & 1 & 2\end{bmatrix}\begin{bmatrix}1 \\ 2 \\ 2.5\end{bmatrix}$$
$$\frac{1}{6}\begin{bmatrix}5 & -3 \\ -3 & 3\end{bmatrix}\begin{bmatrix}5.5 \\ 7\end{bmatrix}$$
$$\frac{1}{6}\begin{bmatrix}6.5 \\ 4.5\end{bmatrix}$$
$$\begin{bmatrix}1.08333 \\ 0.750\end{bmatrix}$$

These are the coefficients $C_0$ and $C_1$ that I can use to create the line equation:

$$y = 1.08333 + 0.750x$$
