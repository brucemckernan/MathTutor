Two integration techniques that handle different structural patterns. Trig substitution is for integrands containing $\sqrt{a^2 - x^2}$, $\sqrt{a^2 + x^2}$, or $\sqrt{x^2 - a^2}$. Partial fractions is for rational functions — when the degree of the numerator is less than the degree of the denominator.

## Q2

Evaluate $\displaystyle\int \frac{x^2}{\sqrt{x^2 + 4}} \, dx$.

**Your answer:**

I recognise the pattern $\sqrt{a^2 + b^2}$ which alows me to use trigonometric substituion with $\sqrt{x^2 + 4}$ as the hypoteneuse. This means I can make x the opposite side and 2 the adjacent side to give me x over a constant with the tangent relationship:

$$tan \theta = \frac{x}{2} , x=2tan \theta, dx = 2sec^2\theta\, d\theta$$
$$\int \frac{4tan^2\theta}{\sqrt{4tan^2\theta + 4}}2sec^2\theta\, d\theta$$

I then simplify as much as I can before I integrate:

$$\int \frac{4tan^2\theta}{\sqrt{4(tan^2\theta + 1)}}2sec^2\theta\, d\theta$$
$$\int \frac{4tan^2\theta}{\sqrt{4sec^2\theta}}2sec^2\theta\, d\theta$$

At this point I use absolute value for $\left|sec \theta \right|$ as I dont know the angle of $\theta$ and therefore $sec \theta$ could be negative
$$\int \frac{4tan^2\theta}{4\left|sec\theta\right|}2sec^2\theta\, d\theta$$

Howwever in the next step I cancel $\left|sec \theta \right|$ with one power of $sec^2 \theta$ as a result the expression can take on any value including negative so I drop the absolute value (Note to claude ... I feel a little unconfident about this chain of reasoning and would appreicate your guidance). I also use the $tan^2 \theta = sec^2\ theta - 1$ to get rid of the tangent at this point to give me an expression only in terms of secant.

$$4 \int tan^2\theta sec\theta \, d\theta$$
$$4 \int \left(sec^2\theta - 1\right) sec\theta \, d\theta$$
$$4 \int \left(sec^3\theta - sec\theta\right)\, d\theta$$

I then split up the integral on subtraction:

$$4 \left[ \int sec^3\theta \, d\theta - \int sec\theta\, d\theta \right]$$

solve the simple integral piece $\int sec \theta\, d\theta$:

$$[1] \rightarrow 4 \int tan^2\theta sec\theta \, d\theta = 4 \left[ \int sec^3\theta \, d\theta - ln \left|sec\theta+tan\theta\right| \right]$$

I then take the remaining integrand piece and split off one power of secant to set up using IBP to solve:

$$\int sec^3\theta \rightarrow \int sec^2\theta sec \theta$$
$$u = sec\theta ,\ du = sec \theta \tan \theta \, d\theta ,\ dv = sec^2\theta ,\ v = tan\theta$$
$$sec\theta\ tan\theta - \int tan\theta\ sec\theta\ tan\theta\,d\theta$$
$$sec\theta\ tan\theta - \int tan^2\theta\ sec\theta\ ,d\theta$$

At this point I see that I have an unresolved integral piece that matches the left hand side of the equation marked above as [1]. So I plug the whole expression back into that equation:
$$4 \int tan^2\theta sec\theta \, d\theta = 4 \left[ sec\theta\ tan\theta - \int tan^2\theta\ sec\theta\ ,d\theta - ln \left|sec\theta+tan\theta\right| \right]$$

I distribute the constant 4 on the right hand side:

$$4 \int tan^2\theta sec\theta \, d\theta = 4 sec\theta\ tan\theta - 4\ \int tan^2\theta\ sec\theta\ ,d\theta - 4\ ln \left|sec\theta+tan\theta\right|$$

I then add the outstanding integral expression to both sides to eliminate it on the right hand side, and then simplify to give me my final ingtegral result:

$$8 \int tan^2\theta sec\theta \, d\theta = 4 sec\theta\ tan\theta - 4\ ln \left|sec\theta+tan\theta\right|$$
$$\int tan^2\theta sec\theta \, d\theta = \frac{1}{2}\ sec\theta\ tan\theta - \frac{1}{2}\ ln \left|sec\theta+tan\theta\right| + C$$
