# Session 08 — Trigonometric Substitution and Partial Fractions

Two integration techniques that handle different structural patterns. Trig substitution is for integrands containing $\sqrt{a^2 - x^2}$, $\sqrt{a^2 + x^2}$, or $\sqrt{x^2 - a^2}$. Partial fractions is for rational functions — when the degree of the numerator is less than the degree of the denominator.
