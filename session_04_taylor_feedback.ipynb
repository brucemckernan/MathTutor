#### Why direct substitution works — and when to use it

**The key insight:** a Maclaurin series is an identity, not just a formula about the letter $x$.

> **Side note — what "identity" means here.**
>
> In mathematics there is a sharp distinction between an *equation* and an *identity*:
>
> - An **equation** is true only for *specific* values of the variable. For example, $x^2 = 4$ is only true at $x = \pm 2$, and $e^x = 2$ is only true at $x = \ln 2$.
> - An **identity** is true for *every* value of the variable within its domain. "True for all real $x$" is the defining property. For example, $(a + b)^2 = a^2 + 2ab + b^2$ holds for all real $a, b$, and $\sin^2 x + \cos^2 x = 1$ holds for all real $x$.
>
> | Expression | Type | Why |
> |---|---|---|
> | $(a+b)^2 = a^2 + 2ab + b^2$ | Identity | True for all real $a, b$ |
> | $\sin^2 x + \cos^2 x = 1$ | Identity | True for all real $x$ |
> | $e^x = \sum x^n/n!$ | Identity | True for all real $x$ |
> | $x^2 = 4$ | Equation | Only true at $x = \pm 2$ |
> | $e^x = 2$ | Equation | Only true at $x = \ln 2$ |
>
> **Domain matters.** Some identities hold on a restricted domain rather than all of $\mathbb{R}$. The series $\frac{1}{1-x} = \sum_{n=0}^{\infty} x^n$ is an identity, but only on $|x| < 1$ where the series converges. The defining requirement is just that the equation holds for *all* values in its natural domain — not necessarily all reals.
>
> You will sometimes see $\equiv$ used instead of $=$ to signal "this is an identity", though most texts just use $=$ and rely on context.
>
> The Maclaurin series $e^x = \displaystyle\sum_{n=0}^{\infty}\frac{x^n}{n!}$ is an identity: both sides are equal for every real $x$. This is why substitution is valid — because the equation holds for **all** $x$, the variable is just a placeholder you can fill with any expression. Replacing $x$ with $-x^2$ throughout gives a new identity:
> $$e^{-x^2} = \sum_{n=0}^{\infty}\frac{(-x^2)^n}{n!}$$
> If the series were only true for *some* values of $x$, substituting a different expression would not be valid — you would lose the guarantee that both sides are equal.

Setting $u = -x^2$:

$$e^{-x^2} = \sum_{n=0}^{\infty}\frac{(-x^2)^n}{n!} = \sum_{n=0}^{\infty}\frac{(-1)^n x^{2n}}{n!} = 1 - x^2 + \frac{x^4}{2!} - \frac{x^6}{3!} + \cdots$$

That's it. The series for $e^u$ converges for all real $u$, so this is valid for all $x$.

**Why differentiation gives the same answer (but is harder):** the Maclaurin series is unique — there is only one power series centred at 0 that represents a given function. So your differentiation approach and the substitution approach must produce the same coefficients. Differentiation is just a much longer path to the same destination.

---

#### Rules of thumb for when substitution applies

Spot the pattern $f(g(x))$ where you already know the series for $f$. Then replace every $x$ in that series with $g(x)$.

| Function | Substitution | Series |
|---|---|---|
| $e^{-x^2}$ | $u = -x^2$ in $e^u$ | $1 - x^2 + \frac{x^4}{2!} - \frac{x^6}{3!} + \cdots$ |
| $\sin(x^2)$ | $u = x^2$ in $\sin u$ | $x^2 - \frac{x^6}{3!} + \frac{x^{10}}{5!} - \cdots$ |
| $\cos(3x)$ | $u = 3x$ in $\cos u$ | $1 - \frac{9x^2}{2!} + \frac{81x^4}{4!} - \cdots$ |
| $\frac{1}{1+x^2}$ | $u = -x^2$ in $\frac{1}{1-u}$ | $1 - x^2 + x^4 - x^6 + \cdots$ |

**The one thing to check — radius of convergence.** The substituted series is valid only where the original series converges *for the substituted expression*. For $e^u$ this is all of $\mathbb{R}$ so there is never a restriction. For the geometric series $\frac{1}{1-u} = \sum u^n$ you need $|u| < 1$, so substituting $u = -x^2$ requires $|-x^2| < 1$, i.e. $|x| < 1$.

**Rule of thumb in one sentence:** if you can write your function as $f(\text{something})$ and you know the series for $f$, substitute — do not differentiate.

#### Part (c) — ✓ Correct

Your term-by-term integration is exactly right, and your final evaluation is correct:

$$1 - \frac{1}{3} + \frac{1}{10} - \frac{1}{42} = 0.7429$$

The code cell below verifies this and shows how adding more terms closes in on the true value.

```python
import numpy as np
from scipy import integrate

# Four-term approximation (your answer)
approx_4 = 1 - 1/3 + 1/10 - 1/42
print(f"4-term approximation : {approx_4:.6f}")

# General term: integral of (-1)^n * x^(2n) / n! from 0 to 1 = (-1)^n / (n! * (2n+1))
def series_approx(num_terms):
    from math import factorial
    return sum((-1)**n / (factorial(n) * (2*n + 1)) for n in range(num_terms))

for k in [4, 6, 8, 10]:
    print(f"{k}-term approximation : {series_approx(k):.6f}")

# True value via numerical integration
true_val, _ = integrate.quad(lambda x: np.exp(-x**2), 0, 1)
print(f"True value           : {true_val:.6f}")
```

## Q2 — Limit via Taylor series

$$\lim_{x \to 0} \frac{1 - \cos x}{x^2}$$

### Feedback

**Result: ✓ Correct — limit is $\dfrac{1}{2}$**

Your reasoning is exactly right. Here is a line-by-line commentary.

---

#### Substitution step — ✓

You replaced $\cos x$ with its Maclaurin series and grouped the higher-order terms under $O(x^4)$:

$$\lim_{x \to 0}\frac{1 - \left(1 - \dfrac{x^2}{2} + O(x^4)\right)}{x^2}$$

That is the correct move. Using big-O here is good practice — it signals "I know there are more terms but I'm consciously setting them aside."

---

#### Small notational point: $-O(x^4)$ vs $+O(x^4)$

After distributing the minus sign you wrote:

$$\frac{\dfrac{x^2}{2} - O(x^4)}{x^2}$$

Strictly, $O(x^4)$ already absorbs the sign (it stands for "some function bounded by a constant times $x^4$ near zero", with no implied sign). The conventional writing is:

$$\frac{\dfrac{x^2}{2} + O(x^4)}{x^2}$$

This is a minor notational point and does not affect the answer — your algebra is correct. Just be aware that $-O(x^4)$ and $+O(x^4)$ are the same class; most texts write $+O$ or just $O$ after distributing.

---

#### Cancellation and the key step — ✓

$$\lim_{x \to 0}\frac{\dfrac{x^2}{2} + O(x^4)}{x^2} = \lim_{x \to 0}\left(\frac{1}{2} + O(x^2)\right)$$

Dividing numerator and denominator by $x^2$: the $\frac{x^2}{2}$ term becomes the constant $\frac{1}{2}$, and the $O(x^4)$ terms become $O(x^2)$ — this is the precise reason you can discard them.

---

#### Why the $O(x^2)$ terms vanish — a tighter statement

Your explanation ("higher-degree terms go to zero far faster") captures the right intuition. Here is the precise version:

> After dividing by $x^2$, the remaining higher-order terms form an $O(x^2)$ expression. As $x \to 0$, any $O(x^2)$ quantity $\to 0$. So:
>
> $$\lim_{x \to 0}\left(\frac{1}{2} + O(x^2)\right) = \frac{1}{2} + 0 = \frac{1}{2}$$

The key word is *after* dividing. Before dividing, both $\frac{x^2}{2}$ and $O(x^4)$ go to zero — so comparing their "speeds" would not tell you the limit. It is the division by $x^2$ that promotes $\frac{x^2}{2}$ to the constant $\frac{1}{2}$ while demoting $O(x^4)$ to $O(x^2) \to 0$.

---

#### Why this beats L'Hôpital

L'Hôpital would require two applications:

1. Differentiate numerator and denominator: $\dfrac{\sin x}{2x}$ — still $\frac{0}{0}$.
2. Differentiate again: $\dfrac{\cos x}{2} \to \dfrac{1}{2}$.

Taylor series gives the answer in one algebraic step with no repeated differentiation. For limits of the form $\dfrac{f(x) - f(0) - f'(0)x}{x^2}$ (or any $\frac{0}{0}$ form where the Taylor expansion is known), Taylor is almost always faster.

```python
import sympy as sp

x = sp.Symbol('x')

# Verify the limit symbolically
expr = (1 - sp.cos(x)) / x**2
lim = sp.limit(expr, x, 0)
print(f"Symbolic limit: {lim}")

# Show the Taylor expansion making the answer visible
cos_series = sp.series(sp.cos(x), x, 0, 6)
print(f"\ncos x = {cos_series}")

numerator_series = sp.series(1 - sp.cos(x), x, 0, 6)
print(f"\n1 - cos x = {numerator_series}")

ratio_series = sp.series((1 - sp.cos(x)) / x**2, x, 0, 4)
print(f"\n(1 - cos x) / x² = {ratio_series}")
```

**Key takeaway:** Your method is exactly the one to use for $\frac{0}{0}$ limits when you know the Taylor series. The one refinement: after cancellation, the reason the leftover terms vanish is that dividing by $x^2$ converts $O(x^4) \to O(x^2) \to 0$ — not just that "higher powers are smaller" in isolation.
