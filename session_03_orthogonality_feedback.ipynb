## Q1 — Projection onto a vector

**Question:** Let $\mathbf{u} = (2,1,-1)^T$, $\mathbf{v} = (1,3,1)^T$. (a) Scalar projection of $\mathbf{v}$ onto $\mathbf{u}$. (b) Vector projection. (c) Component of $\mathbf{v}$ orthogonal to $\mathbf{u}$; verify.

### Result: Correct — one arithmetic slip that doesn't propagate ✓

All three parts are right. There is a single typo in the stated vector projection (third component), but your subtraction in part (c) uses the correct value, so the orthogonal component and the verification are both correct.

---

### (a) Scalar projection

$$\text{scalar proj}_{\mathbf{u}}\mathbf{v} = \frac{\mathbf{v}\cdot\mathbf{u}}{\|\mathbf{u}\|}$$

$$\|\mathbf{u}\| = \sqrt{4+1+1} = \sqrt{6} \qquad \mathbf{v}\cdot\mathbf{u} = 2+3-1 = 4$$

$$\text{scalar projection} = \frac{4}{\sqrt{6}} = \frac{4\sqrt{6}}{6} = \frac{2\sqrt{6}}{3} \checkmark$$

---

### (b) Vector projection

$$\text{proj}_{\mathbf{u}}\mathbf{v} = \frac{\mathbf{v}\cdot\mathbf{u}}{\|\mathbf{u}\|^2}\,\mathbf{u} = \frac{4}{6}\begin{pmatrix}2\\1\\-1\end{pmatrix} = \frac{2}{3}\begin{pmatrix}2\\1\\-1\end{pmatrix} = \begin{pmatrix}\tfrac{4}{3}\\[4pt]\tfrac{2}{3}\\[4pt]-\tfrac{2}{3}\end{pmatrix}$$

**The slip:** You wrote $\left(\tfrac{4}{3},\tfrac{2}{3},-\tfrac{1}{3}\right)$. The third component is $\tfrac{2}{3}\times(-1) = -\tfrac{2}{3}$, not $-\tfrac{1}{3}$. This is a transcription error — your part (c) subtraction gives the correct answer $(−\tfrac{1}{3},\tfrac{7}{3},\tfrac{5}{3})$, which is only consistent with using $-\tfrac{2}{3}$, not $-\tfrac{1}{3}$. So you did the arithmetic right; you just mis-stated the vector projection.

---

### (c) Orthogonal component

$$\mathbf{v}_{\perp} = \mathbf{v} - \text{proj}_{\mathbf{u}}\mathbf{v} = \begin{pmatrix}1\\3\\1\end{pmatrix} - \begin{pmatrix}\tfrac{4}{3}\\[4pt]\tfrac{2}{3}\\[4pt]-\tfrac{2}{3}\end{pmatrix} = \begin{pmatrix}-\tfrac{1}{3}\\[4pt]\tfrac{7}{3}\\[4pt]\tfrac{5}{3}\end{pmatrix} \checkmark$$

**Verification:**

$$\mathbf{v}_{\perp}\cdot\mathbf{u} = \left(-\tfrac{1}{3}\right)(2) + \left(\tfrac{7}{3}\right)(1) + \left(\tfrac{5}{3}\right)(-1) = -\tfrac{2}{3}+\tfrac{7}{3}-\tfrac{5}{3} = 0 \checkmark$$

---

### The geometry behind this

The projection formula decomposes $\mathbf{v}$ into two perpendicular pieces:

$$\mathbf{v} = \underbrace{\text{proj}_{\mathbf{u}}\mathbf{v}}_{\text{along }\mathbf{u}} + \underbrace{\mathbf{v}_{\perp}}_{\perp\,\mathbf{u}}$$

The scalar projection gives the *signed length* of the shadow of $\mathbf{v}$ onto the line through $\mathbf{u}$. The vector projection gives the actual *vector* along $\mathbf{u}$ of that length. Subtracting it leaves only the part of $\mathbf{v}$ that points away from $\mathbf{u}$ — this is the residual, and it must be perpendicular by construction.

This decomposition is the geometric core of Gram-Schmidt (Q2), projection matrices (Q3), and least squares (Q4).

```python
import numpy as np
u = np.array([2, 1, -1])
v = np.array([1, 3, 1])
scalar_proj = np.dot(v, u) / np.linalg.norm(u)
vec_proj = (np.dot(v, u) / np.dot(u, u)) * u
v_perp = v - vec_proj
print(f"scalar projection: {scalar_proj:.6f}  (= 2√6/3 ≈ {2*np.sqrt(6)/3:.6f})")
print(f"vector projection: {vec_proj}")
print(f"orthogonal component: {v_perp}")
print(f"dot product check: {np.dot(v_perp, u):.10f}")
```

## Q3 — Projection matrix

**Question:** Let $A = \begin{pmatrix}1&0\\1&1\\0&1\end{pmatrix}$. Build the projection matrix $P$ onto $\text{col}(A)$, project $\mathbf{b}=(1,2,3)^T$, compute the residual, and verify it is orthogonal to both columns.

### Result: Fully correct ✓

Flawless execution from formula to final check — every matrix multiplication, inverse, and dot product is right.

---

### Formula and setup

$$P = A(A^TA)^{-1}A^T$$

This requires $A$ to have linearly independent columns so that $A^TA$ is invertible. The columns of $A$ are $(1,1,0)^T$ and $(0,1,1)^T$ — neither is a scalar multiple of the other, so they are independent. ✓

**One small note on your reasoning:** you cited "off-diagonal zeros" as evidence of independence. But $A^TA = \begin{pmatrix}2&1\\1&2\end{pmatrix}$ has non-zero off-diagonal entries — the matrix $A$ itself has zeros in it, but that's not what guarantees invertibility. The clean check is $\det(A^TA) = 4-1=3 \neq 0$.

---

### Step-by-step

$$A^TA = \begin{pmatrix}2&1\\1&2\end{pmatrix}, \qquad \det = 3, \qquad (A^TA)^{-1} = \tfrac{1}{3}\begin{pmatrix}2&-1\\-1&2\end{pmatrix} \checkmark$$

$$A(A^TA)^{-1} = \begin{pmatrix}1&0\\1&1\\0&1\end{pmatrix}\cdot\tfrac{1}{3}\begin{pmatrix}2&-1\\-1&2\end{pmatrix} = \tfrac{1}{3}\begin{pmatrix}2&-1\\1&1\\-1&2\end{pmatrix} \checkmark$$

$$P = \tfrac{1}{3}\begin{pmatrix}2&-1\\1&1\\-1&2\end{pmatrix}\begin{pmatrix}1&1&0\\0&1&1\end{pmatrix} = \begin{pmatrix}\tfrac{2}{3}&\tfrac{1}{3}&-\tfrac{1}{3}\\[4pt]\tfrac{1}{3}&\tfrac{2}{3}&\tfrac{1}{3}\\[4pt]-\tfrac{1}{3}&\tfrac{1}{3}&\tfrac{2}{3}\end{pmatrix} \checkmark$$

---

### Projection and residual

$$P\mathbf{b} = \begin{pmatrix}\tfrac{2}{3}+\tfrac{2}{3}-1\\[4pt]\tfrac{1}{3}+\tfrac{4}{3}+1\\[4pt]-\tfrac{1}{3}+\tfrac{2}{3}+2\end{pmatrix} = \begin{pmatrix}\tfrac{1}{3}\\[4pt]\tfrac{8}{3}\\[4pt]\tfrac{7}{3}\end{pmatrix} \checkmark$$

$$\mathbf{b} - P\mathbf{b} = \begin{pmatrix}1\\2\\3\end{pmatrix} - \begin{pmatrix}\tfrac{1}{3}\\\tfrac{8}{3}\\\tfrac{7}{3}\end{pmatrix} = \begin{pmatrix}\tfrac{2}{3}\\[4pt]-\tfrac{2}{3}\\[4pt]\tfrac{2}{3}\end{pmatrix} \checkmark$$

---

### Orthogonality verification

$$\mathbf{a}_1\cdot(\mathbf{b}-P\mathbf{b}) = (1,1,0)\cdot\left(\tfrac{2}{3},-\tfrac{2}{3},\tfrac{2}{3}\right) = \tfrac{2}{3}-\tfrac{2}{3}+0 = 0 \checkmark$$

$$\mathbf{a}_2\cdot(\mathbf{b}-P\mathbf{b}) = (0,1,1)\cdot\left(\tfrac{2}{3},-\tfrac{2}{3},\tfrac{2}{3}\right) = 0-\tfrac{2}{3}+\tfrac{2}{3} = 0 \checkmark$$

This is exactly the condition $A^T(\mathbf{b}-P\mathbf{b}) = \mathbf{0}$, which is the normal equation — the residual lies in the left nullspace of $A$, orthogonal to the entire column space.

---

### Two key properties to remember

$P$ is always **symmetric** ($P = P^T$) and **idempotent** ($P^2 = P$). The first reflects that projection is a self-adjoint operation; the second says projecting twice is the same as projecting once — once you're in the subspace, you stay there. These are worth verifying numerically (see code below).

```python
import numpy as np

A = np.array([[1,0],[1,1],[0,1]], dtype=float)
b = np.array([1,2,3], dtype=float)

P = A @ np.linalg.inv(A.T @ A) @ A.T
Pb = P @ b
residual = b - Pb

print("P =")
print(np.round(P, 6))
print()
print(f"Pb = {np.round(Pb, 6)}")
print(f"b - Pb = {np.round(residual, 6)}")
print()
print("Orthogonality checks:")
print(f"  a1 · residual = {np.dot(A[:,0], residual):.2e}")
print(f"  a2 · residual = {np.dot(A[:,1], residual):.2e}")
print()
print(f"Symmetric: P == P^T  →  {np.allclose(P, P.T)}")
print(f"Idempotent: P^2 == P →  {np.allclose(P @ P, P)}")
```

### Addendum — The projection matrix approach to least squares

#### The problem

Given a system $A\mathbf{x} = \mathbf{b}$ where $A$ is $m \times n$ with $m > n$ (more equations than unknowns), there is generally no exact solution. Instead we ask: **find $\hat{\mathbf{x}}$ such that $A\hat{\mathbf{x}}$ is as close to $\mathbf{b}$ as possible**, minimising the squared error $\|A\hat{\mathbf{x}} - \mathbf{b}\|^2$.

This is the least squares problem. In statistics, $A$ is the design matrix $X$, $\mathbf{b}$ is the response vector $\mathbf{y}$, and $\hat{\mathbf{x}}$ collects the regression coefficients $\hat{\boldsymbol{\beta}}$.

---

#### Why projection gives the answer

$\mathbf{b}$ is simply the observed data — your target vector as it is. It lives in the ambient space $\mathbb{R}^m$, and it will generally **not** lie in $\text{col}(A)$. That is precisely why no exact solution exists: $\mathbf{b}$ has components pointing outside the column space that no choice of coefficients can reach.

The set of all vectors of the form $A\mathbf{x}$ (as $\mathbf{x}$ ranges over $\mathbb{R}^n$) is exactly $\text{col}(A)$. So the question becomes: **which vector in $\text{col}(A)$ is closest to $\mathbf{b}$?**

The answer from geometry is the **orthogonal projection** of $\mathbf{b}$ onto $\text{col}(A)$:

$$P\mathbf{b} = A(A^TA)^{-1}A^T\mathbf{b}$$

Projection strips away every component of $\mathbf{b}$ that points outside $\text{col}(A)$ — the part orthogonal to the subspace. What remains, $P\mathbf{b}$, is the closest point inside $\text{col}(A)$ to $\mathbf{b}$, and the residual $\mathbf{b} - P\mathbf{b}$ is the irreducible error: no choice of coefficients can reduce it further, because $P\mathbf{b}$ is already the best the column space can offer.

**Why is it the minimum?** The residual $\mathbf{b} - P\mathbf{b}$ is perpendicular to $\text{col}(A)$, and a perpendicular is always the shortest path from a point to a subspace. Any other candidate vector in $\text{col}(A)$ would form the hypotenuse of a right triangle with the residual — necessarily longer.

$$\|\mathbf{b} - \mathbf{v}\|^2 = \|\mathbf{b} - P\mathbf{b}\|^2 + \|P\mathbf{b} - \mathbf{v}\|^2 \geq \|\mathbf{b} - P\mathbf{b}\|^2 \quad \text{for any } \mathbf{v} \in \text{col}(A)$$

This is Pythagoras applied to the right triangle formed by the residual and the deviation within the column space.

---

#### The key identity: $P\mathbf{b} = A\hat{\mathbf{x}}$

The projection can be written in two equivalent ways:

$$P\mathbf{b} = \underbrace{A(A^TA)^{-1}A^T}_{\text{projection matrix}}\mathbf{b} = A\underbrace{(A^TA)^{-1}A^T\mathbf{b}}_{\hat{\mathbf{x}}}$$

So $P\mathbf{b}$ is just $A$ applied to $\hat{\mathbf{x}} = (A^TA)^{-1}A^T\mathbf{b}$. In other words, **$\hat{\mathbf{x}}$ is the coefficient vector that expresses the projection in terms of the columns of $A$**.

The residual is therefore:

$$\mathbf{b} - P\mathbf{b} = \mathbf{b} - A\hat{\mathbf{x}}$$

which is the error you want to minimise.

---

#### From projection geometry to the normal equation

The defining property of orthogonal projection is that the residual is **perpendicular to the entire column space of $A$** — that is, perpendicular to every column of $A$:

$$A^T(\mathbf{b} - P\mathbf{b}) = \mathbf{0}$$

Substituting $P\mathbf{b} = A\hat{\mathbf{x}}$:

$$A^T(\mathbf{b} - A\hat{\mathbf{x}}) = \mathbf{0}$$

Expanding:

$$A^T\mathbf{b} - A^TA\hat{\mathbf{x}} = \mathbf{0}$$

$$\boxed{A^TA\,\hat{\mathbf{x}} = A^T\mathbf{b}}$$

This is the **normal equation**. Solve it for $\hat{\mathbf{x}}$ to get the least-squares solution. If $A^TA$ is invertible (i.e. $A$ has linearly independent columns), the unique solution is $\hat{\mathbf{x}} = (A^TA)^{-1}A^T\mathbf{b}$.

In statistical notation with design matrix $X$, response $\mathbf{y}$, and coefficients $\hat{\boldsymbol{\beta}}$, this is the same equation written as $X^T({\mathbf{y}} - X\hat{\boldsymbol{\beta}}) = \mathbf{0}$ or equivalently $X^TX\hat{\boldsymbol{\beta}} = X^T\mathbf{y}$.

---

#### Summary: two equivalent viewpoints

| Viewpoint | Language | Key equation |
|---|---|---|
| **Geometric** | Project $\mathbf{b}$ onto $\text{col}(A)$; residual $\perp$ column space | $A^T(\mathbf{b} - A\hat{\mathbf{x}}) = \mathbf{0}$ |
| **Algebraic** | Minimise $\|A\mathbf{x} - \mathbf{b}\|^2$; set gradient to zero | $A^TA\hat{\mathbf{x}} = A^T\mathbf{b}$ |

Both routes reach the same normal equation. The projection matrix makes the geometry explicit; the algebraic route is faster to apply. In Q4 you will work through a concrete least-squares example using the normal equation directly.

### Addendum — Solving the GLM via the projection matrix

The normal equation approach you used finds $\hat\beta$ directly. There is an equivalent route that builds the projection matrix $P$ first, obtains the fitted values $\hat{\mathbf{y}} = P\mathbf{y}$, and then recovers $\hat\beta$. Doing both makes the geometry concrete.

---

#### Step 1 — Build $P = X(X^TX)^{-1}X^T$

Using the matrices already computed:

$$X(X^TX)^{-1} = \begin{bmatrix}1&0\\1&1\\1&2\end{bmatrix} \cdot \frac{1}{6}\begin{bmatrix}5&-3\\-3&3\end{bmatrix} = \frac{1}{6}\begin{bmatrix}5&-3\\2&0\\-1&3\end{bmatrix}$$

$$P = X(X^TX)^{-1}X^T = \frac{1}{6}\begin{bmatrix}5&-3\\2&0\\-1&3\end{bmatrix}\begin{bmatrix}1&1&1\\0&1&2\end{bmatrix} = \frac{1}{6}\begin{bmatrix}5&2&-1\\2&2&2\\-1&2&5\end{bmatrix}$$

$P$ is $3\times 3$: it acts on vectors in $\mathbb{R}^3$ (the data space) and projects them onto the 2-dimensional column space of $X$ — the set of all exact linear fits.

---

#### Step 2 — Project $\mathbf{y}$ to get fitted values $\hat{\mathbf{y}}$

$$\hat{\mathbf{y}} = P\mathbf{y} = \frac{1}{6}\begin{bmatrix}5&2&-1\\2&2&2\\-1&2&5\end{bmatrix}\begin{bmatrix}1\\2\\2.5\end{bmatrix} = \frac{1}{6}\begin{bmatrix}5+4-2.5\\2+4+5\\-1+4+12.5\end{bmatrix} = \frac{1}{6}\begin{bmatrix}6.5\\11\\15.5\end{bmatrix} = \begin{bmatrix}\tfrac{13}{12}\\[4pt]\tfrac{11}{6}\\[4pt]\tfrac{31}{12}\end{bmatrix}$$

These are the model's predictions at $x=0,1,2$ respectively — the three points that lie exactly on the best-fit line.

---

#### Step 3 — Compute the residual and verify orthogonality

$$\mathbf{e} = \mathbf{y} - \hat{\mathbf{y}} = \begin{bmatrix}1\\2\\2.5\end{bmatrix} - \begin{bmatrix}\tfrac{13}{12}\\[4pt]\tfrac{11}{6}\\[4pt]\tfrac{31}{12}\end{bmatrix} = \begin{bmatrix}-\tfrac{1}{12}\\[4pt]\tfrac{1}{6}\\[4pt]-\tfrac{1}{12}\end{bmatrix}$$

The residual must be orthogonal to every column of $X$ — i.e. $X^T\mathbf{e} = \mathbf{0}$:

$$\text{col}_1: \quad (1,1,1)\cdot\mathbf{e} = -\tfrac{1}{12}+\tfrac{2}{12}-\tfrac{1}{12} = 0 \checkmark$$
$$\text{col}_2: \quad (0,1,2)\cdot\mathbf{e} = 0+\tfrac{1}{6}-\tfrac{2}{12} = \tfrac{2}{12}-\tfrac{2}{12} = 0 \checkmark$$

This orthogonality is not a numerical coincidence — it is the defining property of an orthogonal projection, and it is equivalent to the normal equations.

---

#### Step 4 — Recover $\hat\beta$ from $\hat{\mathbf{y}} = X\hat\beta$

The fitted values satisfy $\hat{\mathbf{y}} = X\hat\beta$ exactly (by construction — $\hat{\mathbf{y}}$ lies in $\text{col}(X)$). Apply the left inverse:

$$\hat\beta = (X^TX)^{-1}X^T\hat{\mathbf{y}} = (X^TX)^{-1}X^TP\mathbf{y}$$

But $X^TP = X^T$ (because $P$ projects onto $\text{col}(X)$ and $X^T$ already "sees" only that subspace), so this collapses back to:

$$\hat\beta = (X^TX)^{-1}X^T\mathbf{y} = \begin{bmatrix}\tfrac{13}{12}\\\tfrac{3}{4}\end{bmatrix}$$

Same answer, same formula — but now you can see *why*: $P$ replaces $\mathbf{y}$ with its projection, and the left inverse then reads off the coefficients from that projected vector.

---

#### Contrasting the two approaches

| | **Normal equations** | **Projection matrix** |
|---|---|---|
| **What you build** | $X^TX$ and $X^T\mathbf{y}$ — two small objects | $P = X(X^TX)^{-1}X^T$ — a full $n\times n$ matrix |
| **What you solve for directly** | $\hat\beta$ (the coefficients) | $\hat{\mathbf{y}}$ (the fitted values) |
| **Key step** | Solve $(X^TX)\hat\beta = X^T\mathbf{y}$ | Multiply $P\mathbf{y}$ |
| **Geometric emphasis** | Algebraic: residual $\perp$ column space | Explicit: residual as $\mathbf{y} - P\mathbf{y}$ |
| **Computational cost** | Cheap — only $p\times p$ system ($p$ = parameters) | Expensive — builds $n\times n$ matrix ($n$ = observations) |
| **When it's useful** | Fitting: you want $\hat\beta$ and will use it repeatedly | Geometry/theory: reasoning about the residual subspace, hat matrix, leverage |
| **In statistics** | Standard OLS implementation | $P$ is the *hat matrix* $H$ — $H_{ii}$ measures leverage of observation $i$ |

**The key insight:** both routes are two descriptions of the same geometric act — projecting $\mathbf{y}$ onto $\text{col}(X)$. The normal equations give you the coordinates of that projection in the basis of $X$'s columns; the projection matrix gives you the projected vector directly. In practice you almost always use the normal equations (or a numerically stable equivalent like QR decomposition); the projection matrix is most valuable as a conceptual and diagnostic tool.

```python
import numpy as np

X = np.array([[1,0],[1,1],[1,2]], dtype=float)
y = np.array([1, 2, 2.5])

# Projection matrix (hat matrix)
P = X @ np.linalg.inv(X.T @ X) @ X.T
print("P (hat matrix) =")
print(np.round(P * 6) / 6)   # show exact sixths
print()

y_hat = P @ y
residual = y - y_hat
print(f"ŷ = {np.round(y_hat, 6)}")
print(f"e = y - ŷ = {np.round(residual, 6)}")
print()
print("Orthogonality of residual to col(X):")
print(f"  col1 · e = {np.dot(X[:,0], residual):.2e}")
print(f"  col2 · e = {np.dot(X[:,1], residual):.2e}")
print()

# Recover beta from fitted values
beta = np.linalg.inv(X.T @ X) @ X.T @ y_hat
print(f"β recovered from ŷ: C0={beta[0]:.6f}, C1={beta[1]:.6f}")
print()

# Hat matrix properties
print(f"Symmetric:   P == P^T  → {np.allclose(P, P.T)}")
print(f"Idempotent:  P² == P   → {np.allclose(P @ P, P)}")
print(f"Diagonal of P (leverage values): {np.round(np.diag(P), 4)}")
print(f"  (sum of diagonal = rank(X) = {np.trace(P):.1f})")
```
