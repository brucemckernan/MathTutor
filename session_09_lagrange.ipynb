# Session 09 — Constrained Optimisation: Lagrange Multipliers

Focus: optimising a function subject to equality constraints using the method of Lagrange multipliers. The key equation is $\nabla f = \lambda \nabla g$ (gradients are parallel at a constrained extremum). Also: interpreting the Lagrange multiplier $\lambda$ as a sensitivity — how much the optimal value changes per unit change in the constraint.

## Q1 — Warmup: one constraint in $\mathbb{R}^2$

Find the point on the line $x + 2y = 5$ closest to the origin. Use Lagrange multipliers (not substitution). Verify your answer geometrically — the constraint and the level curves of $f$ should be tangent at the solution.

**Your answer:**
\begin{gather*}
& f(x, y)=x^2+y^2, \quad g(x, y)=x+2 y \\
& f_x=2 x \quad g_x=1 \\
& \begin{array}{ll}
f_y=2 y & g_y=2
\end{array} \\
& \nabla f(x, y)=2 x \hat{i}+2 y \hat{k} \quad \nabla g(x, y)=1 \hat{i}+2 \hat{k} \\
& \nabla f(x, y)=\lambda g(x, y) \\
& 2 x \hat{i}+2 y \hat{k}=\lambda(1 \hat{i}+2 \hat{k}) \\
& 2 x \hat{i}+2 y \hat{k}=\lambda \hat{i}+2 \lambda \hat{k} \\
& 2 x=\lambda, \quad 2 y=2 \lambda, \quad x+2 y=5 \\
& y=\lambda \quad x=5-2 y \\
& 2(5-2 y)=\lambda \\
& 2(5-2 \lambda)=\lambda \\
& 10-4 \lambda=\lambda \\
& \lambda=2 \\
& 2 x=2 \rightarrow x=1, y=\lambda \rightarrow y=2
\end{gather*}

I want to verify geometrically I take the point on the circle and work out the tangent line equation by working out the radius vector and then based on the assumption this is perpendicular to the tangent line at that point I set up the equation for a tangent line which when resolved is the same equation as the constraint line and therfore shares the same normal

\begin{gather*}
& p(1,2) \\

\end{gather*}


## Q2 — Optimisation on a sphere

Find the maximum and minimum values of $f(x, y, z) = x + 2y + 3z$ subject to the constraint $x^2 + y^2 + z^2 = 1$.

**Your answer:**

We have a 3D spherical constraint intersecting with a 4D object (in this case at it turns out they intersect at eight points). I made the spherical constraint into a funtion g(x), found the partial derivatives and constructed the gradients for f and g. Applied the Lagrange formula and worked the series of equations to find a value for $\lambda$ and then worked backwards to find values for x, y and z. Because these were all $\pm \sqrt{}$ I then used all combinations of the points of intersection with the original funtion f to find the values at those points and from that determined the largest and smallest value and thus the maxima and minima.

\begin{gather*}
&\begin{array}{ll}
f(x, y, z)=x+2 y+3 z & x^2+y^2+z^2=1 \\
& g(x, y, z)=x^2+y^2+z^2 \\
f_x=1 & g_x=2 x \\
f_y=2 & g_y=2 y \\
f_z=3 & g_z=2 z \\
\nabla f(x, y, z)=1 \hat{i}+2 \hat{j}+3 \hat{k} & \nabla g(x, y, z)=2 x \hat{i}+2 y \hat{j}+2 z \hat{k}
\end{array}\\
& \nabla f(x, y, z)=\lambda \nabla g(x, y, z) \\
& 1 \hat{i}+2 \hat{j}+3 \hat{k}=\lambda(2 x \hat{i}+2 y \hat{j}+2 z \hat{k}) \\
& 1 \hat{i}+2 \hat{j}+3 \hat{k}=2 x \lambda \hat{i}+2 y \lambda \hat{j}+2 z \lambda \hat{k} \\
& 1=2 x \lambda \quad x=\frac{1}{2 \lambda} \\
& 2=2 y \lambda \quad y=\frac{2}{2 \lambda}=\frac{1}{\lambda} \\
& 3=2 z \lambda \quad z=\frac{3}{2 \lambda} \\
& x^2+y^2+z^2=1 \\
& 1=\left(\frac{1}{2 \lambda}\right)^2+\left(\frac{1}{\lambda}\right)^2+\left(\frac{3}{2 \lambda}\right)^2 \\
& 1=\frac{1}{4 \lambda^2}+\frac{1}{\lambda^2}+\frac{9}{4 \lambda^2} \\
& 1=\frac{1}{\lambda^2}\left(\frac{1}{4}+\frac{1}{1}+\frac{9}{4}\right) \\
& \lambda^2=\left(\frac{1}{4}+\frac{4}{4}+\frac{9}{4}\right) \\
& \lambda^2=\frac{7}{2} \\
& \lambda= \pm \sqrt{\frac{7}{2}} \\
& x= \pm \frac{1}{2 \sqrt{\frac{7}{2}}} \cdot \frac{\sqrt{\frac{7}{2}}}{\sqrt{\frac{7}{2}}}= \pm \frac{\sqrt{\frac{7}{2}}}{7} \\
& y= \pm \frac{1}{\sqrt{\frac{7}{2}}} \cdot \frac{\sqrt{\frac{7}{2}}}{\sqrt{\frac{7}{2}}}= \pm \frac{\sqrt{\frac{7}{2}}}{\frac{7}{2}}= \pm \frac{2 \sqrt{\frac{7}{2}}}{7} \\
& z= \pm \frac{3}{2 \sqrt{7 / 2}} \cdot \frac{\sqrt{\frac{7}{2}}}{\sqrt{7 / 2}}= \pm \frac{3 \sqrt{\frac{7}{2}}}{7}= \pm 3 \frac{\sqrt{7 / 2}}{7} \\
& f(x, y, z)=x+2 y+3 z \\
& \quad x= \pm \frac{\sqrt{\frac{7}{2}}}{7}, y= \pm \frac{2 \sqrt{\frac{7}{2}}}{7} \quad z=\pm \frac{3 \sqrt{7 / 2}}{7} \\
& f\left(\frac{\sqrt{\frac{7}{2}}}{7}, \frac{2 \sqrt{\frac{7}{2}}}{7}, \frac{3 \sqrt{\frac{7}{2}}}{7}\right)=2 \sqrt{7 / 2} \\
& f\left(\frac{\sqrt{\frac{7}{2}}}{7}, \frac{2 \sqrt{\frac{7}{2}}}{7},-\frac{3 \sqrt{7 / 2}}{7}\right)= -\frac{4 \sqrt{7 / 2}}{7} \\
& f\left(\frac{\sqrt{\frac{7}{2}}}{7},-2 \frac{\sqrt{\frac{7}{2}}}{7}, \frac{3 \sqrt{7 / 2}}{7}\right)=\frac{6 \sqrt{7 / 2}}{7} \\
& f\left(\frac{\sqrt{\frac{7}{2}}}{7},-\frac{2 \sqrt{\frac{7}{2}}}{7},-\frac{3 \sqrt{7 / 2}}{7}\right)=-12 \frac{\sqrt{7 / 2}}{7} \\
& f\left(-\frac{\sqrt{\frac{7}{2}}}{7}, \frac{2 \sqrt{7 / 2}}{7}, \frac{3 \sqrt{7 / 2}}{7}\right)=12 \frac{\sqrt{7 / 2}}{7} \\
& f\left(-\frac{\sqrt{7 / 2}}{7}, \frac{2 \sqrt{7 / 2}}{7},-\frac{3 \sqrt{7 / 2}}{7}\right)=-6 \frac{\sqrt{7 / 2}}{7} \\
& f\left(-\frac{\sqrt{7 / 2}}{7},-2 \frac{\sqrt{7 / 2}}{7}, \frac{3 \sqrt{7 / 2}}{7}\right)=4 \frac{\sqrt{7 / 2}}{7} \\
& f\left(-\frac{\sqrt{7 / 2}}{7},-\frac{2 \sqrt{7 / 2}}{7},-\frac{3 \sqrt{7 / 2}}{7}\right)=-2 \sqrt{7 / 2}
\end{gather*}

Maximum value at:
$$f\left(\frac{\sqrt{\frac{7}{2}}}{7}, \frac{2 \sqrt{\frac{7}{2}}}{7}, \frac{3 \sqrt{\frac{7}{2}}}{7}\right)=2 \sqrt{7 / 2}$$

Minimum value at:
$$f\left(-\frac{\sqrt{7 / 2}}{7},-\frac{2 \sqrt{7 / 2}}{7},-\frac{3 \sqrt{7 / 2}}{7}\right)=-2 \sqrt{7 / 2}$$

# Q3 — Two constraints

Find the point on the intersection of the planes $x + y + z = 1$ and $x + 2y = 0$ that is closest to the origin. (This requires two Lagrange multipliers, one for each constraint.)

**Your answer:**

As previously the closest point to the orgin is going to be tangent to a sphere centered at the origin. therefore I make a constraint function $f(x,y,z) = x^2 + y ^ 2 + z^2$. I also know I have to introduce multiple  $\lambda$ as I have more than one constraint (in this case 2 constraints that form a line of intersection between two planes). I know that $\nabla f(x, y, z)=\lambda_1 \nabla g(x, y, z)+\lambda_2 \nabla h(x, y, z)$ and therefore work out the gradients of my three equations and complete this formula. I then have to work through a system of equations to find both $\lambda_1 and \lambda_2$ and then work backwards to find the x, y, z components of the point.1

\begin{gather*}
& f(x, y, z)=x^2+y^2+z^2 \\
& g(x, y, z)=x+y+z \\
& h(x, y, z)=x+2 y \\
& f_x=2 x \quad g_x=1 \quad h_x=1 \\
& f_y=2 y \quad g_y=1 \quad h_y=1 \\
& f_z=2 z \quad g_z=1 \quad h_z=0 \\
& \nabla f(x, y, z)=\lambda_1 \nabla g(x, y, z)+\lambda_2 \nabla h(x, y, z) \\
& 2 x \hat{i}+2 y \hat{j}+2 z \hat{k}=\lambda_1(1 \hat{i}+1 \hat{j}+1 \hat{k})+\lambda_2(1 \hat{i}+1 \hat{j}) \\
& 2 x \hat{i}+2 y \hat{j}+2 z \hat{k}=\lambda_1 \hat{i}+\lambda_1 \hat{j}+\lambda_1 \hat{k}+\lambda_2 \hat{i}+\lambda_2 \hat{j} \\
&\begin{array}{ll} \\
2 x=\lambda_1+\lambda_2 & x=\frac{\lambda_1+\lambda_2}{2} \\
2 y=\lambda_1+\lambda_2 & y=\frac{\lambda_1+\lambda_2}{2} \\
2 z=\lambda_1 & z=\frac{\lambda_1}{2} \\
x+y+z=1 & x+2 y=0 \\
\frac{1}{2}\left(\lambda_1+\lambda_2+\lambda_1+\lambda_2+\lambda_1\right)=1 \\
\frac{1}{2}\left(3 \lambda_1+2 \lambda_2\right)=1 \\
\frac{3 \lambda_1}{2}+\lambda_2=1 \\
\lambda_2=1-\frac{3 \lambda_1}{2} \\
\lambda_1=\frac{2}{3}\left(1-\lambda_2\right) \\
\lambda_2=1-\frac{3}{2} \cdot \frac{\frac{2}{3}\left(1-\lambda_2\right)}{2} \\
\lambda_2=1-\frac{1}{2}\left(1-\lambda_2\right) \\
\lambda_2=1-\frac{1}{2}-\frac{1}{2} \lambda_2
\end{array}\\
& \lambda_2=\frac{1}{2}-\frac{1}{2} \lambda_2 \\
& \lambda_2+\frac{1}{2} \lambda_2=\frac{1}{2} \\
& \frac{3}{2} \lambda_2=\frac{1}{2} \\
& \lambda_2=\frac{1}{3} \\
& \lambda_1=\frac{2}{3}\left(1-\lambda_2\right) \\
& \lambda_1=\frac{2}{3}\left(1-\frac{1}{3}\right) \\
& \lambda_1=\frac{2}{3} \cdot \frac{2}{3}=\frac{4}{9} \\
& x=\frac{\lambda_1+\lambda_2}{2}=\frac{\frac{4}{9}+\frac{1}{3}}{2}=\frac{7}{9} \cdot \frac{1}{2}=\frac{7}{18} \\
& y=\frac{\lambda_1+\lambda_2}{2}=\frac{\frac{4}{9}+\frac{1}{3}}{2}=\frac{7}{9} \cdot \frac{1}{2}=\frac{7}{18} \\
& z=\frac{\lambda_1}{2}=\frac{\frac{4}{9}}{2}=\frac{4}{9} \cdot \frac{1}{2}=\frac{2}{9} \\
& \left(\frac{7}{18}, \frac{7}{18}, \frac{2}{9}\right)
\end{gather*}



## Q4 — Interpreting $\lambda$

In Q1, you found the minimum value of $f(x,y) = x^2 + y^2$ subject to $x + 2y = 5$.

(a) What is the value of $\lambda$ you found?

(b) Now change the constraint to $x + 2y = 6$ and find the new minimum (you do not need to redo the full calculation — use $\lambda$).

(c) Explain in one or two sentences what $\lambda$ measures in general.

**Your answer:**

(a) in Q1 I found that $\lambda = 2$

(b) if I use $x + 2y = 6$ then I know that $y = \lambda = 2$, and $ x = 6 - 2\lambda - 2$  and the new minimum point is (2,2)

(c) lambda is telling us how much the "maximum/minimum" will change for a unit increase/decrease in $\lambda$ in this case the extrema will move to new coordinates at (2,2) with a value of v + \lambda where v is the oringal maximim/minimum under the orginal constraint
