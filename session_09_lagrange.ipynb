Focus: optimising a function subject to equality constraints using the method of Lagrange multipliers. The key equation is $\nabla f = \lambda \nabla g$ (gradients are parallel at a constrained extremum). Also: interpreting the Lagrange multiplier $\lambda$ as a sensitivity — how much the optimal value changes per unit change in the constraint.

& f(x, y)=x^2+y^2, \quad g(x, y)=x+2 y \\

f_y=2 y & g_y=2
\end{array} \\
& \nabla f(x, y)=2 x \hat{i}+2 y \hat{k} \quad \nabla g(x, y)=1 \hat{i}+2 \hat{k} \\
& \nabla f(x, y)=\lambda g(x, y) \\
& 2 x \hat{i}+2 y \hat{k}=\lambda(1 \hat{i}+2 \hat{k}) \\
& 2 x \hat{i}+2 y \hat{k}=\lambda \hat{i}+2 \lambda \hat{k} \\
& 2 x=\lambda, \quad 2 y=2 \lambda, \quad x+2 y=5 \\
& y=\lambda \quad x=5-2 y \\
& 2(5-2 y)=\lambda \\
& 2(5-2 \lambda)=\lambda \\
& 10-4 \lambda=\lambda \\
& \lambda=2 \\
& 2 x=2 \rightarrow x=1, y=\lambda \rightarrow y=2
\end{gather*}
