## Q7 — Key identities and series for Fourier analysis

This question builds the series vocabulary used directly in Fourier analysis. No prior knowledge of Fourier theory is needed — everything follows from the power series tools above.

**(a) Euler's formula via the exponential series.**

The series $e^x = \displaystyle\sum_{n=0}^\infty \frac{x^n}{n!}$ converges absolutely for all real $x$, and the same is true for complex $x$. Substitute $x = i\theta$ where $i^2 = -1$, and use the repeating pattern $i^0 = 1,\ i^1 = i,\ i^2 = -1,\ i^3 = -i,\ i^4 = 1, \ldots$ to separate the series into real and imaginary parts. Show that:
$$e^{i\theta} = \cos\theta + i\sin\theta$$
Write down the Maclaurin series for $\cos\theta$ and $\sin\theta$ that emerge naturally from this separation.

**(b) Comparison test for series of functions.**

A Fourier series involves terms like $\dfrac{\cos(n\theta)}{n^2}$ or $\dfrac{\sin(n\theta)}{n}$.

- Show that $\displaystyle\sum_{n=1}^\infty \frac{\cos(n\theta)}{n^2}$ converges **absolutely** for all $\theta \in \mathbb{R}$, by exhibiting a convergent majorant series.
- Explain why the same approach fails for $\displaystyle\sum_{n=1}^\infty \frac{\cos(n\theta)}{n}$, and what this tells you about the convergence of the second series.

**(c) The Basel problem and its relatives.**

Euler proved in 1734 that $\displaystyle\sum_{n=1}^\infty \frac{1}{n^2} = \frac{\pi^2}{6}$. This result is central to Fourier analysis — it emerges naturally from Parseval's identity applied to a specific Fourier series. Accept it here and use it to evaluate:

- (i) $\displaystyle\sum_{n=1}^\infty \frac{1}{(2n)^2}$ (sum over even indices)
- (ii) $\displaystyle\sum_{n=1}^\infty \frac{1}{(2n-1)^2}$ (sum over odd indices only)

*Hint for (ii):* Split $\displaystyle\sum_{n=1}^\infty \frac{1}{n^2}$ into its even-indexed and odd-indexed terms and use (i).

**(d) Telescoping and index shifting.**

- Show that $\displaystyle\sum_{n=1}^N\left(\frac{1}{n} - \frac{1}{n+1}\right)$ telescopes, and find its closed form for finite $N$.
- Hence find the sum of $\displaystyle\sum_{n=1}^\infty\left(\frac{1}{n} - \frac{1}{n+1}\right)$.
- Rewrite $\displaystyle\sum_{n=0}^\infty a_{n+2}$ as a sum starting at $n=2$. More generally, if $\displaystyle\sum_{n=0}^\infty a_n$ converges to $S$, express $\displaystyle\sum_{n=0}^\infty a_{n+k}$ in terms of $S$ and the first $k$ terms.
