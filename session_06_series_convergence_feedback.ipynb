### Part (a) — $\displaystyle\sum_{n=1}^\infty \frac{n^2}{2^n}$

#### Divergence test — ✓ Conclusion correct, presentation muddled

You concluded the limit is 0 and the test is inconclusive — both correct. But the intermediate step you wrote,

$$\frac{n^2}{2^n} = \frac{\frac{1}{2^n}}{n^2}$$

is not a valid equality (the right side is $\frac{1}{n^2 \cdot 2^n}$, not $\frac{n^2}{2^n}$). The cleanest way to see the limit is 0 is simply that exponentials dominate polynomials:

$$\lim_{n \to \infty} \frac{n^2}{2^n} = 0 \qquad \text{(exponential beats polynomial)}$$

or, if you want to show the work, apply L'Hôpital twice treating $n$ as continuous.

#### Ratio test — ✓ Setup correct, ✗ one algebraic error

Your setup and cancellation of powers of 2 are correct:

$$\left|\frac{a_{n+1}}{a_n}\right| = \frac{(n+1)^2}{2^{n+1}} \cdot \frac{2^n}{n^2} = \frac{(n+1)^2}{2n^2}$$

Then you wrote this as $\left(\dfrac{n+1}{2n}\right)^2$. This is wrong:

$$\left(\frac{n+1}{2n}\right)^2 = \frac{(n+1)^2}{4n^2} \neq \frac{(n+1)^2}{2n^2}$$

The correct factoring is:

$$\frac{(n+1)^2}{2n^2} = \frac{1}{2} \cdot \left(\frac{n+1}{n}\right)^2 = \frac{1}{2}\left(1 + \frac{1}{n}\right)^2$$

Taking the limit:

$$L = \lim_{n \to \infty} \frac{1}{2}\left(1 + \frac{1}{n}\right)^2 = \frac{1}{2} \cdot 1^2 = \frac{1}{2}$$

The limit is $\frac{1}{2}$, not $\frac{1}{4}$. Since $L = \frac{1}{2} < 1$, the conclusion — **the series converges absolutely** — is correct despite the error.

### Part (b) — $\displaystyle\sum_{n=1}^\infty \frac{n!}{n^n}$

#### Divergence test — ✗ Wrong on every step; conclusion reversed

You claimed:

$$\lim_{n \to \infty}\frac{n!}{n^n} = \frac{(n-1)!}{n} = \frac{1 - \frac{1}{n}}{1} = 1$$

None of these steps are valid:

1. $\frac{n!}{n^n} \neq \frac{(n-1)!}{n}$. These are not equal for any $n$. (For example at $n=3$: $\frac{6}{27} \neq \frac{2}{3}$.)
2. $\frac{(n-1)!}{n}$ does not simplify to $\frac{1 - 1/n}{1}$ — factorials do not behave like polynomials.

**The true limit.** Write out what $\frac{n!}{n^n}$ actually looks like:

$$\frac{n!}{n^n} = \frac{1 \cdot 2 \cdot 3 \cdots n}{n \cdot n \cdot n \cdots n} = \frac{1}{n} \cdot \frac{2}{n} \cdot \frac{3}{n} \cdots \frac{n}{n}$$

Every factor is $\leq 1$, and the first factor is $\frac{1}{n} \to 0$. So:

$$0 \leq \frac{n!}{n^n} \leq \frac{1}{n} \to 0$$

The limit is **0**, not 1. The divergence test is **inconclusive** — you cannot stop here.

#### Correct approach: ratio test

$$\frac{a_{n+1}}{a_n} = \frac{(n+1)!}{(n+1)^{n+1}} \cdot \frac{n^n}{n!} = \frac{(n+1) \cdot n^n}{(n+1)^{n+1}} = \frac{n^n}{(n+1)^n} = \left(\frac{n}{n+1}\right)^n = \left(1 - \frac{1}{n+1}\right)^n \to \frac{1}{e}$$

Since $L = \frac{1}{e} < 1$, the series **converges absolutely** — the opposite of your conclusion.

The $\left(1 - \frac{1}{n}\right)^n \to e^{-1}$ limit is worth knowing by heart; it appears frequently with factorials.

| Claimed limit = 1 | ✗ | True limit is 0 |
| "Divergence test conclusive" | ✗ | Limit = 0 means inconclusive |
| Conclusion: diverges | ✗ | Series converges; $L = 1/e < 1$ by ratio test |

### Part (c) — $\displaystyle\sum_{n=1}^\infty \frac{n^3}{3^n}$

#### Divergence test — ✓ Good; L'Hôpital chain is correct

Listing terms to build intuition is a good habit. Two denominators are slightly off:

- $n = 3$: $3^3 = 27$, not 36, so the term is $\frac{27}{27} = 1$ (same value by coincidence, but the denominator is wrong)
- $n = 4$: $3^4 = 81$, not 144, so the term is $\frac{64}{81}$, not $\frac{64}{144}$

The L'Hôpital chain is correct and well-presented — much cleaner than (a):

$$\frac{n^3}{3^n} \xrightarrow{L'} \frac{3n^2}{3^n \ln 3} \xrightarrow{L'} \frac{6n}{3^n(\ln 3)^2} \xrightarrow{L'} \frac{6}{3^n(\ln 3)^3} \to 0$$

Limit is 0, divergence test inconclusive. ✓

#### Ratio test — ✓ Setup correct, ✗ same factoring error as (a)

Cancelling $3^n$ from $\frac{3^n}{3^{n+1}}$ correctly gives $\frac{(n+1)^3}{3n^3}$. ✓

Then you factored this as $\left(\dfrac{n+1}{3n}\right)^3$. **This is wrong** — the same pattern as (a):

$$\left(\frac{n+1}{3n}\right)^3 = \frac{(n+1)^3}{27n^3} \neq \frac{(n+1)^3}{3n^3}$$

Putting the 3 inside the bracket before cubing multiplies it by itself three times ($3^3 = 27$), inflating the denominator ninefold.

**The rule:** when you have $\dfrac{(n+1)^k}{c \cdot n^k}$, the constant $c$ stays **outside** the power:

$$\frac{(n+1)^k}{c \cdot n^k} = \frac{1}{c}\left(\frac{n+1}{n}\right)^k$$

not $\left(\dfrac{n+1}{cn}\right)^k = \dfrac{(n+1)^k}{c^k n^k}$.

Applied here:

$$\frac{(n+1)^3}{3n^3} = \frac{1}{3}\left(\frac{n+1}{n}\right)^3 = \frac{1}{3}\left(1 + \frac{1}{n}\right)^3 \to \frac{1}{3} \cdot 1 = \frac{1}{3}$$

The correct limit is $L = \dfrac{1}{3}$, not $\dfrac{1}{27}$.

Since $\frac{1}{3} < 1$, the conclusion — **the series converges absolutely** — is correct. But the limit value is wrong for the second time due to the same error.

#### The pattern to fix

| Part | After cancellation | Wrong factoring | Correct factoring | Wrong $L$ | Correct $L$ |

| (a) | $\dfrac{(n+1)^2}{2n^2}$ | $\left(\dfrac{n+1}{2n}\right)^2 = \dfrac{(n+1)^2}{4n^2}$ | $\dfrac{1}{2}\left(\dfrac{n+1}{n}\right)^2$ | $\dfrac{1}{4}$ | $\dfrac{1}{2}$ |
| (c) | $\dfrac{(n+1)^3}{3n^3}$ | $\left(\dfrac{n+1}{3n}\right)^3 = \dfrac{(n+1)^3}{27n^3}$ | $\dfrac{1}{3}\left(\dfrac{n+1}{n}\right)^3$ | $\dfrac{1}{27}$ | $\dfrac{1}{3}$ |

The fix is always the same: pull the constant out as a separate factor, then write the remaining fraction to the power.

**Correct.**

Your ratio test:

$$\frac{a_{n+1}}{a_n} = \frac{(n+1)!}{5^{n+1}} \cdot \frac{5^n}{n!} = \frac{(n+1) \cdot n!}{5 \cdot 5^n} \cdot \frac{5^n}{n!} = \frac{n+1}{5}$$

### Part (f) — $\displaystyle\sum_{n=1}^\infty \frac{(n+1)!}{n^n}$

**Two errors; wrong conclusion. The series converges.**

#### Error 1 — dropped a factor of $(n+1)$ from the denominator

Your cancellation of $(n+2)!/(n+1)! = n+2$ is correct. But $(n+1)^{n+1} = (n+1)^n \cdot (n+1)$, so:

$$\frac{a_{n+1}}{a_n} = \frac{(n+2)!}{(n+1)^{n+1}} \cdot \frac{n^n}{(n+1)!} = \frac{(n+2) \cdot n^n}{(n+1)^{n+1}} = \frac{(n+2) \cdot n^n}{(n+1)^n \cdot (n+1)}$$

You wrote $\dfrac{n^n(n+2)}{(n+1)^n}$, omitting the extra $(n+1)$ in the denominator. That factor matters: it keeps the numerator and denominator growing at the same rate, rather than letting the numerator escape to infinity.

#### Error 2 — $1^\infty$ is not 1

You wrote: "the $\frac{1}{n} = 0$ meaning the denominator becomes $1^\infty$ which is still 1."

$1^\infty$ is an indeterminate form — you cannot substitute $\frac{1}{n} = 0$ inside the exponent while keeping the exponent $n \to \infty$. The two limits compete. The true result is one of the most important limits in analysis:

$$\lim_{n \to \infty}\left(1 + \frac{1}{n}\right)^n = e \approx 2.718$$

You saw this limit in part (b) in its mirror form $\left(1 - \frac{1}{n+1}\right)^n \to e^{-1}$. It's the same family.

#### Correct limit

Rewriting the ratio with both factors explicit:

$$\frac{a_{n+1}}{a_n} = \frac{n+2}{n+1} \cdot \left(\frac{n}{n+1}\right)^n = \frac{n+2}{n+1} \cdot \left(1 - \frac{1}{n+1}\right)^n$$

As $n \to \infty$:
- $\dfrac{n+2}{n+1} \to 1$
- $\left(1 - \dfrac{1}{n+1}\right)^n \to e^{-1}$

$$L = 1 \cdot \frac{1}{e} = \frac{1}{e} \approx 0.368 < 1 \implies \textbf{converges absolutely}$$

This is the same $L = 1/e$ result as part (b). A useful pattern: any time the ratio collapses to something of the form $\left(\frac{n}{n+1}\right)^n$ or $\left(1 \pm \frac{c}{n}\right)^n$, the answer involves $e$.

#### Summary for (f)

| Step | Mark | Note |
|---|---|---|
| Ratio test setup | ✓ | |
| $(n+2)!/(n+1)! = n+2$ | ✓ | |
| Simplification: missing $(n+1)$ factor | ✗ | Denominator should be $(n+1)^{n+1}$, not $(n+1)^n$ |
| $(1 + 1/n)^n \to 1$ | ✗ | $1^\infty$ indeterminate; true limit is $e$ |
| Conclusion: diverges | ✗ | Series converges; $L = 1/e < 1$ |

```python
import numpy as np
from math import factorial

# --- Part (a) verification ---
ns = np.arange(1, 51)
terms_a = ns**2 / 2**ns
print("Part (a): n^2 / 2^n")
print(f"  Sum of first 50 terms: {np.cumsum(terms_a)[-1]:.8f}  (exact: 6)")
ratios_a = (ns[1:] + 1)**2 / (2 * ns[1:]**2)
print(f"  Ratio at n=50: {ratios_a[-1]:.6f}  (limit = 1/2 = 0.5)")

print()

# --- Part (b) verification ---
terms_b = [factorial(n) / n**n for n in range(1, 21)]
print("Part (b): n! / n^n")
print(f"  Partial sum (n=1..20): {sum(terms_b):.8f}  (converges)")
ratios_b = [(n / (n + 1))**n for n in range(1, 21)]
print(f"  Ratio at n=20: {ratios_b[-1]:.6f}  (limit = 1/e = {1/np.e:.6f})")

print()

# --- Part (c) verification ---
terms_c = ns**3 / 3**ns
print("Part (c): n^3 / 3^n")
print(f"  First 6 terms: {[f'{n**3}/{3**n}' for n in range(1,7)]}")
print(f"  Decimal:       {[round(n**3/3**n, 4) for n in range(1,7)]}")
print(f"  Sum of first 50 terms: {np.cumsum(terms_c)[-1]:.8f}")
ratios_c = (ns[1:] + 1)**3 / (3 * ns[1:]**3)
print(f"  Ratio at n=50: {ratios_c[-1]:.6f}  (limit = 1/3 = {1/3:.6f}, NOT 1/27 = {1/27:.6f})")

print()

# --- Part (d) verification ---
terms_d = [5**n / factorial(n) for n in range(1, 51)]
print("Part (d): 5^n / n!")
print(f"  Partial sum (n=1..50): {sum(terms_d):.8f}  (converges; exact value = e^5 - 1 = {np.e**5 - 1:.6f})")
ratios_d = [5 / (n + 1) for n in range(1, 21)]
print(f"  Ratio at n=20: {ratios_d[-1]:.6f}  (limit = 0)")

print()

# --- Part (e) verification ---
terms_e = [factorial(n) / 5**n for n in range(1, 16)]
print("Part (e): n! / 5^n")
print(f"  First 6 terms: {[round(factorial(n)/5**n, 4) for n in range(1,7)]}  (growing without bound)")
ratios_e = [(n + 1) / 5 for n in range(1, 16)]
print(f"  Ratio at n=14: {ratios_e[-1]:.4f}  (limit = infinity)")

print()

# --- Part (f) verification ---
print("Part (f): (n+1)! / n^n")
terms_f = [factorial(n + 1) / (n ** n) for n in range(1, 31)]
print(f"  Partial sum (n=1..30): {sum(terms_f):.8f}  (converges)")
# Show the ratio converging to 1/e
ratios_f = [(n + 2) / (n + 1) * (n / (n + 1))**n for n in range(1, 31)]
print(f"  Ratio at n=30: {ratios_f[-1]:.6f}  (limit = 1/e = {1/np.e:.6f})")
print(f"  (1 + 1/n)^n at n=100: {(1 + 1/100)**100:.6f}  →  e = {np.e:.6f}  (NOT 1)")
```
