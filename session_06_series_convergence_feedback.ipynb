### Part (a) — $\displaystyle\sum_{n=1}^\infty \frac{n^2}{2^n}$

#### Divergence test — ✓ Conclusion correct, presentation muddled

You concluded the limit is 0 and the test is inconclusive — both correct. But the intermediate step you wrote,

$$\frac{n^2}{2^n} = \frac{\frac{1}{2^n}}{n^2}$$

is not a valid equality (the right side is $\frac{1}{n^2 \cdot 2^n}$, not $\frac{n^2}{2^n}$). The cleanest way to see the limit is 0 is simply that exponentials dominate polynomials:

$$\lim_{n \to \infty} \frac{n^2}{2^n} = 0 \qquad \text{(exponential beats polynomial)}$$

or, if you want to show the work, apply L'Hôpital twice treating $n$ as continuous.

#### Ratio test — ✓ Setup correct, ✗ one algebraic error

Your setup and cancellation of powers of 2 are correct:

$$\left|\frac{a_{n+1}}{a_n}\right| = \frac{(n+1)^2}{2^{n+1}} \cdot \frac{2^n}{n^2} = \frac{(n+1)^2}{2n^2}$$

Then you wrote this as $\left(\dfrac{n+1}{2n}\right)^2$. This is wrong:

$$\left(\frac{n+1}{2n}\right)^2 = \frac{(n+1)^2}{4n^2} \neq \frac{(n+1)^2}{2n^2}$$

The correct factoring is:

$$\frac{(n+1)^2}{2n^2} = \frac{1}{2} \cdot \left(\frac{n+1}{n}\right)^2 = \frac{1}{2}\left(1 + \frac{1}{n}\right)^2$$

Taking the limit:

$$L = \lim_{n \to \infty} \frac{1}{2}\left(1 + \frac{1}{n}\right)^2 = \frac{1}{2} \cdot 1^2 = \frac{1}{2}$$

The limit is $\frac{1}{2}$, not $\frac{1}{4}$. Since $L = \frac{1}{2} < 1$, the conclusion — **the series converges absolutely** — is correct despite the error.

The intuition you gave at the end ("the difference between terms is decreasing by an order of magnitude 2 for each step") is essentially right: each term is roughly $\frac{1}{2}$ the size of the previous one, so the series behaves like a geometric series with ratio $\frac{1}{2}$.

### Part (b) — $\displaystyle\sum_{n=1}^\infty \frac{n!}{n^n}$

#### Divergence test — ✗ Wrong on every step; conclusion reversed

You claimed:

$$\lim_{n \to \infty}\frac{n!}{n^n} = \frac{(n-1)!}{n} = \frac{1 - \frac{1}{n}}{1} = 1$$

None of these steps are valid:

1. $\frac{n!}{n^n} \neq \frac{(n-1)!}{n}$. These are not equal for any $n$. (For example at $n=3$: $\frac{6}{27} \neq \frac{2}{3}$.)
2. $\frac{(n-1)!}{n}$ does not simplify to $\frac{1 - 1/n}{1}$ — factorials do not behave like polynomials.

**The true limit.** Write out what $\frac{n!}{n^n}$ actually looks like:

$$\frac{n!}{n^n} = \frac{1 \cdot 2 \cdot 3 \cdots n}{n \cdot n \cdot n \cdots n} = \frac{1}{n} \cdot \frac{2}{n} \cdot \frac{3}{n} \cdots \frac{n}{n}$$

Every factor is $\leq 1$, and the first factor is $\frac{1}{n} \to 0$. So:

$$0 \leq \frac{n!}{n^n} \leq \frac{1}{n} \to 0$$

The limit is **0**, not 1. The divergence test is **inconclusive** — you cannot stop here.

#### Correct approach: ratio test

$$\frac{a_{n+1}}{a_n} = \frac{(n+1)!}{(n+1)^{n+1}} \cdot \frac{n^n}{n!} = \frac{(n+1) \cdot n^n}{(n+1)^{n+1}} = \frac{n^n}{(n+1)^n} = \left(\frac{n}{n+1}\right)^n$$

Rewrite as:

$$\left(\frac{n}{n+1}\right)^n = \left(1 - \frac{1}{n+1}\right)^n$$

As $n \to \infty$, this is the standard limit $\left(1 - \frac{1}{n}\right)^n \to e^{-1}$:

$$L = \lim_{n \to \infty}\left(1 - \frac{1}{n+1}\right)^n = e^{-1} = \frac{1}{e} \approx 0.368$$

Since $L = \frac{1}{e} < 1$, the series **converges absolutely**.

This is the opposite of your conclusion. The error cascaded from the invalid divergence test calculation — once that limit was wrong, everything that followed was wrong.

#### The $e^{-1}$ limit — where it comes from

The standard result is $\lim_{n\to\infty}\left(1 + \frac{x}{n}\right)^n = e^x$. Setting $x = -1$ gives $\left(1 - \frac{1}{n}\right)^n \to e^{-1}$. The shift from $n$ to $n+1$ in the denominator makes no difference to the limit. This limit is worth knowing by heart as it appears frequently in ratio test calculations involving factorials.

#### Summary for (b)

| Step | Mark | Note |
|---|---|---|
| $\frac{n!}{n^n} = \frac{(n-1)!}{n}$ | ✗ | Not a valid equality |
| $\frac{(n-1)!}{n} = \frac{1-1/n}{1}$ | ✗ | Factorials do not simplify this way |
| Claimed limit = 1 | ✗ | True limit is 0 |
| "Divergence test conclusive" | ✗ | Limit = 0 means inconclusive |
| Conclusion: diverges | ✗ | Series converges; $L = 1/e < 1$ by ratio test |

```python
import numpy as np

# --- Part (a) verification ---
# Partial sums of n^2 / 2^n
ns = np.arange(1, 51)
terms_a = ns**2 / 2**ns
partial_sums_a = np.cumsum(terms_a)
print("Part (a): n^2 / 2^n")
print(f"  Sum of first 50 terms: {partial_sums_a[-1]:.8f}")
print(f"  Exact value (6): 6.0  (the series sums to 6)")

# Ratio test: check the ratio (n+1)^2 / (2 n^2) -> 1/2
ratios_a = (ns[1:] + 1)**2 / (2 * (ns[1:])**2)
print(f"  Ratio |a_{{n+1}}/a_n| at n=50: {ratios_a[-1]:.6f}  (limit = 0.5)")

print()

# --- Part (b) verification ---
# n! / n^n — compute carefully to avoid overflow
from math import factorial
terms_b = [factorial(n) / n**n for n in range(1, 21)]
partial_sums_b = np.cumsum(terms_b)
print("Part (b): n! / n^n")
print(f"  Partial sum (n=1..20): {partial_sums_b[-1]:.8f}")
print(f"  Term at n=1: {terms_b[0]:.6f}")
print(f"  Term at n=5: {terms_b[4]:.6f}")
print(f"  Term at n=10: {terms_b[9]:.10f}")
print(f"  Term at n=20: {terms_b[19]:.15f}  (-> 0, confirming limit = 0)")

# Ratio test: (n/(n+1))^n -> 1/e
ratios_b = [(n / (n + 1))**n for n in range(1, 21)]
print(f"  Ratio |a_{{n+1}}/a_n| at n=20: {ratios_b[-1]:.6f}")
print(f"  1/e = {1/np.e:.6f}  (limit of ratio)")
```
