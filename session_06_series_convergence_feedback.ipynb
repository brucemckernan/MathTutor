You concluded the limit is 0 and the test is inconclusive тАФ both correct. But the intermediate step you wrote,

$$\frac{n^2}{2^n} = \frac{\frac{1}{2^n}}{n^2}$$

is not a valid equality (the right side is $\frac{1}{n^2 \cdot 2^n}$, not $\frac{n^2}{2^n}$). The cleanest way to see the limit is 0 is simply that exponentials dominate polynomials:

$$\lim_{n \to \infty} \frac{n^2}{2^n} = 0 \qquad \text{(exponential beats polynomial)}$$

or, if you want to show the work, apply L'H├┤pital twice treating $n$ as continuous.

$$\frac{(n+1)^2}{2n^2} = \frac{1}{2} \cdot \left(\frac{n+1}{n}\right)^2 = \frac{1}{2}\left(1 + \frac{1}{n}\right)^2$$

$$L = \lim_{n \to \infty} \frac{1}{2}\left(1 + \frac{1}{n}\right)^2 = \frac{1}{2} \cdot 1^2 = \frac{1}{2}$$

### Part (b) тАФ $\displaystyle\sum_{n=1}^\infty \frac{n!}{n^n}$

#### Divergence test тАФ тЬЧ Wrong on every step; conclusion reversed

You claimed:

$$\lim_{n \to \infty}\frac{n!}{n^n} = \frac{(n-1)!}{n} = \frac{1 - \frac{1}{n}}{1} = 1$$

None of these steps are valid:

1. $\frac{n!}{n^n} \neq \frac{(n-1)!}{n}$. These are not equal for any $n$. (For example at $n=3$: $\frac{6}{27} \neq \frac{2}{3}$.)
2. $\frac{(n-1)!}{n}$ does not simplify to $\frac{1 - 1/n}{1}$ тАФ factorials do not behave like polynomials.

**The true limit.** Write out what $\frac{n!}{n^n}$ actually looks like:

$$\frac{n!}{n^n} = \frac{1 \cdot 2 \cdot 3 \cdots n}{n \cdot n \cdot n \cdots n} = \frac{1}{n} \cdot \frac{2}{n} \cdot \frac{3}{n} \cdots \frac{n}{n}$$

Every factor is $\leq 1$, and the first factor is $\frac{1}{n} \to 0$. So:

$$0 \leq \frac{n!}{n^n} \leq \frac{1}{n} \to 0$$

The limit is **0**, not 1. The divergence test is **inconclusive** тАФ you cannot stop here.

#### Correct approach: ratio test

$$\frac{a_{n+1}}{a_n} = \frac{(n+1)!}{(n+1)^{n+1}} \cdot \frac{n^n}{n!} = \frac{(n+1) \cdot n^n}{(n+1)^{n+1}} = \frac{n^n}{(n+1)^n} = \left(\frac{n}{n+1}\right)^n = \left(1 - \frac{1}{n+1}\right)^n \to \frac{1}{e}$$

Since $L = \frac{1}{e} < 1$, the series **converges absolutely** тАФ the opposite of your conclusion.

The $\left(1 - \frac{1}{n}\right)^n \to e^{-1}$ limit is worth knowing by heart; it appears frequently with factorials.

**The rule:** when you have $\dfrac{(n+1)^k}{c \cdot n^k}$, the constant $c$ stays **outside** the power:

Applied here:

$$\frac{(n+1)^3}{3n^3} = \frac{1}{3}\left(\frac{n+1}{n}\right)^3 = \frac{1}{3}\left(1 + \frac{1}{n}\right)^3 \to \frac{1}{3} \cdot 1 = \frac{1}{3}$$

$$\frac{a_{n+1}}{a_n} = \frac{(n+2)!}{(n+1)^{n+1}} \cdot \frac{n^n}{(n+1)!} = \frac{(n+2) \cdot n^n}{(n+1)^{n+1}} = \frac{(n+2) \cdot n^n}{(n+1)^n \cdot (n+1)}$$

#### Error 2 тАФ $1^\infty$ is not 1

You wrote: "the $\frac{1}{n} = 0$ meaning the denominator becomes $1^\infty$ which is still 1."

$1^\infty$ is an indeterminate form тАФ you cannot substitute $\frac{1}{n} = 0$ inside the exponent while keeping the exponent $n \to \infty$. The two limits compete. The true result is one of the most important limits in analysis:

$$\lim_{n \to \infty}\left(1 + \frac{1}{n}\right)^n = e \approx 2.718$$

You saw this limit in part (b) in its mirror form $\left(1 - \frac{1}{n+1}\right)^n \to e^{-1}$. It's the same family.

$$\frac{a_{n+1}}{a_n} = \frac{n+2}{n+1} \cdot \left(\frac{n}{n+1}\right)^n = \frac{n+2}{n+1} \cdot \left(1 - \frac{1}{n+1}\right)^n$$

As $n \to \infty$:
- $\dfrac{n+2}{n+1} \to 1$
- $\left(1 - \dfrac{1}{n+1}\right)^n \to e^{-1}$

$$L = 1 \cdot \frac{1}{e} = \frac{1}{e} \approx 0.368 < 1 \implies \textbf{converges absolutely}$$

This is the same $L = 1/e$ result as part (b). A useful pattern: any time the ratio collapses to something of the form $\left(\frac{n}{n+1}\right)^n$ or $\left(1 \pm \frac{c}{n}\right)^n$, the answer involves $e$.

## Q2 тАФ Comparison and limit comparison

(a) Show that $\displaystyle\sum_{n=1}^\infty \frac{1}{n^2 + n}$ converges using the comparison test.

(b) Determine the convergence of $\displaystyle\sum_{n=1}^\infty \frac{2n+1}{n^3 - n + 1}$ using the limit comparison test.

### Feedback

**Both parts correct.**

---

### Part (a) тАФ $\displaystyle\sum_{n=1}^\infty \frac{1}{n^2 + n}$

#### Comparison choice тАФ тЬУ

$\frac{1}{n^2} \geq \frac{1}{n^2+n}$ for all $n \geq 1$ because $n^2 + n > n^2$. The inequality is valid and the direction is right. тЬУ

#### Conclusion тАФ тЬУ

$\sum \frac{1}{n^2}$ is a p-series with $p = 2 > 1$, so it converges. By the Direct Comparison Test, $\sum \frac{1}{n^2+n}$ converges. тЬУ

#### One unnecessary step

You applied the divergence test to $\frac{1}{n^2}$ (the comparison series) before citing it as a p-series. The divergence test only rules out convergence тАФ it can never establish it тАФ and the comparison series isn't what you're testing anyway. The logic you need is simply:

1. $0 \leq \dfrac{1}{n^2+n} \leq \dfrac{1}{n^2}$ for all $n \geq 1$
2. $\displaystyle\sum \frac{1}{n^2}$ converges (p-series, $p=2$)
3. Therefore $\displaystyle\sum \frac{1}{n^2+n}$ converges by DCT

#### Bonus тАФ exact sum via telescoping

This series is one of the rare cases where you can find the exact sum, not just decide convergence.

**Step 1 тАФ partial fractions.** Factor the denominator: $n^2 + n = n(n+1)$, then decompose:

$$\frac{1}{n(n+1)} = \frac{A}{n} + \frac{B}{n+1}$$

Multiplying both sides by $n(n+1)$: $1 = A(n+1) + Bn$. Setting $n = 0$ gives $A = 1$; setting $n = -1$ gives $B = -1$. So:

$$\frac{1}{n^2 + n} = \frac{1}{n} - \frac{1}{n+1}$$

**Step 2 тАФ write out the partial sum.** Expand the sum up to $N$ using this form:

$$S_N = \sum_{n=1}^N \left(\frac{1}{n} - \frac{1}{n+1}\right) = \left(\frac{1}{1} - \frac{1}{2}\right) + \left(\frac{1}{2} - \frac{1}{3}\right) + \left(\frac{1}{3} - \frac{1}{4}\right) + \cdots + \left(\frac{1}{N} - \frac{1}{N+1}\right)$$

**Step 3 тАФ telescoping cancellation.** Every interior term cancels: the $+\frac{1}{2}$ from the first pair is cancelled by the $-\frac{1}{2}$ from the second pair, and so on for every $\frac{1}{k}$ with $2 \leq k \leq N$. Only the very first and very last terms survive:

$$S_N = 1 - \frac{1}{N+1}$$

**Step 4 тАФ take the limit:**

$$\sum_{n=1}^\infty \frac{1}{n^2+n} = \lim_{N \to \infty} S_N = \lim_{N \to \infty}\left(1 - \frac{1}{N+1}\right) = 1$$

The series converges to exactly 1. The comparison test gives convergence but not the value; telescoping gives both.

#### How to recognise a telescoping candidate

The telescoping structure only becomes visible *after* partial fractions тАФ so the real skill is knowing when to try partial fractions in the first place. Two reliable signals:

**Signal 1 тАФ the denominator is a product of two linear factors with roots differing by a fixed integer.**

| Denominator | Factors | Gap | Telescopes |
|---|---|---|---|
| $n(n+1)$ | $n$ and $n+1$ | 1 | $\frac{1}{n} - \frac{1}{n+1}$ |
| $n(n+2)$ | $n$ and $n+2$ | 2 | $\frac{1}{2}\!\left(\frac{1}{n} - \frac{1}{n+2}\right)$ тАФ cancels every other term |
| $(2n-1)(2n+1)$ | consecutive odd integers | 2 | $\frac{1}{2}\!\left(\frac{1}{2n-1} - \frac{1}{2n+1}\right)$ |

Whenever you can factor the denominator this way, partial fractions will split the term into a difference $f(n) - f(n+k)$, and the sum will telescope.

**Signal 2 тАФ the term is already visibly a difference $f(n) - f(n+1)$.**

Sometimes you don't even need partial fractions. For example $\sum (\sqrt{n} - \sqrt{n+1})$ telescopes directly. If you can rewrite $a_n = f(n) - f(n+1)$ for any function $f$, the partial sum collapses to $f(1) - f(N+1)$ immediately.

**Practical checklist** when you encounter a rational series term $\frac{P(n)}{Q(n)}$:
1. Is the degree of $Q$ exactly one more than the degree of $P$? (i.e. $\frac{\text{const}}{n^2+\ldots}$ or $\frac{\text{linear}}{n^3+\ldots}$, etc.)
2. Does $Q$ factor into two (or more) linear factors in $n$?

If both are yes, partial fractions are worth attempting тАФ and the result will often be a telescoping difference.

The integral test requires you to state (before integrating) that $f(x) = \frac{\ln x}{x^2}$ is:

To verify decreasing, differentiate:

$$f'(x) = \frac{\frac{1}{x} \cdot x^2 - \ln x \cdot 2x}{x^4} = \frac{1 - 2\ln x}{x^3}$$

$f'(x) < 0$ when $\ln x > \frac{1}{2}$, i.e. $x > \sqrt{e} \approx 1.65$. So $f$ is actually *increasing* on $(1, \sqrt{e})$ and only decreasing afterwards. The integral test doesn't require monotone decreasing on the whole interval тАФ just *eventually* decreasing. You can apply it from $x = 2$ onwards (or simply note $f$ is decreasing for $x > \sqrt{e}$).

This is a subtle point, but it's worth stating in your solution so the argument is airtight.
