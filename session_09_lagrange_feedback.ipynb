You wrote:

In $\mathbb{R}^2$ the two basis vectors are $\hat{i}$ (x-direction) and $\hat{j}$ (y-direction). $\hat{k}$ is the z-direction; using it here is a slip. This didn't affect the calculation since you immediately component-matched correctly, but it's worth being precise:

$$\nabla f = 2x\,\hat{i} + 2y\,\hat{j}, \qquad \nabla g = \hat{i} + 2\hat{j}$$

### Notation slip 2 — $\nabla f = \lambda\, g$ vs $\nabla f = \lambda\, \nabla g$

You wrote:
$$\nabla f(x,y) = \lambda\, g(x,y)$$

This conflates the scalar function $g$ with its gradient $\nabla g$. On the left is a vector; on the right $\lambda g(x,y)$ is a scalar — the equation is dimensionally inconsistent. The correct Lagrange condition is:

$$\nabla f = \lambda\, \nabla g$$

i.e., the **gradients** are parallel at a constrained extremum, not the function and the gradient. The line immediately after your condition — component form — is correct, so you clearly knew what you were doing; the slip was in writing the condition itself.

### Geometric verification

Excellent. Your argument is clean and correct:

1. The level curve of $f$ through $(1,2)$ is the circle $x^2 + y^2 = 5$.
2. The radius vector from the origin to $(1,2)$ is $\mathbf{r} = \langle 1, 2 \rangle$, which is the outward normal to the circle at that point.
3. The tangent line to the circle at $(1,2)$ is perpendicular to $\mathbf{r}$, giving $1(x-1) + 2(y-2) = 0 \Rightarrow x + 2y = 5$.
4. This is exactly the constraint line — so the circle and the line are tangent at $(1,2)$, confirming it is the extremum.

This is precisely the geometric content of Lagrange multipliers: $\nabla f \parallel \nabla g$ means the level curve of $f$ and the constraint curve share the same normal (equivalently, are tangent) at the solution.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5, 5))

# Constraint line x + 2y = 5
x = np.linspace(-1, 4, 300)
ax.plot(x, (5 - x) / 2, 'b-', label=r'$x + 2y = 5$')

# Level curve of f through (1, 2): x^2 + y^2 = 5
theta = np.linspace(0, 2 * np.pi, 300)
ax.plot(np.sqrt(5) * np.cos(theta), np.sqrt(5) * np.sin(theta),
        'r-', label=r'$x^2+y^2=5$')

# Solution point and radius vector
ax.plot(1, 2, 'ko', markersize=7, zorder=5)
ax.annotate(r'$(1,2)$', (1, 2), textcoords='offset points', xytext=(8, 4))
ax.annotate('', xy=(1, 2), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', color='green', lw=1.5))
ax.plot(0, 0, 'k+', markersize=8)

ax.set_xlim(-2.5, 4)
ax.set_ylim(-2.5, 4)
ax.set_aspect('equal')
ax.axhline(0, color='k', lw=0.5)
ax.axvline(0, color='k', lw=0.5)
ax.legend()
ax.set_title('Level curve tangent to constraint at solution')
plt.tight_layout()
plt.show()
```

## Q2 — Optimisation on a sphere

### Result: Correct answers — one important conceptual error in the method

**Notation improvement:** You correctly wrote $\nabla f = \lambda \nabla g$ this time. Good — this was the slip in Q1.

---

### Setup and $\lambda$ — correct

Gradients are right:

$$\nabla f = \langle 1, 2, 3 \rangle, \qquad \nabla g = \langle 2x, 2y, 2z \rangle$$

The Lagrange conditions give:

$$x = \frac{1}{2\lambda}, \quad y = \frac{1}{\lambda}, \quad z = \frac{3}{2\lambda}$$

Substituting into the constraint:

$$\frac{1}{4\lambda^2} + \frac{1}{\lambda^2} + \frac{9}{4\lambda^2} = \frac{14}{4\lambda^2} = 1 \implies \lambda^2 = \frac{7}{2}$$

All correct.

---

### Conceptual error — x, y, z are not independently $\pm$

You described "eight points of intersection" and evaluated $f$ at all eight sign combinations. **This is not how Lagrange multipliers work**, and it's the main thing to take from this question.

The variables $x$, $y$, $z$ are not free to take signs independently. They are determined by $\lambda$ through:

$$x = \frac{1}{2\lambda}, \quad y = \frac{1}{\lambda}, \quad z = \frac{3}{2\lambda}$$

All three share the same $\lambda$. So a single sign choice for $\lambda$ fixes the signs of all three simultaneously:

- $\lambda = +\sqrt{7/2}$ gives $x > 0,\ y > 0,\ z > 0$
- $\lambda = -\sqrt{7/2}$ gives $x < 0,\ y < 0,\ z < 0$

There are exactly **two critical points**, not eight. A mixed-sign combination like $(+x, -y, +z)$ cannot arise from any single value of $\lambda$ — such a point lies on the sphere but does not satisfy $\nabla f = \lambda \nabla g$, so it is not a candidate extremum from the Lagrange system.

You happened to get the right answer because the true maximum and minimum (all-positive and all-negative) were among the eight you tested, and they did come out largest and smallest. But the six intermediate evaluations were extraneous — they're not critical points.

**The correct argument after finding $\lambda = \pm\sqrt{7/2}$:** evaluate $f$ at the two critical points:

$$f\bigl|_{\lambda = +\sqrt{7/2}} = x + 2y + 3z = \frac{1}{2\lambda} + \frac{2}{\lambda} + \frac{9}{2\lambda} = \frac{14}{2\lambda} = \frac{7}{\lambda} = \frac{7}{\sqrt{7/2}} = \sqrt{14}$$

$$f\bigl|_{\lambda = -\sqrt{7/2}} = -\sqrt{14}$$

So $\max = \sqrt{14}$, $\min = -\sqrt{14}$ (noting $2\sqrt{7/2} = \sqrt{2 \cdot 7} = \sqrt{14}$).

---

### A cleaner way to see the answer

Notice that $f = x + 2y + 3z = \nabla f \cdot \mathbf{r}$ where $\mathbf{r} = \langle x,y,z\rangle$. By the Cauchy–Schwarz inequality:

$$|f| = |\langle 1,2,3\rangle \cdot \mathbf{r}| \leq \|\langle 1,2,3\rangle\| \cdot \|\mathbf{r}\| = \sqrt{14} \cdot 1$$

Equality holds when $\mathbf{r} \parallel \langle 1,2,3\rangle$, i.e. at $\mathbf{r} = \pm\frac{1}{\sqrt{14}}\langle 1,2,3\rangle$. This is the same answer and shows why the extreme values are always $\pm\|\nabla f\|$ for a linear $f$ on the unit sphere.

### Elaboration — The evaluation bar notation $f\big|_{\lambda = a}$

This notation means "the value of the expression on the left, evaluated at the condition on the right." It is a compact substitute for writing out a substitution explicitly.

**Three contexts where you'll see it:**

**1. Definite integrals** (most common)
$$\int_0^1 2x\,dx = \left[x^2\right]_0^1 = 1^2 - 0^2 = 1$$
The bar with subscript/superscript means: evaluate the antiderivative at the upper limit, subtract the value at the lower limit.

**2. Evaluating a function at a point**
$$f\big|_{x=3} \quad \text{means} \quad f(3)$$
Useful when the expression is long and writing $f(3)$ would lose clarity about what is being substituted.

**3. Specifying which case you are in** (what appeared in Q2 feedback)
$$f\big|_{\lambda = +\sqrt{7/2}}$$
means: take the expression for $f$ and substitute $\lambda = +\sqrt{7/2}$ throughout. It is shorthand for "in the case $\lambda = +\sqrt{7/2}$."

**When to use it vs alternatives:**

| Situation | Bar notation | Alternative |
|---|---|---|
| Definite integral | $\left[F(x)\right]_a^b$ | Almost always use this — it is standard |
| Simple substitution | $f\big|_{x=3}$ | Just write $f(3)$ if the variable is obvious |
| Evaluating at a named case | $f\big|_{\lambda=+\sqrt{7/2}}$ | "When $\lambda = +\sqrt{7/2}$, $f = \ldots$" in words |

In Q2 it was used stylistically — the prose alternative is equally valid. The bar is most essential in the definite integral context; elsewhere it is a matter of taste.

### Elaboration — The Cauchy–Schwarz inequality

**Statement:** For any two vectors $\mathbf{u}$ and $\mathbf{v}$ in $\mathbb{R}^n$:

$$|\mathbf{u} \cdot \mathbf{v}| \leq \|\mathbf{u}\|\,\|\mathbf{v}\|$$

with equality if and only if $\mathbf{u}$ and $\mathbf{v}$ are parallel.

---

**Geometric derivation** (most transparent)

Recall the geometric definition of the dot product: $\mathbf{u} \cdot \mathbf{v} = \|\mathbf{u}\|\,\|\mathbf{v}\|\cos\theta$, where $\theta$ is the angle between them. Therefore:

$$|\mathbf{u} \cdot \mathbf{v}| = \|\mathbf{u}\|\,\|\mathbf{v}\|\,|\cos\theta| \leq \|\mathbf{u}\|\,\|\mathbf{v}\|$$

since $|\cos\theta| \leq 1$ always, with equality when $\theta = 0$ or $\pi$ (i.e. parallel vectors). Done.

---

**Algebraic derivation** (works without assuming the geometric dot product formula — useful in more abstract settings)

Consider $h(t) = \|\mathbf{u} - t\mathbf{v}\|^2 \geq 0$ for all real $t$. Expanding:

$$h(t) = \|\mathbf{u}\|^2 - 2t(\mathbf{u}\cdot\mathbf{v}) + t^2\|\mathbf{v}\|^2 \geq 0$$

This is a quadratic in $t$ that is never negative, so its discriminant must satisfy $\Delta \leq 0$:

$$4(\mathbf{u}\cdot\mathbf{v})^2 - 4\|\mathbf{u}\|^2\|\mathbf{v}\|^2 \leq 0 \implies (\mathbf{u}\cdot\mathbf{v})^2 \leq \|\mathbf{u}\|^2\|\mathbf{v}\|^2$$

Taking square roots gives Cauchy–Schwarz. Equality holds when $h(t) = 0$ for some $t$, meaning $\mathbf{u} = t\mathbf{v}$ — i.e. they are parallel.

---

**How it applied in Q2**

Your objective $f = x + 2y + 3z$ is a dot product:

$$f = \underbrace{\langle 1, 2, 3\rangle}_{\mathbf{u}} \cdot \underbrace{\langle x, y, z\rangle}_{\mathbf{v}}$$

The constraint $x^2 + y^2 + z^2 = 1$ says $\|\mathbf{v}\| = 1$. Cauchy–Schwarz immediately gives:

$$|f| \leq \|\langle 1,2,3\rangle\| \cdot 1 = \sqrt{1^2 + 2^2 + 3^2} = \sqrt{14}$$

This tells us the **maximum value** of $f$ is $\sqrt{14}$. But where is that maximum achieved? That is a separate question answered by the equality condition.

**Finding the location of the maximum — where $\frac{1}{\sqrt{14}}$ comes from**

Cauchy–Schwarz achieves equality when $\mathbf{v}$ is parallel to $\mathbf{u}$, meaning:

$$\langle x, y, z\rangle = c\,\langle 1, 2, 3\rangle \quad \text{for some scalar } c$$

We have one free parameter $c$, but we also have a constraint to satisfy: $\langle x,y,z\rangle$ must lie on the unit sphere, i.e. $\|\langle x,y,z\rangle\| = 1$. Substituting:

$$\|c\,\langle 1,2,3\rangle\| = 1 \implies |c|\,\|\langle 1,2,3\rangle\| = 1 \implies |c|\,\sqrt{14} = 1 \implies |c| = \frac{1}{\sqrt{14}}$$

So $c = +\dfrac{1}{\sqrt{14}}$ or $c = -\dfrac{1}{\sqrt{14}}$, giving the two critical points:

$$\langle x,y,z\rangle = \pm\frac{1}{\sqrt{14}}\langle 1,2,3\rangle$$

In words: the maximum is achieved at the point on the unit sphere that lies directly along $\langle 1,2,3\rangle$ (or directly opposite for the minimum). The factor $\frac{1}{\sqrt{14}}$ is simply what you need to scale $\langle 1,2,3\rangle$ so that it has length 1 — it is the normalization factor, or equivalently $\frac{1}{\|\langle 1,2,3\rangle\|}$.

So the two quantities $\sqrt{14}$ and $\frac{1}{\sqrt{14}}$ refer to different things:
- $\sqrt{14} = \|\langle 1,2,3\rangle\|$ is the **maximum value of $f$**
- $\frac{1}{\sqrt{14}}$ is the **scaling factor applied to $\langle 1,2,3\rangle$** to make it a unit vector (the location of the maximum)

**General pattern:** Whenever you maximise a linear function $\mathbf{c} \cdot \mathbf{x}$ subject to $\|\mathbf{x}\| = 1$, the maximum value is $\|\mathbf{c}\|$ and it is achieved at $\mathbf{x} = \frac{\mathbf{c}}{\|\mathbf{c}\|}$ (the unit vector in the direction of $\mathbf{c}$).

---

**Where you will encounter Cauchy–Schwarz**

| Context | What it says |
|---|---|
| Optimising linear $f$ on a sphere | $\max f = \|\nabla f\|$ (Q2 above) |
| Triangle inequality | $\|\mathbf{u}+\mathbf{v}\| \leq \|\mathbf{u}\|+\|\mathbf{v}\|$ follows from C-S as a lemma |
| Statistics | Correlation coefficient $r \in [-1,1]$ because $r = \frac{\mathbf{u}\cdot\mathbf{v}}{\|\mathbf{u}\|\|\mathbf{v}\|}$ and C-S bounds this by 1 |
| Least squares / projections | Bounding approximation error (Session 03 territory) |

The inequality generalises beyond $\mathbb{R}^n$ to any inner product space, including spaces of functions (where the dot product becomes an integral $\int f g$). That is why it appears throughout analysis, probability, and quantum mechanics — but the vector version above is the form to keep in hand for now.
