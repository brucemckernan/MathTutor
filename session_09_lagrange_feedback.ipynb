**Setup and algebra** 
You correctly identified $f(x,y) = x^2 + y^2$ (distance-squared from origin) and constraint $g(x,y) = x + 2y = 5$. The system of equations from $\nabla f = \lambda \nabla g$ was set up and solved correctly:

Eliminating $x$ and $\lambda$ gives $\lambda = 2$, $x = 1$, $y = 2$. The closest point is $(1, 2)$, with $f(1,2) = 5$. All correct.

### Notation slip 1 — $\hat{k}$ vs $\hat{j}$

You wrote:
$$\nabla f = 2x\,\hat{i} + 2y\,\hat{k}, \qquad \nabla g = 1\,\hat{i} + 2\,\hat{k}$$

In $\mathbb{R}^2$ the two basis vectors are $\hat{i}$ (x-direction) and $\hat{j}$ (y-direction). $\hat{k}$ is the z-direction; using it here is a slip. This didn't affect the calculation since you immediately component-matched correctly, but it's worth being precise:

$$\nabla f = 2x\,\hat{i} + 2y\,\hat{j}, \qquad \nabla g = \hat{i} + 2\hat{j}$$

---

### Notation slip 2 — $\nabla f = \lambda\, g$ vs $\nabla f = \lambda\, \nabla g$

You wrote:
$$\nabla f(x,y) = \lambda\, g(x,y)$$

This conflates the scalar function $g$ with its gradient $\nabla g$. On the left is a vector; on the right $\lambda g(x,y)$ is a scalar — the equation is dimensionally inconsistent. The correct Lagrange condition is:

$$\nabla f = \lambda\, \nabla g$$

i.e., the **gradients** are parallel at a constrained extremum, not the function and the gradient. The line immediately after your condition — component form — is correct, so you clearly knew what you were doing; the slip was in writing the condition itself.

### Geometric verification

Excellent. Your argument is clean and correct:

1. The level curve of $f$ through $(1,2)$ is the circle $x^2 + y^2 = 5$.
2. The radius vector from the origin to $(1,2)$ is $\mathbf{r} = \langle 1, 2 \rangle$, which is the outward normal to the circle at that point.
3. The tangent line to the circle at $(1,2)$ is perpendicular to $\mathbf{r}$, giving $1(x-1) + 2(y-2) = 0 \Rightarrow x + 2y = 5$.
4. This is exactly the constraint line — so the circle and the line are tangent at $(1,2)$, confirming it is the extremum.

This is precisely the geometric content of Lagrange multipliers: $\nabla f \parallel \nabla g$ means the level curve of $f$ and the constraint curve share the same normal (equivalently, are tangent) at the solution.

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(5, 5))

# Constraint line x + 2y = 5
x = np.linspace(-1, 4, 300)
ax.plot(x, (5 - x) / 2, 'b-', label=r'$x + 2y = 5$')

# Level curve of f through (1, 2): x^2 + y^2 = 5
theta = np.linspace(0, 2 * np.pi, 300)
ax.plot(np.sqrt(5) * np.cos(theta), np.sqrt(5) * np.sin(theta),
        'r-', label=r'$x^2+y^2=5$')

# Solution point and radius vector
ax.plot(1, 2, 'ko', markersize=7, zorder=5)
ax.annotate(r'$(1,2)$', (1, 2), textcoords='offset points', xytext=(8, 4))
ax.annotate('', xy=(1, 2), xytext=(0, 0),
            arrowprops=dict(arrowstyle='->', color='green', lw=1.5))
ax.plot(0, 0, 'k+', markersize=8)

ax.set_xlim(-2.5, 4)
ax.set_ylim(-2.5, 4)
ax.set_aspect('equal')
ax.axhline(0, color='k', lw=0.5)
ax.axvline(0, color='k', lw=0.5)
ax.legend()
ax.set_title('Level curve tangent to constraint at solution')
plt.tight_layout()
plt.show()
```
