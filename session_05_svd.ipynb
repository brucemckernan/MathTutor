**Your answer:**

I use the determinant of this square vector to find its eigenvalues (from the equation $Av_i = \sigma v_i$, eventually from this $(A - \lambda I) v = 0$ and this only being true if the determinant of $A - \lambda I$ equals zero)

$$det\left[\begin{pmatrix}11 - \lambda & 7 \\ 7 & 11-\lambda

## Q2 — SVD and the four fundamental subspaces

For a matrix $A$ with SVD $A = U \Sigma V^T$, express each of the four fundamental subspaces (column space, null space, row space, left null space) in terms of the columns of $U$ and $V$. How does the rank of $A$ appear in the SVD?

**Your answer:**

The definition of rank is the number of non zero singular values, so in the case of the SVD I have just completed  the matrix is rank 2. This makes sense as I have a 3 x 2 matrix which represents a 2D plane embedded in a 3D space and rank is the minimum of (m, n) in this case n=2. So I have two orthogonal singularvectors in the columns of U spanning the column space of A (although there are no vectors in the left null space of this matrix, if there had been then they would have been in columns r < = n of U where r is the rank of the matrix. Simillarly the rows of V and therefore columns of $v^T$ hold singular vectors that span the row space of A. There are no vectors in the null space of A but if there had been they would have been in columns r <= n where r is the rank of the matrix.

## Q3 — Python: visualising the transformation

For the matrix $B = \begin{pmatrix}2 & 1\\0 & 1\end{pmatrix}$:

(a) Write code to compute the SVD using `numpy.linalg.svd`.

(b) Plot the unit circle and its image under $B$ on the same axes.

(c) Draw the semi-axes of the resulting ellipse, scaled by the singular values and oriented by the left singular vectors ($U$).

(d) Add annotations showing $\sigma_1$ and $\sigma_2$.

This is the picture from Session 01 B2 — now produce it yourself.

```python
import matplotlib.pyplot as plt
import numpy as np
import numpy.linalg as la

B = np.array([[2, 1], [0, 1]], dtype=float)
U, s, Vh = la.svd(B, full_matrices=True)

t = np.linspace(0, 2 * np.pi, 400)
circle = np.column_stack([np.cos(t), np.sin(t)])
ellipse = circle @ B.T

fig, ax = plt.subplots(figsize=(6, 6))
ax.set_xlim(-4, 4)
ax.set_ylim(-4, 4)
ax.set_aspect('equal')
ax.grid(True)
ax.plot(circle[:, 0], circle[:, 1], label='Unit circle')
ax.plot(ellipse[:, 0], ellipse[:, 1], label='Image under B')

for i in range(2):
    tip = s[i] * U[:, i]
    ax.annotate('', xy=tip, xytext=(0, 0),
                arrowprops=dict(arrowstyle='->', color='red', lw=2))
    ax.text(tip[0] + 0.1, tip[1] + 0.1, f'σ{i+1} = {s[i]:.2f}', color='red')

ax.legend()
plt.title('SVD: unit circle → ellipse under B')
plt.show()
```
