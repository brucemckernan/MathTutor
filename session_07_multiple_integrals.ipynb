## Q1 — Changing order of integration

Evaluate $\displaystyle\int_0^1 \int_y^1 e^{x^2} \, dx \, dy$ by switching the order of integration. (The integral is not directly evaluable in the given order.)

**Your answer:**
First I drew out the region being integrated over on the xy plane. This is a trianlge bounded by x=y (or y=x) and the points (0,0), (1,0), (1,1). From this I see that I can change the uppper bound of y to be a function of x (y=x) andd its lower bound is still a conostant 0. In addition I can rewrite the bounds of x as constants x = 0 and x = 1. This lets me reorder the integration so that I can integrate with respect ot x on the "outside". This means I should be able to use a u-sub after introducing an 'x' into the expression with the inside integration with respect to y:

$$\int_{x=0}^{x=1} \int_{y=0}^{y=x} e^{x^2}\,dydx$$
$$\int_{x=0}^{x=1} \left[ e^{x^2} y\right]_{y=0}^{y=x}\,dydx$$

integrate with respect to y and evaluate:

$$\int_{x=0}^{x=1} \left[ e^{x^2}\times x - e^{x^2} \times 0\right]\,dx$$
$$\int_{x=0}^{x=1} \left[ e^{x^2} x\right]\,dx$$


I use u-sub of $u = x^2$ and change bounds. Integrate with respect to u and evaluate:

$$\int_{u=0}^{u=1} \left[ e^{u} x\right]\,\frac{du}{2x}$$
$$\frac{1}{2}\int_{u=0}^{u=1} e^{u}\,du$$
$$\frac{1}{2}\left[ e^{u} \right]_{u=0}^{u=1}$$

## Q2 — Double integral over a general region

Evaluate $\displaystyle\iint_R xy \, dA$ where $R$ is the region bounded by $y = x$ and $y = x^2$.

Sketch the region, identify the limits, and evaluate.

**Your answer:**

So I drew out the region and set the equatoins equal to each other to find their intercepts. They intercept at the points (0,0) and (1,1) creating a bounding rectangle, with the y=x^2 parabola below y=x until point (1,1). As both directions lie between the same two functions I can integrate in either direction first. However handling $y=x^2$ seems somewhat easier than handling $x=\sqrt{y}$ I chose to integrate in the y direction first with $y=x^2$ as the lower bound and $y=x$ as the upper bound and the outer integral with respect to x with the lower bound the constant 0 and upper bound the constant 1:

setting up the integral and region:
$$\int_{x=0}^{x=1} \int_{y={x^2}}^{y=x} xy \, dydx$$

integrating with respect to y:

$$\int_{x=0}^{x=1} \left[ \frac{1}{2} xy^2 \right]_{y={x^2}}^{y=x} \, dx$$
$$\frac{1}{2} \int_{x=0}^{x=1} \left[ xy^2 \right]_{y={x^2}}^{y=x} \, dx$$

evaluating the inner integral:

$$\frac{1}{2} \int_{x=0}^{x=1} \left[ x^3 - x^5 \right]\, dx$$

integrating with respect to x:

$$\frac{1}{2} \left[ \frac{x^4}{4} - \frac{x^6}{6} \right]_{x=0}^{x=1}$$

evaluating the outer integral and simplifying:

$$\frac{1}{2} \left[ \left(\frac{1^4}{4} - \frac{1^6}{6} \right)- \left( \frac{0^4}{4} - \frac{0^6}{6}\right)\right] \rightarrow \frac{1}{2} \left(\frac{1}{4} - \frac{1}{6} \right) = \frac{1}{2} \times \frac{2}{24} = \frac{1}{24}$$



## Q3 — Polar coordinates

Evaluate $\displaystyle\iint_R (x^2 + y^2) \, dA$ where $R$ is the disk $x^2 + y^2 \leq 4$.

Convert to polar coordinates and evaluate.

**Your answer:**

First I define the region in terms of polar coordinates using the identity $r^2 = x^2 + y^2$ and from this I can find out that the radius of the region goes from the origin to the circular edge at r=2. As the disc is a circular region on the xy-plane centered at the origin I can sweep out the whole region with an angle $\theta$ going from 0 (the x-asxis) to $2\pi$. This gives me the lower and upper bounds of the inner integral of a polaer coordinate integral which is always r as well as the outer integral bounds for $\theta$. I then use the same identity $r^2 = x^2 + y^2$ to convert the integrand to polar coordinates (this is a parabolic surface above the xy-plane):

$$r^2 = x^2 + y^2$$
$$r^2 \leq 4 \rightarrow r = \sqrt{4} = 2$$
$$\int_{\theta=0}^{\theta=2\pi}\int_{r=0}^{r=2}(r^2) \, rdrd\theta$$
$$\int_{\theta=0}^{\theta=2\pi}\int_{r=0}^{r=2}(r^3) \, drd\theta$$
$$\int_{\theta=0}^{\theta=2\pi}\left[ \frac{r^4}{4}\right]_{r=0}^{r=2} \, d\theta$$
$$\int_{\theta=0}^{\theta=2\pi}\frac{1}{4}\left[ r^4\right]_{r=0}^{r=2} \, d\theta$$
$$\int_{\theta=0}^{\theta=2\pi}\frac{1}{4}\left[ 2^4 - 0^4\right] \, d\theta$$
$$\int_{\theta=0}^{\theta=2\pi}4 \, d\theta$$
$$\left[ 4 \theta\right]_{\theta=0}^{\theta=2\pi}$$
$$\left[ 8\pi - 0\pi \right] = 8\pi$$

therefore I have found that the volume under the parabolic surface above the xy-plane bounded by the circular region is $8\pi$
