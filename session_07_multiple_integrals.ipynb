## Q1 — Changing order of integration

Evaluate $\displaystyle\int_0^1 \int_y^1 e^{x^2} \, dx \, dy$ by switching the order of integration. (The integral is not directly evaluable in the given order.)

First I drew out the region being integrated over on the xy plane. This is a trianlge bounded by x=y (or y=x) and the points (0,0), (1,0), (1,1). From this I see that I can change the uppper bound of y to be a function of x (y=x) andd its lower bound is still a conostant 0. In addition I can rewrite the bounds of x as constants x = 0 and x = 1. This lets me reorder the integration so that I can integrate with respect ot x on the "outside". This means I should be able to use a u-sub after introducing an 'x' into the expression with the inside integration with respect to y:

$$\int_{x=0}^{x=1} \int_{y=0}^{y=x} e^{x^2}\,dydx$$
$$\int_{x=0}^{x=1} \left[ e^{x^2} y\right]_{y=0}^{y=x}\,dydx$$

integrate with respect to y and evaluate:

$$\int_{x=0}^{x=1} \left[ e^{x^2}\times x - e^{x^2} \times 0\right]\,dx$$
$$\int_{x=0}^{x=1} \left[ e^{x^2} x\right]\,dx$$


I use u-sub of $u = x^2$ and change bounds. Integrate with respect to u and evaluate:

$$\int_{u=0}^{u=1} \left[ e^{u} x\right]\,\frac{du}{2x}$$
$$\frac{1}{2}\int_{u=0}^{u=1} e^{u}\,du$$
$$\frac{1}{2}\left[ e^{u} \right]_{u=0}^{u=1}$$
$$\frac{1}{2}\left( e^{1} - e^{0} \right) = \frac{e - 1}{2} = 0.85914$$
