## Q1 — Ratio test

Determine whether the following series converge or diverge. State which test you are using and show the limit clearly.

(a) $\displaystyle\sum_{n=1}^\infty \frac{n^2}{2^n}$

(b) $\displaystyle\sum_{n=1}^\infty \frac{n!}{n^n}$

(c) $\displaystyle\sum_{n=1}^\infty \frac{n^3}{3^n}$

(d) $\displaystyle\sum_{n=1}^\infty \frac{5^n}{n!}$

(e) $\displaystyle\sum_{n=1}^\infty \frac{n!}{5^n}$

(f) $\displaystyle\sum_{n=1}^\infty \frac{(n+1)!}{n^n}$

**Your answer:**

For question 1(a) I first do a divergence test to see if we can rule out convergence:

$$\lim_{n \to \infty}\frac{n^2}{2^n} = \frac{\frac{1}{2^n}}{n^2} = \frac{0}{\infty} = 0$$

As the limit of the cdivergence test is 0 it is inconclusive. As we have an expression with exponentials it is reasonable to try the ratio test:

$$\lim_{n \to \infty}\left| \frac{\frac{(n+1)^2}{2^{n+1}}}{\frac{n^2}{2^n}}\right|$$
$$\lim_{n \to \infty}\left| \frac{(n+1)^2}{2^{n+1}}\times{\frac{2^n}{n^2}}\right|$$
$$\lim_{n \to \infty}\left| \frac{(n+1)^2}{2}\times{\frac{1}{n^2}}\right|$$
$$\lim_{n \to \infty}\left| \frac{(n+1)^2}{2n^2}\right|$$
$$\lim_{n \to \infty}\left| \left(\frac{(n+1)}{2n}\right)^2\right|$$

As we have an expression squared we can drop the absolute value as it will always be positive:

$$\lim_{n \to \infty}\left(\frac{(n+1)}{2n}\right)^2$$
$$\lim_{n \to \infty}\left(\frac{(1+\frac{1}{n})}{2}\right)^2$$

with n isolated we can let n go to infinity and evaluate the limit: 

$$\lim_{n \to \infty}\left(\frac{(1+0)}{2}\right)^2 = \frac{1}{2}^2 = \frac{1}{4}$$

The limit exists and it is less than 1 therefore by the ratio test we can conclude the series absolutely converges (as the ratio of the nth term to the nth + 1 term shows us that the difference between terms is decreasing by an order of magnitude 2 for each step)

for question 1(b) I first do a divergence test to see if we can rule out convergence:

$$\lim_{n=1 \to \infty}\frac{n!}{n^n} = \frac{(n-1)!}{n} = \frac{1 - \frac{1}{n}}{1} = 1$$

As the divergence test returns a result > 0 it means I can conclude the series diverges without going any further.

## Q2 — Comparison and limit comparison

(a) Show that $\displaystyle\sum_{n=1}^\infty \frac{1}{n^2 + n}$ converges using the comparison test.

(b) Determine the convergence of $\displaystyle\sum_{n=1}^\infty \frac{2n+1}{n^3 - n + 1}$ using the limit comparison test. State clearly what you are comparing to.

**Your answer:**

For part (a) I choose to compare with the series $\frac{1}{n^2}$ as:

$$\frac{1}{n^2 + n} \leq \frac{1}{n^2}$$ for all n

using the divergence test first in inconclusive:

$$\lim_{n \to \infty} \frac{1}{n^2} = 0$$

However $\frac{1}{n^2}$ is a p-series with p > 1 and therefore it converges. Therefore as the orignal series is always less than this series I can conclude that it also converges.

For part (b) I pull out the dominant terms of the numerator and denominator to create $b_n$:

$$b_n = \frac{2n}{n^3} = \frac{2}{n^2} = 2 \times \frac{1}{n^2}$$

$b_n$ is therefore a p-series with P > 1 and therefore it converges. I just need to show that the original series shares the same fate. Therefore I use the limit comparisson test:

$$\lim_{n \to \infty}\frac{\frac{2n+1}{n^3 - n + 1}}{\frac{2}{n^2}}$$
$$\frac{2n+1}{n^3 - n + 1} \times \frac{n^2}{2}$$
$$\frac{2n^3 + n^2}{2n^3 - 2n + 2}$$
$$\frac{2 + \frac{1}{n}}{2 - \frac{2}{n^2} + \frac{2}{n^3}}$$
$$\frac{2 + 0}{2 - 0 + 0}=\frac{2}{2} = 1$$

the result is $ 0 \lt 1 \lt \infty$ therefore both series share the same fate and converge.


## Q3 — Integral test

Use the integral test to determine whether $\displaystyle\sum_{n=1}^\infty \frac{\ln n}{n^2}$ converges or diverges. (You will need IBP from Session 02.)

**Your answer:**

I set up the integral by replacing n with a continuous variable x

$$\int_{1}^{\infty} \frac{ln x}{x^2} \, dx$$

As this is integrating a product of functions of x I use IBP setting up values for u and dv by using LIATE:

$$u = ln x, du = \frac{1}{x}, du = \frac{1}{x^2}, v = -\frac{1}{x}$$

And setting up the IBP formula

$$ln x \times -\frac{1}{x} - \int_{1}^{\infty} \frac{1}{x} \times \frac{1}{x} \, dx$$
$$-\frac{ln x}{x} - \int_{1}^{\infty} - \frac{1}{x^2} \, dx$$
$$- \frac{ln x}{x} - \left[\frac{1}{x}\right]_{1}^{\infty}$$

To evaluate an improper integral I use a limit replacing the infinity upperbound with a variable b that can approach infinity:

$$- \frac{ln x}{x} - \lim_{b \to \infty}\left[\frac{1}{x}\right]_{1}^{b}$$
$$- \frac{ln x}{x} - \lim_{b \to \infty}\left[\frac{1}{b} - 1\right]$$
$$1 - \frac{ln x}{x}$$

I then use a limit to handle the remining $\frac{ln x}{x}$ in the expression:

$$\lim_{x \to \infty} 1 - \frac{ln x}{x} \rightarrow  1 - \lim_{x \to \infty} \frac{ln x}{x}$$

As the integral converges on a value I can conlude that the series also converges:

$$1 - \lim_{x \to \infty} \frac{\frac{1}{x}}{1} = 1 - \frac{0}{1} = 1 - 0 = 1$$
