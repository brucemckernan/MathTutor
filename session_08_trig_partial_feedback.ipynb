### Two points worth internalising

**1. Why $\sin$, not $\cos$?**

Your triangle intuition (mapping $x$ to the opposite side) is a valid mnemonic, but the algebraic reason is cleaner: you need $a^2 - x^2$ to factor as a perfect square so the square root disappears. Substituting $x = a\sin\theta$ gives

$$a^2 - x^2 = a^2 - a^2\sin^2\theta = a^2(1 - \sin^2\theta) = a^2\cos^2\theta$$

and $\sqrt{a^2\cos^2\theta} = a\cos\theta$ (on the principal branch). You could equally set $x = a\cos\theta$ — the algebra works — but you'd get $-\sin\theta$ terms and the simplification is slightly messier. $\sin$ is the conventional choice.

The three standard patterns and their substitutions:

| Form in integrand | Substitution | Identity used |
|---|---|---|
| $\sqrt{a^2 - x^2}$ | $x = a\sin\theta$ | $1 - \sin^2\theta = \cos^2\theta$ |
| $\sqrt{a^2 + x^2}$ | $x = a\tan\theta$ | $1 + \tan^2\theta = \sec^2\theta$ |
| $\sqrt{x^2 - a^2}$ | $x = a\sec\theta$ | $\sec^2\theta - 1 = \tan^2\theta$ |

Worth committing to memory — Session 8 Q2 uses the second pattern.

**2. The sign assumption on $\cos\theta$ — in detail**

At the step $\sqrt{9\cos^2\theta} = 3\cos\theta$ you implicitly used the fact that

$$\sqrt{u^2} = |u|, \quad\text{not always }u.$$

So the algebraically correct statement is $\sqrt{9\cos^2\theta} = 3|\cos\theta|$. You can drop the absolute value only when you know $\cos\theta \geq 0$.

**Why the principal branch makes it valid.**

When we write $x = 3\sin\theta$ and then back-substitute $\theta = \arcsin(x/3)$, we commit to the principal branch of arcsin:

$$\theta \in \left[-\frac{\pi}{2},\, \frac{\pi}{2}\right].$$

Cosine is non-negative on that entire interval, so $\cos\theta \geq 0$ throughout, and $|\cos\theta| = \cos\theta$ with no caveat.

**Stepwise example on the principal branch ($\theta = \pi/6$).**

Take $x = 3/2$. Then $\sin\theta = 1/2 \Rightarrow \theta = \pi/6$ (principal branch).

| Quantity | Value |
|---|---|
| $\cos(\pi/6)$ | $\sqrt{3}/2 > 0$ ✓ |
| $\sqrt{9 - x^2}$ directly | $\sqrt{9 - 9/4} = \sqrt{27/4} = \tfrac{3\sqrt{3}}{2}$ |
| $3\cos(\pi/6)$ | $3 \cdot \tfrac{\sqrt{3}}{2} = \tfrac{3\sqrt{3}}{2}$ ✓ |
| Integrand after substitution | $\dfrac{3\cos\theta}{3\cos\theta}\,d\theta = 1\,d\theta$ ✓ |

Everything agrees; no absolute value needed.

**Counter-example on a different branch ($\theta = 5\pi/6$).**

$\sin(5\pi/6) = 1/2$ as well — same $x = 3/2$ — but now $\theta = 5\pi/6$ (second quadrant, outside the principal branch).

| Quantity | Value |
|---|---|
| $\cos(5\pi/6)$ | $-\sqrt{3}/2 < 0$ ✗ |
| $\sqrt{9\cos^2(5\pi/6)}$ | $3\left|-\tfrac{\sqrt{3}}{2}\right| = \tfrac{3\sqrt{3}}{2}$ (always positive) |
| $3\cos(5\pi/6)$ (no absolute value) | $-\tfrac{3\sqrt{3}}{2}$ ← **negative**, contradicts a square root |

If you wrote $\sqrt{9\cos^2\theta} = 3\cos\theta$ here without checking, you would assign a **negative value to a square root** — which is impossible. The correct denominator is $3|\cos\theta| = +\tfrac{3\sqrt{3}}{2}$.

Now the simplified integrand is

$$\frac{3\cos\theta}{3|\cos\theta|}\,d\theta = \frac{-\tfrac{3\sqrt{3}}{2}}{+\tfrac{3\sqrt{3}}{2}}\,d\theta = -1\,d\theta$$

so integration gives $-\theta + C$, not $+\theta + C$ — a **sign flip in the answer**. (The errors from the wrong sign and the wrong back-substitution happen to cancel for indefinite integrals, but the intermediate steps are wrong and the approach breaks for definite integrals with bounds in the wrong region.)

**The takeaway.** The principal-branch convention $\theta \in [-\pi/2, \pi/2]$ is not just a notational habit — it is what licenses you to write $\sqrt{\cos^2\theta} = \cos\theta$. State it explicitly once at the top of any trig-substitution working and the rest follows cleanly.

```python
from sympy import *

x = symbols('x')

integrand = 1 / sqrt(9 - x**2)
result = integrate(integrand, x)
print("Antiderivative:", result)

# Confirm it matches arcsin(x/3)
expected = asin(x / 3)
print("Matches arcsin(x/3)?:", simplify(result - expected) == 0)
```

### Issue 2 — The absolute value and the principal branch

You wrote: *"I cancel $|\sec\theta|$ with one power of $\sec^2\theta$ — as a result the expression can take on any value including negative so I drop the absolute value."*

This reasoning is back to front. Dropping the absolute value makes the expression *less* restricted, not more. The correct reason to drop it is:

When $x = 2\tan\theta$ and we back-substitute $\theta = \arctan(x/2)$, we commit to the **principal branch of arctan**, $\theta\in(-\pi/2,\,\pi/2)$. On that interval $\cos\theta > 0$, so $\sec\theta = 1/\cos\theta > 0$, which means $|\sec\theta| = \sec\theta$ from the very start. State the principal branch at the top of your working and write $\sec\theta$ throughout — no separate sign check is ever needed.

Concretely: $\dfrac{\sec^2\theta}{|\sec\theta|} = \dfrac{\sec^2\theta}{\sec\theta} = \sec\theta$ (positive) because $\sec\theta > 0$ on the principal branch.

**Does restricting to the principal branch limit where the antiderivative is valid?**

No — and this is worth understanding clearly. The restriction is on $\theta$ (the intermediate variable), not on $x$ (the variable in the final answer). Once you back-substitute and write the result in terms of $x$, the antiderivative is valid across the full natural domain of the original integrand.

The reason is that each principal branch is chosen so that it covers every $x$ value the integrand can take, with no gaps:

| Substitution | Principal branch of $\theta$ | $x$ values covered | Natural domain of integrand |
|---|---|---|---|
| $x = a\sin\theta$ | $\theta \in [-\pi/2,\, \pi/2]$ | $x \in [-a,\, a]$ | $\sqrt{a^2-x^2}$ defined on $(-a,\,a)$ ✓ |
| $x = a\tan\theta$ | $\theta \in (-\pi/2,\, \pi/2)$ | $x \in (-\infty,\, \infty)$ | $\sqrt{a^2+x^2}$ defined on all $\mathbb{R}$ ✓ |
| $x = a\sec\theta$ | $\theta \in [0,\pi/2)\cup(\pi/2,\pi]$ | $\lvert x\rvert \geq a$ | $\sqrt{x^2-a^2}$ defined on $\lvert x\rvert \geq a$ ✓ |

The angles outside the principal branch — say $\theta \in (\pi/2, 3\pi/2)$ — do not add any new $x$ values. They only repeat values already covered. For example with $x = 2\tan\theta$:

- $\theta = \pi$ (excluded): $\tan\pi = 0$, so $x = 0$ — already given by $\theta = 0$.
- $\theta = 3\pi/4$ (excluded): $\tan(3\pi/4) = -1$, so $x = -2$ — already given by $\theta = -\pi/4$.

Trig functions are periodic and not globally one-to-one. The principal branch picks exactly **one representative $\theta$ for each $x$ value** and discards the redundant repeats — which is also what makes the back-substitution well-defined. For Q2, the final antiderivative

$$\frac{x\sqrt{x^2+4}}{2} - 2\ln\!\left|x+\sqrt{x^2+4}\right| + C$$

is valid for all real $x$. The $\theta$-level bookkeeping was scaffolding for the intermediate computation; once you are back in $x$ it has no further role.

### Issue 3 — Missing factor of 4: step-by-step

The key is to keep track of **what you are actually trying to find**. After the substitution the original integral becomes:

$$\underbrace{\int\frac{x^2}{\sqrt{x^2+4}}\,dx}_{\text{what we want}} = \underbrace{4\int\tan^2\theta\sec\theta\,d\theta}_{\text{call this }I}$$

So $I = 4\int\tan^2\theta\sec\theta\,d\theta$ is what we need to find — it carries the factor of 4.

**Step 1.** You correctly applied $\tan^2\theta = \sec^2\theta - 1$ and split:

$$I = 4\int(\sec^3\theta - \sec\theta)\,d\theta = 4\int\sec^3\theta\,d\theta - 4\ln|\sec\theta+\tan\theta| \quad \cdots [1]$$

**Step 2.** IBP on $\int\sec^3\theta$:

$$\int\sec^3\theta\,d\theta = \sec\theta\tan\theta - \int\tan^2\theta\sec\theta\,d\theta$$

**Step 3.** Substitute step 2 into [1]:

$$I = 4\!\left[\sec\theta\tan\theta - \int\tan^2\theta\sec\theta\,d\theta\right] - 4\ln|\sec\theta+\tan\theta|$$

$$I = 4\sec\theta\tan\theta - 4\int\tan^2\theta\sec\theta\,d\theta - 4\ln|\sec\theta+\tan\theta|$$

**Step 4.** Notice that $I = 4\int\tan^2\theta\sec\theta\,d\theta$, so the term $4\int\tan^2\theta\sec\theta\,d\theta$ on the right is just $I$ again. Add $I$ to both sides:

$$2I = 4\sec\theta\tan\theta - 4\ln|\sec\theta+\tan\theta|$$

**Step 5.** Divide both sides by 2 to isolate $I$:

$$I = 2\sec\theta\tan\theta - 2\ln|\sec\theta+\tan\theta| + C$$

**What went wrong in your working.**

You reached the same equation as Step 4, written as:

$$8\int\tan^2\theta\sec\theta\,d\theta = 4\sec\theta\tan\theta - 4\ln|\sec\theta+\tan\theta|$$

This is correct. But then you divided by 8 to get

$$\int\tan^2\theta\sec\theta\,d\theta = \tfrac{1}{2}\sec\theta\tan\theta - \tfrac{1}{2}\ln|\sec\theta+\tan\theta| + C$$

and presented this as the final answer. The problem: you solved for the *bare* integral $\int\tan^2\theta\sec\theta\,d\theta$ — which is $\frac{1}{4}$ of what you needed. The original integral is $4$ times that, so multiply back:

$$I = 4\times\left[\tfrac{1}{2}\sec\theta\tan\theta - \tfrac{1}{2}\ln|\sec\theta+\tan\theta|\right] = 2\sec\theta\tan\theta - 2\ln|\sec\theta+\tan\theta| + C$$

Equivalently, from $8\int\tan^2\theta\sec\theta\,d\theta = 4\sec\theta\tan\theta - 4\ln|\sec\theta+\tan\theta|$, divide by 2 (not 8) to read off $I = 4\int\tan^2\theta\sec\theta\,d\theta$ directly:

$$\underbrace{4\int\tan^2\theta\sec\theta\,d\theta}_{= I} = 2\sec\theta\tan\theta - 2\ln|\sec\theta+\tan\theta| + C \checkmark$$

The safest habit: at the start, write a labelled equation $I = \ldots$ for the original integral. At the end, make sure it is $I$ you are solving for, not a sub-piece of it.

### Issue 4 — Back-substitution (not attempted)

With $x = 2\tan\theta$, draw the reference triangle: opposite $= x$, adjacent $= 2$, hypotenuse $= \sqrt{x^2+4}$. Read off:

$$\tan\theta = \frac{x}{2}, \qquad \sec\theta = \frac{\sqrt{x^2+4}}{2}$$

Substituting into $I = 2\sec\theta\tan\theta - 2\ln|\sec\theta+\tan\theta| + C$:

$$= 2\cdot\frac{\sqrt{x^2+4}}{2}\cdot\frac{x}{2} - 2\ln\!\left|\frac{\sqrt{x^2+4}}{2}+\frac{x}{2}\right| + C$$

$$= \frac{x\sqrt{x^2+4}}{2} - 2\ln\!\left|\frac{x+\sqrt{x^2+4}}{2}\right| + C$$

The $\ln(1/2) = -\ln 2$ that comes out of splitting the log argument is a constant and absorbs into $C$, giving the standard final form:

$$\boxed{\frac{x\sqrt{x^2+4}}{2} - 2\ln\!\left|x + \sqrt{x^2+4}\right| + C}$$
